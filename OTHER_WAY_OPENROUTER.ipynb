```python
import requests
import json
import time
import os
import re


class BookGenerator:
    def __init__(self):
        self.base_url = "http://localhost:11434/api/generate"
        self.model = "gemma3:12b"
        self.story_premise = ""
        self.num_chapters = 0
        self.story_outline = ""
        self.chapters = []
        self.characters = {}  # Changed from list to dictionary for easier lookup
        self.chapter_summaries = {}  # Store detailed summaries of each chapter
        self.settings = {}  # Track settings/locations
        self.plot_events = []  # Track major plot events
        self.world_name = ""  # Consistent world name
        self.chapter_plan = ""  # Store the detailed chapter plan
        self.timeline = {}  # Track time progression between chapters
        self.emotional_arc = {}  # Track emotional tone in chapters
        self.transitions = {}  # Store generated transitions between chapters
        self.recurring_motifs = []  # Track recurring motifs or symbols for continuity

    def get_user_input(self):
        """Get the story premise and number of chapters from the user"""
        print("""
CAPABILITIES: Creates novels or fanfiction with consistency checks
GENERATION TIME: Up to an hour depending on computational resources
STORY INPUT: One paragraph up to 830 characters with your plot idea
"""
        )
        self.story_premise = input("Please provide a paragraph about what your story is about: ")

        while True:
            try:
                self.num_chapters = int(input("How many chapters would you like (minimum 3): "))
                if self.num_chapters >= 3:
                    break
                else:
                    print("Please enter at least 3 chapters.")
            except ValueError:
                print("Please enter a valid number.")

    def generate_text(self, prompt, system_prompt="You are a creative fiction writer."):
        """Make API call to Ollama with the given prompt"""
        data = {
            "model": self.model,
            "prompt": prompt,
            "system": system_prompt,
            "stream": False,
        }

        try:
            response = requests.post(self.base_url, json=data)
            response.raise_for_status()
            return response.json()["response"]
        except requests.exceptions.RequestException as e:
            print(f"Error making request to Ollama: {e}")
            return None

    def extract_characters(self, text):
        """Extract character information from text and create structured data"""
        characters = {}
        # Look for patterns like "CHARACTER NAME: description"
        pattern = r"([A-Z][A-Za-z\s]+):\s+([^\n]+)"
        matches = re.findall(pattern, text)

        for match in matches:
            name = match[0].strip()
            description = match[1].strip()
            characters[name] = {
                "name": name,
                "description": description,
                "first_appearance": 0,
                "status": "alive",
                "development": [],
                "relationships": {},
            }

        return characters

    def extract_world_name(self, outline):
        """Extract consistent world name from the story outline"""
        # Generic pattern to find world names without hardcoding specific ones
        patterns = [
            r"Neo-[A-Za-z]+",
            r"[A-Z][a-z]+land",
            r"[A-Z][a-z]+ Kingdom",
            r"[A-Z][a-z]+ Empire",
            r"[A-Z][a-z]+ Realm",
            r"[A-Z][a-z]+ World",
            r"[A-Z][a-z]+ City"
        ]

        for pattern in patterns:
            matches = re.findall(pattern, outline)
            if matches:
                # Use the most common name found
                counts = {}
                for match in matches:
                    counts[match] = counts.get(match, 0) + 1
                return max(counts, key=counts.get)

        # If no specific world name is found, let the LLM generate one in later steps
        return ""

    def create_story_outline(self):
        """Generate a high-level outline for the entire story with improved structure"""
        system_prompt = """You are a professional novelist and editor.
        You excel at creating compelling story outlines that maintain coherence and proper narrative structure.
        Pay special attention to character development, consistent world-building, and logical plot progression."""
        prompt = f"""Based on the following story premise, create a detailed story outline for a {self.num_chapters}-chapter book.

Story premise: {self.story_premise}

For each chapter, provide:
1. A chapter title
2. Key plot events (3-4 bullet points with specific details)
3. Character development points (which characters appear and how they develop)
4. Setting/location details (be specific and consistent)

Also create a section titled "WORLD BUILDING" with:
1. The name of the main city/setting (create a unique, memorable name)
2. Key locations that will appear multiple times
3. Important technology or cultural elements

Then create a section titled "CHARACTERS" with:
1. Main protagonist (name, detailed description, motivation, arc)
2. Main antagonist (name, detailed description, motivation)
3. Supporting characters (name, role, connection to protagonist)

Also create a section titled "RECURRING MOTIFS" with 3-5 symbols, objects, or phrases that will recur throughout the story to provide thematic continuity.

The outline should have a clear beginning (chapters 1-2), middle (chapters 3 to {self.num_chapters-2}), and end (final {min(2, self.num_chapters-2)} chapters).
Ensure proper rising action, climax, and resolution structure.

For the climax chapters (chapters {self.num_chapters-2} and {self.num_chapters-1}), provide extra detail about:
1. How the final confrontation unfolds
2. What's at stake for each character
3. Specific steps in the resolution

Make sure all character names, settings, and plot elements remain 100% consistent throughout.
"""
        print("Generating detailed story outline...")
        self.story_outline = self.generate_text(prompt, system_prompt)

        # Extract character information
        char_prompt = f"""Based on this story outline, create a detailed character guide:

{self.story_outline}

For EACH character, format as:
CHARACTER NAME: Brief description, role in story, key personality traits, motivation, background

Include EVERY character mentioned in the outline, even minor ones.
"""
        print("Creating detailed character profiles...")
        character_text = self.generate_text(char_prompt, system_prompt)
        self.characters = self.extract_characters(character_text)

        # Extract world name for consistency
        self.world_name = self.extract_world_name(self.story_outline)

        # If no world name was found, ask the LLM to create one
        if not self.world_name:
            world_prompt = f"""Based on this story outline, create a unique and memorable name for the main world/city/setting:

{self.story_outline}

The name should be a single term, creative, and fitting the tone of the story.
Reply with ONLY the world name, nothing else.
"""
            generated_world_name = self.generate_text(world_prompt)
            if generated_world_name:
                 self.world_name = generated_world_name.strip()
            print(f"Generated world name: {self.world_name}")

        # Extract recurring motifs
        motif_prompt = f"""Based on this story outline, identify 3-5 recurring motifs, symbols, or objects that appear throughout the story:

{self.story_outline}

Format as a simple list of items, one per line.
These should be concrete objects, symbols, or phrases that can recur throughout chapters.
"""
        print("Identifying recurring motifs...")
        motifs_text = self.generate_text(motif_prompt)
        if motifs_text:
            self.recurring_motifs = [motif.strip() for motif in motifs_text.strip().split('\n') if motif.strip()]
        print(f"Identified motifs: {', '.join(self.recurring_motifs)}")

        # Create detailed chapter-by-chapter plan
        chapter_plan_prompt = f"""Based on the story outline, create a VERY detailed chapter-by-chapter plan.

STORY OUTLINE: {self.story_outline}

For EACH chapter (1 through {self.num_chapters}), provide:
1. Chapter title
2. Chapter summary (250-300 words)
3. Scene breakdown (list each scene with location and characters present)
4. Character development in this chapter
5. Plot advancement in this chapter
6. Timeline indicators (time of day, date, or how much time has passed since previous chapter)
7. Emotional tone and tension level (1-10) at the end of the chapter
8. How this chapter connects to the next (create a narrative bridge)

Be extremely specific and detailed. This plan will be used to ensure narrative consistency.
"""
        print("Creating detailed chapter plan...")
        self.chapter_plan = self.generate_text(chapter_plan_prompt, system_prompt)

    def create_chapter_summary(self, chapter_num, chapter_content):
        """Create a detailed summary of a chapter after it's written"""
        system_prompt = """You are a literary analyst specializing in narrative structure and continuity.
Create comprehensive, detailed summaries that capture all key elements."""
        prompt = f"""Create a detailed summary of the following chapter content.
This is Chapter {chapter_num} of a {self.num_chapters}-chapter book.

Include in your summary:
1. All key plot developments
2. Character appearances and development
3. Setting details
4. Important dialogue or revelations
5. How this chapter connects to previous chapters
6. Emotional tone at the beginning and end of the chapter

CHAPTER CONTENT:
{chapter_content}

Your summary should be comprehensive enough that another writer could use it to maintain perfect continuity.
"""
        summary = self.generate_text(prompt, system_prompt)
        self.chapter_summaries[chapter_num] = summary
        return summary

    def update_character_tracking(self, chapter_num, chapter_content):
        """Update character tracking data based on a chapter's content"""
        system_prompt = """You are a narrative continuity expert who specializes in tracking character development.
Extract precise information about characters from text."""
        character_names = list(self.characters.keys())
        characters_str = ", ".join(character_names)
        prompt = f"""Based on the following chapter content, track the development of all characters mentioned.

CHAPTER CONTENT:
{chapter_content}

CHARACTERS TO TRACK: {characters_str}

For each character that appears in this chapter, provide:
1. Current status (alive, dead, injured, etc.)
2. Development in this chapter
3. New relationships formed
4. Current location
5. Emotional state at the end of the chapter

Format as:
CHARACTER NAME: status|development|relationships|location|emotional_state

Only include characters who actually appear or are mentioned in this chapter.
"""
        character_updates = self.generate_text(prompt, system_prompt)

        if not character_updates:
            print(f"Warning: No character updates received for Chapter {chapter_num}.")
            return

        # Parse and update character data
        update_pattern = r"([A-Z][A-Za-z\s]+):\s+([^|]+)\|([^|]+)\|([^|]+)\|([^|]+)\|([^\n]+)"
        matches = re.findall(update_pattern, character_updates)

        for match in matches:
            name = match[0].strip()
            if name in self.characters:
                self.characters[name]["status"] = match[1].strip()
                self.characters[name]["development"].append({
                    "chapter": chapter_num,
                    "development": match[2].strip()
                })
                # Update relationship data
                new_relationships = match[3].strip()
                if new_relationships:
                    for other_char in character_names:
                        if other_char != name and other_char in new_relationships:
                            self.characters[name]["relationships"][other_char] = chapter_num

                # Update location and emotional state
                self.characters[name]["location"] = match[4].strip()
                self.characters[name]["emotional_state"] = match[5].strip()

                # Record first appearance if not already set
                if self.characters[name]["first_appearance"] == 0:
                    self.characters[name]["first_appearance"] = chapter_num

    def update_timeline(self, chapter_num, chapter_content):
        """Extract and update timeline information for chapter"""
        system_prompt = """You are a literary analyst specializing in temporal structure in narratives."""

        prompt = f"""Based on the following chapter content, determine:
        1. How much time has passed in this chapter
        2. What time of day/date it ends on
        3. Any specific time markers mentioned

        CHAPTER CONTENT:
        {chapter_content}

        Reply only with the time information in this format:
        TIME_ELAPSED: [time passed during chapter]
        END_TIME: [time of day/date at chapter end]
        TIME_MARKERS: [any specific times mentioned]
        """

        time_info = self.generate_text(prompt, system_prompt)
        self.timeline[chapter_num] = time_info
        return time_info

    def track_emotional_arc(self, chapter_num, chapter_content):
        """Track emotional tone and tension at the end of the chapter"""
        system_prompt = """You are a literary analyst specializing in emotional arcs in storytelling."""

        prompt = f"""Analyze the emotional tone at the end of this chapter:

        CHAPTER CONTENT:
        {chapter_content}

        Determine:
        1. The primary emotion at the chapter's end (tension, relief, hope, despair, etc.)
        2. The level of tension (1-10 scale)
        3. The primary unresolved question or conflict

        Reply in this format:
        EMOTION: [primary emotion]
        TENSION: [level 1-10]
        UNRESOLVED: [main unresolved question]
        """

        emotional_status = self.generate_text(prompt, system_prompt)
        self.emotional_arc[chapter_num] = emotional_status
        return emotional_status

    def create_chapter_transition(self, chapter_num, chapter_content):
        """Create a transition from current chapter to the next"""
        if chapter_num >= self.num_chapters:
            return ""  # No transition needed for the last chapter

        system_prompt = """You are a master storyteller specializing in creating suspenseful
        chapter endings and smooth transitions between chapters."""

        # Extract relevant part of chapter plan for the next chapter
        chapter_plan_prompt = f"""From this detailed chapter plan, extract ONLY the plan
        for Chapter {chapter_num + 1}:

        {self.chapter_plan}

        Include ONLY Chapter {chapter_num + 1}'s detailed plan.
        """
        next_chapter_plan = self.generate_text(chapter_plan_prompt)

        # Get emotional status
        emotional_status = self.emotional_arc.get(chapter_num, "")

        # Get timeline info
        timeline_info = self.timeline.get(chapter_num, "")

        # Choose a recurring motif to include
        if self.recurring_motifs:
            chosen_motif = self.recurring_motifs[chapter_num % len(self.recurring_motifs)]
        else:
            chosen_motif = ""

        prompt = f"""Create a compelling transition paragraph or two to serve as the ending of Chapter {chapter_num}
        and create anticipation for Chapter {chapter_num + 1}.

        CURRENT CHAPTER CONTENT (last 1000 characters):
        {chapter_content[-1000:]}

        NEXT CHAPTER PLAN:
        {next_chapter_plan}

        EMOTIONAL STATE AT CHAPTER END:
        {emotional_status}

        TIMELINE INFORMATION:
        {timeline_info}

        RECURRING MOTIF TO INCLUDE IF POSSIBLE:
        {chosen_motif}

        The transition should:
        1. Conclude the current chapter with a hook, an unanswered question, or tension
        2. Create anticipation for what comes next
        3. Avoid completely resolving all tension from the current chapter
        4. Link to the theme or events of the upcoming chapter without being too explicit
        5. Maintain the emotional tone but hint at potential changes in the next chapter
        6. Include a subtle reference to the recurring motif if possible

        Create only 1-2 paragraphs for this transition. These will be the FINAL paragraphs of the current chapter.
        """

        transition = self.generate_text(prompt, system_prompt)
        self.transitions[chapter_num] = transition
        return transition

    def create_next_chapter_opener(self, chapter_num):
        """Create a strong opening for the next chapter that connects to the previous one"""
        if chapter_num <= 1:
            return ""  # First chapter doesn't need a special opener

        system_prompt = """You are a master storyteller specializing in creating
        engaging chapter openings that connect smoothly to previous events."""

        # Get information about the previous chapter
        prev_chapter_summary = self.chapter_summaries.get(chapter_num - 1, "")
        prev_transition = self.transitions.get(chapter_num - 1, "")
        prev_emotional_status = self.emotional_arc.get(chapter_num - 1, "")
        prev_timeline = self.timeline.get(chapter_num - 1, "")

        # Extract timeline info to know how much time has passed
        timeline_pattern = r"END_TIME:\s*([^\n]+)"
        prev_end_time = ""
        if prev_timeline:
            time_match = re.search(timeline_pattern, prev_timeline)
            if time_match:
                prev_end_time = time_match.group(1).strip()

        # Extract relevant part of chapter plan for this chapter
        chapter_plan_prompt = f"""From this detailed chapter plan, extract ONLY the plan
        for Chapter {chapter_num}:

        {self.chapter_plan}

        Include ONLY Chapter {chapter_num}'s detailed plan.
        """
        this_chapter_plan = self.generate_text(chapter_plan_prompt)

        prompt = f"""Create a compelling opening paragraph for Chapter {chapter_num} that connects
        seamlessly with the end of Chapter {chapter_num - 1}.

        PREVIOUS CHAPTER SUMMARY:
        {prev_chapter_summary}

        END OF PREVIOUS CHAPTER:
        {prev_transition}

        EMOTIONAL STATE AT END OF PREVIOUS CHAPTER:
        {prev_emotional_status}

        TIME AT END OF PREVIOUS CHAPTER:
        {prev_end_time}

        THIS CHAPTER'S PLAN:
        {this_chapter_plan}

        The opening paragraph should:
        1. Establish a clear time connection to the end of the previous chapter
           (e.g., "The next morning...", "Three days later...", "As the sun set on that fateful day...")
        2. Provide a smooth transition that builds on the tension or questions from the previous chapter
        3. Orient readers immediately in the scene (location, characters present)
        4. Avoid summarizing or repeating information from the previous chapter
        5. Set the emotional tone for this chapter

        Create a single strong opening paragraph (3-5 sentences).
        """

        opener = self.generate_text(prompt, system_prompt)
        return opener

    def validate_chapter_consistency(self, chapter_num, chapter_content):
        """Check chapter for consistency issues"""
        system_prompt = """You are a literary editor specializing in narrative consistency.
Your job is to identify and flag any inconsistencies in a narrative."""
        # Create context for consistency check
        previous_summaries = ""
        for i in range(1, chapter_num):
            if i in self.chapter_summaries:
                previous_summaries += f"Chapter {i} Summary: {self.chapter_summaries[i]}\n\n"

        character_status_list = []
        for name, data in self.characters.items():
            if data["first_appearance"] > 0:  # Only include characters who have appeared
                status = data["status"]
                first_app = data["first_appearance"]
                character_status_list.append(f"{name}: First appeared in Chapter {first_app}, Status: {status}")
        character_status_str = "\n".join(character_status_list)


        # Add timeline information
        timeline_info = ""
        for i in range(1, chapter_num):
            if i in self.timeline:
                timeline_info += f"Chapter {i} Timeline: {self.timeline[i]}\n"

        prompt = f"""Analyze this chapter for consistency issues compared to previous chapters.

STORY PREMISE: {self.story_premise}

WORLD NAME: {self.world_name}

PREVIOUS CHAPTERS:
{previous_summaries}

CHARACTER STATUS:
{character_status_str}

TIMELINE INFORMATION:
{timeline_info}

CURRENT CHAPTER {chapter_num} CONTENT:
{chapter_content}

Identify ANY inconsistencies related to:
1. Character names or backgrounds
2. Setting/location names
3. Timeline/sequence of events
4. Plot contradictions
5. Sudden introduction of new characters without proper context
6. Time of day or elapsed time inconsistencies

If any inconsistencies are found, list them in order of severity.
If no inconsistencies are found, respond with "CONSISTENT".
"""
        consistency_check = self.generate_text(prompt, system_prompt)
        return consistency_check

    def fix_chapter_inconsistencies(self, chapter_num, chapter_content, issues):
        """Fix identified consistency issues in a chapter"""
        system_prompt = """You are a professional novelist and editor who excels at maintaining narrative consistency.
Fix all inconsistencies while preserving the core narrative."""
        # Create context for the fix
        previous_summaries = ""
        for i in range(1, chapter_num):
            if i in self.chapter_summaries:
                previous_summaries += f"Chapter {i} Summary: {self.chapter_summaries[i]}\n\n"

        character_status_list = []
        for name, data in self.characters.items():
            if data["first_appearance"] > 0:  # Only include characters who have appeared
                status = data["status"]
                first_app = data["first_appearance"]
                dev_list = []
                if "development" in data and data["development"]:
                    dev_list = [d.get("development", "N/A") for d in data["development"] if isinstance(d, dict)]
                dev = "; ".join(dev_list)
                character_status_list.append(f"{name}: First appeared in Chapter {first_app}, Status: {status}, Development: {dev}")
        character_status_str = "\n".join(character_status_list)


        # Add timeline information
        timeline_info = ""
        for i in range(1, chapter_num):
            if i in self.timeline:
                timeline_info += f"Chapter {i} Timeline: {self.timeline[i]}\n"

        prompt = f"""Rewrite this chapter to fix all the identified consistency issues while maintaining the same overall plot and character development.

STORY PREMISE: {self.story_premise}

WORLD NAME: {self.world_name} (use this name consistently for the city/world)

PREVIOUS CHAPTERS:
{previous_summaries}

CHARACTER STATUS:
{character_status_str}

TIMELINE INFORMATION:
{timeline_info}

CURRENT CHAPTER CONTENT:
{chapter_content}

CONSISTENCY ISSUES TO FIX:
{issues}

Guidelines for fixing:
1. Maintain the same overall plot progression and key events
2. Ensure all character names, backgrounds, and statuses match previous chapters
3. Use the established world name ({self.world_name}) consistently
4. Provide proper context for any character that hadn't appeared before
5. Ensure timeline and causality make sense
6. Keep appropriate transitions between scenes within the chapter

Rewrite the complete chapter while fixing all issues.
"""
        fixed_chapter = self.generate_text(prompt, system_prompt)
        return fixed_chapter

    def generate_chapter(self, chapter_num):
        """Generate a single chapter with enhanced context awareness and consistency checks"""
        system_prompt = """You are a celebrated novelist known for writing engaging, coherent chapters
with natural flow and character development. Your chapters have clear narrative structure and
maintain perfect consistency with previously established elements."""
        # Build context from previous chapters
        context_summaries = []
        for i in range(1, chapter_num):
            if i in self.chapter_summaries:
                context_summaries.append(f"Chapter {i} Summary: {self.chapter_summaries[i]}")
        context = "\n\n".join(context_summaries)

        # Create character context
        character_context_list = []
        for name, data in self.characters.items():
            if data["first_appearance"] > 0 and data["first_appearance"] < chapter_num:
                # Only include characters who have already appeared
                status = data["status"]
                desc = data["description"]
                character_context_list.append(f"{name}: {desc}, Status: {status}")
        characters_in_chapter_str = "\n".join(character_context_list)


        # Add timeline information
        timeline_context = ""
        for i in range(1, chapter_num):
            if i in self.timeline:
                timeline_context += f"Chapter {i} Timeline: {self.timeline[i]}\n"

        # Add emotional arc information
        emotional_context = ""
        if chapter_num > 1 and (chapter_num - 1) in self.emotional_arc:
            emotional_context = f"End of previous chapter emotional state: {self.emotional_arc[chapter_num - 1]}"

        # Create chapter opener for chapters after the first
        chapter_opener = ""
        if chapter_num > 1:
            chapter_opener = self.create_next_chapter_opener(chapter_num)

        # Extract relevant part of chapter plan for this chapter
        chapter_plan_prompt = f"""From this detailed chapter plan, extract ONLY the plan for Chapter {chapter_num}:

{self.chapter_plan}

Include ONLY Chapter {chapter_num}'s detailed plan.
"""
        this_chapter_plan = self.generate_text(chapter_plan_prompt)

        # Choose a recurring motif to include
        if self.recurring_motifs:
            chosen_motif = self.recurring_motifs[chapter_num % len(self.recurring_motifs)]
            motif_instruction = f"Include the recurring motif '{chosen_motif}' somewhere in this chapter in a natural way."
        else:
            motif_instruction = ""

        prompt = f"""Write Chapter {chapter_num} of a novel based on the following guidelines:

STORY PREMISE: {self.story_premise}

WORLD NAME: {self.world_name} (use this name consistently throughout)

OVERALL STORY OUTLINE: {self.story_outline}

THIS CHAPTER'S DETAILED PLAN:
{this_chapter_plan}

CHARACTERS IN THIS STORY SO FAR:
{characters_in_chapter_str}

PREVIOUS CHAPTERS SUMMARY:
{context}

TIMELINE INFORMATION:
{timeline_context}

EMOTIONAL CONTEXT:
{emotional_context}

SUGGESTED CHAPTER OPENING:
{chapter_opener}

{motif_instruction}

GUIDELINES FOR THIS CHAPTER:
- Write a complete chapter with a clear beginning, middle, and end
- Each chapter should be approximately 2500-3000 words
- Include descriptive elements, dialogue, and character development
- Maintain consistent tone, voice, and character personalities
- This is chapter {chapter_num} of {self.num_chapters}, so structure it accordingly
- Do NOT repeat world-building elements that have been established in previous chapters
- Each chapter should advance the plot while having its own mini-arc
- Include chapter title at the beginning (e.g., "## Chapter {chapter_num}: [Your Chapter Title Here]")
- ONLY introduce new characters if they are essential and provide proper context for them
- Refer to the world/city consistently as {self.world_name}
- Begin by directly continuing from the previous chapter's events or revealing their consequences
- End the chapter with a compelling hook or unanswered question
- The last paragraph should create tension that leads into the next chapter

CHAPTER STRUCTURE GUIDELINES:
- If this is not the first chapter, begin by using the suggested chapter opening to create a smooth transition
- Don't repeat the world description that was given in previous chapters
- End the chapter with a compelling hook or unanswered question
- The last paragraph should create tension that leads into the next chapter

Format the chapter with proper paragraph structure and dialogue formatting. Start with the chapter title.
The chapter title should be formatted as a Markdown H2 header, like this: "## Chapter {chapter_num}: [Your Chapter Title Here]".
Make sure the LLM generates this title as the first line of the chapter content.
"""
        print(f"Generating Chapter {chapter_num}...")
        chapter_content = self.generate_text(prompt, system_prompt)

        if not chapter_content:
            print(f"Error: Failed to generate content for Chapter {chapter_num}.")
            return f"## Chapter {chapter_num}: Generation Failed\n\n[Content generation failed for this chapter]"


        # Check for consistency issues
        print(f"Validating Chapter {chapter_num} for consistency...")
        consistency_check = self.validate_chapter_consistency(chapter_num, chapter_content)

        # If issues found, fix them
        if consistency_check and "CONSISTENT" not in consistency_check.upper():
            print(f"Consistency issues found in Chapter {chapter_num}. Issues:\n{consistency_check}\nFixing...")
            fixed_content = self.fix_chapter_inconsistencies(chapter_num, chapter_content, consistency_check)
            if fixed_content:
                chapter_content = fixed_content
                print(f"Chapter {chapter_num} fixed for consistency.")
            else:
                print(f"Error: Failed to fix consistency issues for Chapter {chapter_num}. Using original content.")
        else:
            print(f"Chapter {chapter_num} is consistent with previous narrative.")

        # Create summary and update character tracking
        print(f"Creating detailed summary for Chapter {chapter_num}...")
        self.create_chapter_summary(chapter_num, chapter_content)

        print(f"Updating character tracking for Chapter {chapter_num}...")
        self.update_character_tracking(chapter_num, chapter_content)

        print(f"Updating timeline information for Chapter {chapter_num}...")
        self.update_timeline(chapter_num, chapter_content)

        print(f"Analyzing emotional arc for Chapter {chapter_num}...")
        self.track_emotional_arc(chapter_num, chapter_content)

        # Add transition if not the last chapter
        if chapter_num < self.num_chapters:
            print(f"Creating transitional ending for Chapter {chapter_num}...")
            transition = self.create_chapter_transition(chapter_num, chapter_content)
            if transition:
                # Extract the last paragraph
                paragraphs = chapter_content.strip().split('\n\n')
                # Replace the last paragraph with our transition
                if len(paragraphs) > 1:
                    chapter_content = '\n\n'.join(paragraphs[:-1]) + '\n\n' + transition.strip()
                else:
                    chapter_content = chapter_content.strip() + '\n\n' + transition.strip()
            else:
                print(f"Warning: Failed to generate transition for Chapter {chapter_num}.")


        return chapter_content

    def check_chapter_transitions(self):
        """Check and improve transitions between all chapters after generation"""
        print("Performing final check on chapter transitions...")
        improved_chapters = []

        for i in range(len(self.chapters)):
            current_chapter_text = self.chapters[i]
            if i == 0:
                improved_chapters.append(current_chapter_text)
                continue

            # Get current and previous chapters
            prev_chapter_text = self.chapters[i-1]


            system_prompt = """You are a professional editor specializing in narrative flow and chapter transitions."""

            prompt = f"""Analyze the transition between these consecutive chapters and improve it if needed:

            END OF PREVIOUS CHAPTER (last ~200 words):
            {' '.join(prev_chapter_text.strip().split()[-200:])}

            BEGINNING OF CURRENT CHAPTER (first ~200 words):
            {' '.join(current_chapter_text.strip().split()[:200])}

            If the transition is already smooth and natural, respond with "TRANSITION: SMOOTH".

            Otherwise, provide an improved beginning for the current chapter (first 1-2 paragraphs, approximately 100-150 words) that:
            1. Creates a smoother connection with the previous chapter's ending.
            2. Avoids repeating information already established.
            3. Maintains character and plot consistency.
            4. Progresses the timeline naturally, if applicable.
            5. Sets the tone for the current chapter effectively.

            Start your response with "TRANSITION: REVISED" followed by ONLY the revised opening paragraph(s) for the current chapter.
            Do not include the chapter title in your revised opening.
            """

            transition_check_response = self.generate_text(prompt, system_prompt)

            if transition_check_response and "TRANSITION: REVISED" in transition_check_response:
                # Extract and apply the revised beginning
                try:
                    revised_beginning_text = transition_check_response.split("TRANSITION: REVISED")[1].strip()
                    if revised_beginning_text:
                        # Find the chapter title (e.g., "## Chapter X: Title")
                        title_match = re.match(r"(##\s*Chapter\s*\d+[:\s\w]+\n*)", current_chapter_text)
                        if title_match:
                            chapter_title = title_match.group(1)
                            # Get content after the title
                            content_after_title = current_chapter_text[len(chapter_title):].strip()
                            # Split the content after title into paragraphs to replace the first few
                            paragraphs_after_title = content_after_title.split('\n\n')
                            # Assuming revised_beginning_text replaces 1-2 paragraphs.
                            # This is a heuristic; might need refinement based on LLM output.
                            num_paras_to_replace = 1
                            if '\n\n' in revised_beginning_text: # If LLM gives 2 paragraphs
                                num_paras_to_replace = 2

                            if len(paragraphs_after_title) >= num_paras_to_replace :
                                remaining_content = '\n\n'.join(paragraphs_after_title[num_paras_to_replace:])
                                improved_chapter_text = chapter_title.strip() + '\n\n' + revised_beginning_text + '\n\n' + remaining_content
                                improved_chapters.append(improved_chapter_text.strip())
                                print(f"Transition improved for Chapter {i+1}.")
                            else: # Not enough paragraphs to replace, just prepend. Unlikely.
                                improved_chapters.append(chapter_title.strip() + '\n\n' + revised_beginning_text + '\n\n' + content_after_title)
                        else: # No clear chapter title found, prepend revised beginning (less ideal)
                            print(f"Warning: Chapter title not found in standard format for Chapter {i+1}. Prepending revised opening.")
                            improved_chapters.append(revised_beginning_text + '\n\n' + current_chapter_text.strip())
                    else:
                        print(f"Transition for Chapter {i+1} marked for revision, but no revised text provided.")
                        improved_chapters.append(current_chapter_text)
                except IndexError:
                    print(f"Error parsing revised transition for Chapter {i+1}. Using original.")
                    improved_chapters.append(current_chapter_text)
            else:
                print(f"Transition for Chapter {i+1} is smooth or revision failed.")
                improved_chapters.append(current_chapter_text)

        self.chapters = improved_chapters
        print("Chapter transitions have been optimized.")

    def generate_book(self):
        """Generate the complete book with enhanced consistency checks"""
        self.get_user_input()
        self.create_story_outline()

        if not self.story_outline or not self.chapter_plan:
            print("Critical error: Story outline or chapter plan generation failed. Aborting book generation.")
            return "Book generation failed due to missing outline or plan."

        for i in range(1, self.num_chapters + 1):
            print(f"\n--- Generating Chapter {i} of {self.num_chapters} ---")
            chapter = self.generate_chapter(i)
            if chapter:
                self.chapters.append(chapter)
            else:
                print(f"Critical error: Generation of Chapter {i} failed. Aborting book generation.")
                # Optionally, you could append a placeholder or stop entirely.
                # For now, let's add a placeholder and continue, but this indicates a problem.
                self.chapters.append(f"## Chapter {i}: Content Generation Failed\n\n[This chapter could not be generated.]")


            # Add a delay to prevent overwhelming the API, if needed
            if i < self.num_chapters:
                print("Pausing briefly before generating next chapter...")
                time.sleep(3) # Adjust delay as necessary

        # Perform final check on transitions between chapters
        if self.chapters: # Only check transitions if chapters were generated
            self.check_chapter_transitions()

        return self.compile_book()

    def compile_book(self):
        """Compile all chapters into a complete book, containing only title and chapters."""
        title_prompt = f"""Create a compelling title for a book with the following premise and outline:

Premise: {self.story_premise}

Story Outline Snippet (first few chapter ideas):
{self.story_outline[:1000]}

Respond with ONLY the book title, nothing else.
"""
        book_title = self.generate_text(title_prompt)

        if not book_title:
            book_title = "Untitled Book" # Fallback title
        else:
            book_title = book_title.strip().split('\n')[0] # Take first line
            if book_title.lower().startswith("title:"):
                book_title = book_title[len("title:"):].strip()
            if not book_title: # If stripping "title:" leaves it empty
                book_title = "Untitled Book"


        book = f"# {book_title}\n\n"  # Start with the book title, H1 Markdown

        # Iterate through the chapters and append them directly
        for i, chapter_content in enumerate(self.chapters):
            # Ensure each chapter starts on a new line and has a blank line after it.
            # The chapter_content should ideally contain the chapter title
            # (e.g., "## Chapter 1: The Beginning") as generated by `generate_chapter`.
            # If not, it will just be the chapter text.
            book += f"{chapter_content.strip()}\n\n"

        return book.strip() # Use strip() at the end to remove any trailing newlines


    def save_book(self, book_content, filename="generated_book.md"):
        """Save the generated book to a file"""
        if not book_content or "Book generation failed" in book_content:
            print(f"Book content is missing or incomplete. Not saving to {filename}.")
            return

        try:
            with open(filename, "w", encoding="utf-8") as f:
                f.write(book_content)
            print(f"Book saved as {filename}")
        except IOError as e:
            print(f"Error saving book to {filename}: {e}")
            return


        # Also save metadata for future reference
        metadata = {
            "story_premise": self.story_premise,
            "num_chapters_requested": self.num_chapters,
            "num_chapters_generated": len(self.chapters),
            "world_name": self.world_name,
            "characters": self.characters,
            "chapter_summaries": self.chapter_summaries,
            "recurring_motifs": self.recurring_motifs,
            "timeline": self.timeline,
            "emotional_arc": self.emotional_arc,
            "story_outline_snippet": self.story_outline[:2000] if self.story_outline else "N/A", # Save a snippet
            "chapter_plan_snippet": self.chapter_plan[:2000] if self.chapter_plan else "N/A" # Save a snippet
        }

        try:
            with open("book_metadata.json", "w", encoding="utf-8") as f:
                json.dump(metadata, f, indent=4) # Indent for readability
            print("Book metadata saved as book_metadata.json")
        except IOError as e:
            print(f"Error saving metadata to book_metadata.json: {e}")
        except TypeError as e:
            print(f"Error serializing metadata to JSON: {e}. Some data might not be serializable.")


if __name__ == "__main__":
    generator = BookGenerator()
    book_text_content = generator.generate_book()
    if book_text_content and "Book generation failed" not in book_text_content:
        generator.save_book(book_text_content)
    else:
        print("Book generation was unsuccessful. No file saved.")
```

```python
import requests
import json
import time
import os
import re


class BookGenerator:
    def __init__(self):
        self.base_url = "http://localhost:11434/api/generate"
        self.model = "gemma3:12b" # Consider using a model suited for creative writing if available
        self.story_premise = ""
        self.genre = "" # Added genre attribute
        self.num_chapters = 0
        self.story_outline = ""
        self.chapters = []
        self.characters = {}  # Changed from list to dictionary for easier lookup
        self.chapter_summaries = {}  # Store detailed summaries of each chapter
        self.settings = {}  # Track settings/locations
        self.plot_events = []  # Track major plot events
        self.world_name = ""  # Consistent world name
        self.chapter_plan = ""  # Store the detailed chapter plan
        self.timeline = {}  # Track time progression between chapters
        self.emotional_arc = {}  # Track emotional tone in chapters
        self.transitions = {}  # Store generated transitions between chapters
        self.recurring_motifs = []  # Track recurring motifs or symbols for continuity

    def get_user_input(self):
        """Get the story premise, genre, and number of chapters from the user"""
        print("""
CAPABILITIES: Creates novels or fanfiction with consistency checks
GENERATION TIME: Up to an hour depending on computational resources
STORY INPUT: One paragraph up to 830 characters with your plot idea
"""
        )
        self.story_premise = input("Please provide a paragraph about what your story is about: ")
        # Get genre input
        self.genre = input("Please specify the genre for your story (e.g., Sci-Fi, Fantasy, Mystery): ").strip()
        if not self.genre: # Basic fallback if user enters nothing
            self.genre = "General Fiction"
            print("No genre specified, defaulting to General Fiction.")

        while True:
            try:
                self.num_chapters = int(input("How many chapters would you like (minimum 3): "))
                if self.num_chapters >= 3:
                    break
                else:
                    print("Please enter at least 3 chapters.")
            except ValueError:
                print("Please enter a valid number.")

    def generate_text(self, prompt, system_prompt="You are a creative fiction writer."):
        """Make API call to Ollama with the given prompt"""
        data = {
            "model": self.model,
            "prompt": prompt,
            "system": system_prompt,
            "stream": False,
        }

        try:
            response = requests.post(self.base_url, json=data)
            response.raise_for_status()
            # Basic check for empty or error response from the model itself
            response_data = response.json()
            if "response" not in response_data or not response_data["response"]:
                 print(f"Warning: Received empty response from model for prompt:\n---\n{prompt[:200]}...\n---")
                 return None # Return None for empty response
            return response_data["response"]
        except requests.exceptions.RequestException as e:
            print(f"Error making request to Ollama: {e}")
            return None
        except json.JSONDecodeError as e:
            print(f"Error decoding JSON response from Ollama: {e}")
            print(f"Response text: {response.text}")
            return None


    def extract_characters(self, text):
        """Extract character information from text and create structured data"""
        if not text: # Handle case where text is None or empty
            return {}
        characters = {}
        # Look for patterns like "CHARACTER NAME: description"
        # Made pattern more robust to handle variations in spacing and line breaks
        pattern = r"([A-Z][A-Za-z\s'-]+):\s+([\s\S]*?)(?=\n[A-Z][A-Za-z\s'-]+:|\Z)"
        matches = re.findall(pattern, text)

        for match in matches:
            name = match[0].strip()
            description = match[1].strip()
            if name and description: # Ensure name and description are not empty
                characters[name] = {
                    "name": name,
                    "description": description,
                    "first_appearance": 0, # Will be updated when character first appears
                    "status": "unknown", # Initial status
                    "development": [],
                    "relationships": {},
                    "location": "unknown",
                    "emotional_state": "unknown"
                }
            else:
                print(f"Warning: Skipped potential character entry due to missing name or description: {match}")


        # Fallback if the primary pattern fails - look for capitalized names at start of lines
        if not characters:
             print("Primary character extraction pattern failed. Trying fallback.")
             # Simple fallback: look for capitalized words followed by a colon at the start of a line
             fallback_pattern = r"^\s*([A-Z][A-Za-z\s'-]+):\s*(.*)"
             lines = text.split('\n')
             for line in lines:
                 match = re.match(fallback_pattern, line)
                 if match:
                     name = match.group(1).strip()
                     description = match.group(2).strip()
                     if name and description and name not in characters: # Avoid duplicates
                         characters[name] = {
                             "name": name,
                             "description": description,
                             "first_appearance": 0,
                             "status": "unknown",
                             "development": [],
                             "relationships": {},
                             "location": "unknown",
                             "emotional_state": "unknown"
                         }

        if not characters:
            print("Warning: No characters extracted. Check the format of the character text.")

        return characters

    def extract_world_name(self, outline):
        """Extract consistent world name from the story outline"""
        if not outline: return ""
        # Generic pattern to find world names without hardcoding specific ones
        # Added more patterns and made them case-insensitive where appropriate
        patterns = [
            r"Neo-[A-Za-z]+",
            r"[A-Z][a-z]+land", r"(?i)\bthe\s+Kingdom\s+of\s+([A-Z][a-z]+)\b",
            r"(?i)\bthe\s+([A-Z][a-z]+)\s+Kingdom\b",
            r"(?i)\bthe\s+([A-Z][a-z]+)\s+Empire\b",
            r"(?i)\bthe\s+Realm\s+of\s+([A-Z][a-z]+)\b",
            r"(?i)\bthe\s+([A-Z][a-z]+)\s+Realm\b",
            r"(?i)\b([A-Z][a-z]+)\s+World\b",
            r"(?i)\b([A-Z][a-z]+)\s+City\b",
            r"(?i)\bCity\s+of\s+([A-Z][a-z]+)\b",
            r"(?i)\bPlanet\s+([A-Z][a-z]+)\b", # Added Planet
            r"(?i)\bthe\s+([A-Z][a-z]+)\s+Federation\b", # Added Federation
            r"(?i)\bthe\s+([A-Z][a-z]+)\s+Republic\b", # Added Republic
        ]

        found_names = []
        for pattern in patterns:
            # Find all matches for the pattern
            matches = re.findall(pattern, outline)
            # Add non-empty matches to the list
            found_names.extend([m.strip() for m in matches if isinstance(m, str) and m.strip()])
            # Handle patterns that return tuples (like those with capture groups)
            if matches and isinstance(matches[0], tuple):
                 found_names.extend([item.strip() for tpl in matches for item in tpl if isinstance(item, str) and item.strip()])


        if found_names:
            # Count occurrences of each name
            counts = {}
            for name in found_names:
                counts[name] = counts.get(name, 0) + 1
            # Return the most frequent name
            return max(counts, key=counts.get)

        # If no specific world name is found, let the LLM generate one
        print("No specific world name pattern matched in the outline.")
        return ""

    def create_story_outline(self):
        """Generate a high-level outline for the entire story with improved structure"""
        # Updated system prompt to include genre
        system_prompt = f"""You are a professional novelist and editor specializing in the {self.genre} genre.
You excel at creating compelling story outlines that maintain coherence and proper narrative structure.
Pay special attention to character development, consistent world-building, logical plot progression, and genre conventions for {self.genre}."""

        # Updated main prompt to include genre
        prompt = f"""Based on the following story premise, create a detailed story outline for a {self.num_chapters}-chapter {self.genre} book.

Story premise (Genre: {self.genre}): {self.story_premise}

For each chapter, provide:
1. A chapter title (evocative and relevant to the {self.genre} genre)
2. Key plot events (3-4 bullet points with specific details)
3. Character development points (which characters appear and how they develop)
4. Setting/location details (be specific, consistent, and fitting for {self.genre})

Also create a section titled "WORLD BUILDING" with:
1. The name of the main city/setting (create a unique, memorable name fitting the {self.genre} genre)
2. Key locations that will appear multiple times
3. Important technology, magic systems, or cultural elements relevant to {self.genre}

Then create a section titled "CHARACTERS" with:
1. Main protagonist (name, detailed description, motivation, arc)
2. Main antagonist (name, detailed description, motivation)
3. Supporting characters (name, role, connection to protagonist)

Also create a section titled "RECURRING MOTIFS" with 3-5 symbols, objects, or phrases that will recur throughout the story to provide thematic continuity, relevant to the {self.genre} genre.

The outline should have a clear beginning (chapters 1-2), middle (chapters 3 to {self.num_chapters-2}), and end (final {min(2, self.num_chapters-1)} chapters).
Ensure proper rising action, climax, and resolution structure appropriate for the {self.genre} genre.

For the climax chapters (chapters {self.num_chapters-2} and {self.num_chapters-1}), provide extra detail about:
1. How the final confrontation unfolds
2. What's at stake for each character
3. Specific steps in the resolution

Make sure all character names, settings, and plot elements remain 100% consistent throughout.
"""
        print("Generating detailed story outline...")
        self.story_outline = self.generate_text(prompt, system_prompt)
        if not self.story_outline:
             print("Error: Failed to generate story outline. Cannot proceed.")
             return # Stop if outline generation fails

        # Extract character information
        # Updated system prompt for character extraction
        char_system_prompt = f"You are a literary analyst specializing in character tracking within the {self.genre} genre."
        char_prompt = f"""Based on this story outline for a {self.genre} novel, create a detailed character guide:

{self.story_outline}

For EACH character, format as:
CHARACTER NAME: Brief description, role in story, key personality traits, motivation, background (ensure consistency with the {self.genre} genre)

Include EVERY character mentioned in the outline, even minor ones. Ensure names are spelled consistently.
"""
        print("Creating detailed character profiles...")
        character_text = self.generate_text(char_prompt, char_system_prompt)
        self.characters = self.extract_characters(character_text)
        if not self.characters:
            print("Warning: Character extraction yielded no results. Proceeding without pre-defined characters.")


        # Extract world name for consistency
        self.world_name = self.extract_world_name(self.story_outline)

        # If no world name was found, ask the LLM to create one
        if not self.world_name:
            print("Attempting to generate world name...")
            # Updated system prompt for world name generation
            world_system_prompt = f"You are a creative world-builder specializing in the {self.genre} genre."
            world_prompt = f"""Based on this story outline for a {self.genre} novel, create a unique and memorable name for the main world/city/setting:

{self.story_outline}

The name should be a single term (or a short phrase like 'City of X'), creative, and fitting the tone and genre ({self.genre}) of the story.
Reply with ONLY the world name, nothing else.
"""
            generated_world_name = self.generate_text(world_prompt, world_system_prompt)
            if generated_world_name:
                 # Clean up potential extra text from the response
                 self.world_name = generated_world_name.strip().split('\n')[0]
                 print(f"Generated world name: {self.world_name}")
            else:
                 print("Warning: Could not generate world name. Will proceed without one.")
                 self.world_name = "The Setting" # Fallback

        # Extract recurring motifs
        # Updated system prompt for motif extraction
        motif_system_prompt = f"You are a literary analyst identifying thematic elements in {self.genre} fiction."
        motif_prompt = f"""Based on this story outline for a {self.genre} novel, identify 3-5 recurring motifs, symbols, or objects that appear throughout the story:

{self.story_outline}

Format as a simple list of items, one per line.
These should be concrete objects, symbols, or phrases relevant to the {self.genre} genre that can recur throughout chapters.
"""
        print("Identifying recurring motifs...")
        motifs_text = self.generate_text(motif_prompt, motif_system_prompt)
        if motifs_text:
            # Filter empty lines and clean up potential bullet points/numbering
            self.recurring_motifs = [re.sub(r"^\s*[-\*\d]+\.?\s*", "", motif).strip()
                                     for motif in motifs_text.strip().split('\n')
                                     if motif.strip()]
        if self.recurring_motifs:
            print(f"Identified motifs: {', '.join(self.recurring_motifs)}")
        else:
            print("Warning: No recurring motifs identified.")


        # Create detailed chapter-by-chapter plan
        # Updated system prompt for chapter plan
        plan_system_prompt = f"You are a meticulous story planner specializing in {self.genre} narratives."
        chapter_plan_prompt = f"""Based on the story outline for this {self.genre} novel, create a VERY detailed chapter-by-chapter plan.

STORY OUTLINE: {self.story_outline}

GENRE: {self.genre}

For EACH chapter (1 through {self.num_chapters}), provide:
1. Chapter title (consistent with outline and genre)
2. Chapter summary (250-300 words, detailing key events and character interactions)
3. Scene breakdown (list each scene with location, characters present, and key action/dialogue points)
4. Character development in this chapter (specific changes or revelations for characters)
5. Plot advancement in this chapter (how the main plot moves forward)
6. Timeline indicators (time of day, date, or how much time has passed since previous chapter)
7. Emotional tone and tension level (1-10) at the end of the chapter
8. How this chapter connects to the next (create a narrative bridge or hook)

Be extremely specific and detailed, adhering to the {self.genre} conventions. This plan will be used to ensure narrative consistency.
"""
        print("Creating detailed chapter plan...")
        self.chapter_plan = self.generate_text(chapter_plan_prompt, plan_system_prompt)
        if not self.chapter_plan:
             print("Error: Failed to generate chapter plan. Book generation quality may be affected.")


    def create_chapter_summary(self, chapter_num, chapter_content):
        """Create a detailed summary of a chapter after it's written"""
        if not chapter_content: return None # Handle empty content
        # Updated system prompt
        system_prompt = f"""You are a literary analyst specializing in narrative structure and continuity for the {self.genre} genre.
Create comprehensive, detailed summaries that capture all key elements."""
        prompt = f"""Create a detailed summary of the following chapter content.
This is Chapter {chapter_num} of a {self.num_chapters}-chapter {self.genre} book.

Include in your summary:
1. All key plot developments relevant to the main story and subplots
2. Character appearances, actions, dialogue highlights, and development
3. Setting details and atmosphere established
4. Important dialogue or revelations
5. How this chapter connects to previous chapters and sets up future events
6. Emotional tone at the beginning and end of the chapter
7. Any introduction of new significant elements (characters, items, locations)

CHAPTER CONTENT:
{chapter_content}

Your summary should be comprehensive enough that another writer could use it to maintain perfect continuity for this {self.genre} story.
"""
        summary = self.generate_text(prompt, system_prompt)
        if summary:
            self.chapter_summaries[chapter_num] = summary
        else:
            print(f"Warning: Failed to generate summary for Chapter {chapter_num}.")
            self.chapter_summaries[chapter_num] = "[Summary generation failed]"
        return summary

    def update_character_tracking(self, chapter_num, chapter_content):
        """Update character tracking data based on a chapter's content"""
        if not chapter_content or not self.characters: return # Need content and characters
        # Updated system prompt
        system_prompt = f"""You are a narrative continuity expert specializing in tracking character development in {self.genre} stories.
Extract precise information about characters from text."""
        character_names = list(self.characters.keys())
        characters_str = ", ".join(character_names)
        prompt = f"""Based on the following chapter content from a {self.genre} novel, track the development of all characters mentioned.

CHAPTER CONTENT:
{chapter_content}

CHARACTERS TO TRACK: {characters_str}

For each character that APPEARS or is SIGNIFICANTLY MENTIONED in this chapter, provide updates on:
1. Current status (e.g., alive, injured, captured, location if changed significantly)
2. Key actions or decisions made in this chapter
3. Notable development or changes in motivation/personality
4. Significant interactions or changes in relationships with other characters
5. Emotional state at the end of the chapter within this context

Format as:
CHARACTER NAME: status|actions/decisions|development|relationships|emotional_state

Only include characters who actually appear or are directly impacted in this chapter. Be concise but informative.
If a character appears but has no significant changes in these areas, note that briefly (e.g., "CHARACTER NAME: present|no major changes|...").
"""
        character_updates = self.generate_text(prompt, system_prompt)

        if not character_updates:
            print(f"Warning: No character updates received or generated for Chapter {chapter_num}.")
            return

        # Parse and update character data
        # Improved pattern to handle potential missing fields and variations
        update_pattern = r"([A-Z][A-Za-z\s'-]+):\s*([^|]*)\|([^|]*)\|([^|]*)\|([^|]*)\|([^\n]*)"
        matches = re.findall(update_pattern, character_updates)

        updated_chars_in_chapter = set()

        for match in matches:
            name = match[0].strip()
            if name in self.characters:
                updated_chars_in_chapter.add(name)
                # Update fields only if they provide new information
                status = match[1].strip()
                actions = match[2].strip()
                development = match[3].strip()
                relationships = match[4].strip()
                emotion = match[5].strip()

                if status and "no major changes" not in status.lower():
                    self.characters[name]["status"] = status
                if development and "no major changes" not in development.lower():
                    self.characters[name]["development"].append({
                        "chapter": chapter_num,
                        "development": development,
                        "actions": actions # Store actions too
                    })
                if relationships:
                    # Basic update - just note the interaction mentioned
                    self.characters[name]["relationships"][f"interaction_ch_{chapter_num}"] = relationships

                # Update location and emotional state if provided
                if status and "/" in status: # Check if location is embedded in status
                    parts = status.split('/')
                    self.characters[name]["status"] = parts[0].strip()
                    self.characters[name]["location"] = parts[1].strip()
                elif "location" in status.lower(): # Check if location is explicitly mentioned
                     self.characters[name]["location"] = status # Crude, needs refinement based on LLM output format

                if emotion:
                    self.characters[name]["emotional_state"] = emotion

                # Record first appearance if not already set
                if self.characters[name]["first_appearance"] == 0:
                    self.characters[name]["first_appearance"] = chapter_num
            else:
                 print(f"Warning: Received update for unknown character '{name}' in Chapter {chapter_num}.")

        # Note characters present but not explicitly updated (if any were extracted)
        if updated_chars_in_chapter:
             present_chars = set(re.findall(r'\b(' + '|'.join(re.escape(c) for c in character_names) + r')\b', chapter_content))
             for char_name in present_chars:
                 if char_name in self.characters and char_name not in updated_chars_in_chapter:
                     if self.characters[char_name]["first_appearance"] == 0:
                         self.characters[char_name]["first_appearance"] = chapter_num
                     # Optionally add a note that they were present but not updated
                     # self.characters[char_name]["development"].append({"chapter": chapter_num, "development": "Present, no specific update extracted"})


    def update_timeline(self, chapter_num, chapter_content):
        """Extract and update timeline information for chapter"""
        if not chapter_content: return None
        system_prompt = f"""You are a literary analyst specializing in temporal structure in {self.genre} narratives."""

        prompt = f"""Based on the following chapter content from a {self.genre} novel, determine:
        1. How much time seems to have passed DURING this chapter (e.g., a few hours, one day, indeterminate)
        2. What is the approximate time of day or date at the END of the chapter (e.g., evening, night, Tuesday, Day 5)
        3. Any specific time markers mentioned within the chapter (e.g., "at dawn", "three hours later", "midnight")

        CHAPTER CONTENT:
        {chapter_content}

        Reply only with the time information in this format, keeping it concise:
        TIME_ELAPSED: [estimated time passed during chapter]
        END_TIME: [estimated time/day at chapter end]
        TIME_MARKERS: [list any specific markers mentioned, separated by commas]
        """

        time_info = self.generate_text(prompt, system_prompt)
        if time_info:
            self.timeline[chapter_num] = time_info.strip()
        else:
            print(f"Warning: Failed to generate timeline info for Chapter {chapter_num}.")
            self.timeline[chapter_num] = "TIME_ELAPSED: Unknown\nEND_TIME: Unknown\nTIME_MARKERS: None"
        return self.timeline[chapter_num]


    def track_emotional_arc(self, chapter_num, chapter_content):
        """Track emotional tone and tension at the end of the chapter"""
        if not chapter_content: return None
        system_prompt = f"""You are a literary analyst specializing in emotional arcs in {self.genre} storytelling."""

        prompt = f"""Analyze the emotional tone at the end of this chapter from a {self.genre} novel:

        CHAPTER CONTENT (focus on the ending):
        {chapter_content[-1500:]} # Analyze last part for ending tone

        Determine:
        1. The primary emotion felt by the reader/protagonist at the chapter's end (e.g., tension, relief, hope, despair, mystery, anticipation)
        2. The level of narrative tension (1-10 scale, where 1 is low and 10 is high cliffhanger)
        3. The primary unresolved question or conflict driving interest into the next chapter

        Reply in this format:
        EMOTION: [primary emotion]
        TENSION: [level 1-10]
        UNRESOLVED: [main unresolved question or conflict]
        """

        emotional_status = self.generate_text(prompt, system_prompt)
        if emotional_status:
            self.emotional_arc[chapter_num] = emotional_status.strip()
        else:
            print(f"Warning: Failed to generate emotional arc info for Chapter {chapter_num}.")
            self.emotional_arc[chapter_num] = "EMOTION: Unknown\nTENSION: Unknown\nUNRESOLVED: Unknown"
        return self.emotional_arc[chapter_num]

    def create_chapter_transition(self, chapter_num, chapter_content):
        """Create a transition from current chapter to the next"""
        if chapter_num >= self.num_chapters or not chapter_content:
            return ""  # No transition needed for the last chapter or if content is missing

        # Updated system prompt
        system_prompt = f"""You are a master storyteller specializing in creating suspenseful
        chapter endings and smooth transitions between chapters, particularly for the {self.genre} genre."""

        # Extract relevant part of chapter plan for the next chapter
        plan_extract_prompt = f"""From this detailed chapter plan for a {self.genre} novel, extract ONLY the plan
        for Chapter {chapter_num + 1}:

        {self.chapter_plan}

        Include ONLY Chapter {chapter_num + 1}'s detailed plan (title, summary, scenes, etc.).
        """
        next_chapter_plan = self.generate_text(plan_extract_prompt)
        if not next_chapter_plan:
             print(f"Warning: Could not extract plan for Chapter {chapter_num + 1} for transition generation.")
             next_chapter_plan = "[Next chapter plan not available]"


        # Get emotional status
        emotional_status = self.emotional_arc.get(chapter_num, "EMOTION: Unknown\nTENSION: Unknown\nUNRESOLVED: Unknown")

        # Get timeline info
        timeline_info = self.timeline.get(chapter_num, "END_TIME: Unknown")

        # Choose a recurring motif to include
        chosen_motif = ""
        if self.recurring_motifs:
            # Cycle through motifs based on chapter number
            motif_index = (chapter_num -1) % len(self.recurring_motifs) # Use chapter_num - 1 for 0-based index
            chosen_motif = self.recurring_motifs[motif_index]


        prompt = f"""Create a compelling transition paragraph (or two short ones) to serve as the ending of Chapter {chapter_num}
        of this {self.genre} novel and create anticipation for Chapter {chapter_num + 1}.

        CURRENT CHAPTER CONTENT (last ~1000 characters):
        {chapter_content[-1000:]}

        NEXT CHAPTER ({chapter_num + 1}) PLAN:
        {next_chapter_plan}

        EMOTIONAL STATE AT CURRENT CHAPTER END:
        {emotional_status}

        TIMELINE INFORMATION AT CURRENT CHAPTER END:
        {timeline_info}

        RECURRING MOTIF TO INCLUDE IF POSSIBLE:
        {chosen_motif if chosen_motif else "N/A"}

        The transition should:
        1. Conclude the current chapter with a hook, an unanswered question, or lingering tension appropriate for {self.genre}.
        2. Create anticipation for what comes next based on the next chapter's plan.
        3. Avoid completely resolving all tension from the current chapter.
        4. Link to the theme or events of the upcoming chapter without being too explicit.
        5. Maintain the emotional tone but hint at potential shifts in the next chapter.
        6. Include a subtle, natural reference to the recurring motif ({chosen_motif}) if it fits the scene.

        Generate ONLY the 1-2 transition paragraphs. These will be the FINAL paragraphs of Chapter {chapter_num}.
        Do not add any extra commentary.
        """

        transition = self.generate_text(prompt, system_prompt)
        if transition:
            self.transitions[chapter_num] = transition.strip()
            return self.transitions[chapter_num]
        else:
            print(f"Warning: Failed to generate transition for Chapter {chapter_num}.")
            # Provide a generic fallback transition
            fallback_transition = f"The events of the day left much unresolved. As {self.world_name} settled into uncertainty, the stage was set for what was to come."
            self.transitions[chapter_num] = fallback_transition
            return fallback_transition


    def create_next_chapter_opener(self, chapter_num):
        """Create a strong opening for the next chapter that connects to the previous one"""
        if chapter_num <= 1:
            return ""  # First chapter doesn't need a special opener

        # Updated system prompt
        system_prompt = f"""You are a master storyteller specializing in creating
        engaging chapter openings that connect smoothly to previous events in {self.genre} narratives."""

        # Get information about the previous chapter
        prev_chapter_summary = self.chapter_summaries.get(chapter_num - 1, "[Previous summary unavailable]")
        # Use the generated transition as the end of the previous chapter
        prev_transition = self.transitions.get(chapter_num - 1, "[Previous transition unavailable]")
        prev_emotional_status = self.emotional_arc.get(chapter_num - 1, "EMOTION: Unknown\nTENSION: Unknown\nUNRESOLVED: Unknown")
        prev_timeline = self.timeline.get(chapter_num - 1, "END_TIME: Unknown")

        # Extract timeline info to know how much time has passed
        prev_end_time = "Unknown"
        if prev_timeline:
             # Try to extract END_TIME more reliably
             time_match = re.search(r"END_TIME:\s*([^\n]+)", prev_timeline)
             if time_match:
                 prev_end_time = time_match.group(1).strip()


        # Extract relevant part of chapter plan for this chapter
        plan_extract_prompt = f"""From this detailed chapter plan for a {self.genre} novel, extract ONLY the plan
        for Chapter {chapter_num}:

        {self.chapter_plan}

        Include ONLY Chapter {chapter_num}'s detailed plan (title, summary, scenes, etc.).
        """
        this_chapter_plan = self.generate_text(plan_extract_prompt)
        if not this_chapter_plan:
             print(f"Warning: Could not extract plan for Chapter {chapter_num} for opener generation.")
             this_chapter_plan = "[This chapter plan not available]"


        prompt = f"""Create a compelling opening paragraph (3-5 sentences) for Chapter {chapter_num} of this {self.genre} novel,
        connecting seamlessly with the end of Chapter {chapter_num - 1}.

        PREVIOUS CHAPTER ({chapter_num - 1}) SUMMARY (for context):
        {prev_chapter_summary}

        END OF PREVIOUS CHAPTER ({chapter_num - 1}) (The actual transition text):
        {prev_transition}

        EMOTIONAL STATE AT END OF PREVIOUS CHAPTER:
        {prev_emotional_status}

        TIME AT END OF PREVIOUS CHAPTER:
        {prev_end_time}

        THIS CHAPTER'S ({chapter_num}) PLAN:
        {this_chapter_plan}

        The opening paragraph should:
        1. Establish a clear time connection to the end of the previous chapter based on the timeline ({prev_end_time}) and this chapter's plan (e.g., "The next morning...", "Three days later...", "Moments after the confrontation...").
        2. Provide a smooth transition that builds on the tension, questions, or mood from the previous chapter's ending ({prev_transition}).
        3. Immediately orient readers in the new scene (location, characters present, initial action/observation) as per this chapter's plan.
        4. Avoid summarizing or repeating information from the previous chapter.
        5. Set the emotional tone for this chapter, aligning with the {self.genre} genre.

        Generate ONLY the single strong opening paragraph (3-5 sentences).
        Do not add any extra commentary or the chapter title.
        """

        opener = self.generate_text(prompt, system_prompt)
        if opener:
            return opener.strip()
        else:
            print(f"Warning: Failed to generate opener for Chapter {chapter_num}.")
            # Generic fallback opener
            return f"Time passed in {self.world_name}. Following the recent events, the atmosphere was charged with anticipation."


    def validate_chapter_consistency(self, chapter_num, chapter_content):
        """Check chapter for consistency issues"""
        if not chapter_content: return "CONSISTENT" # Cannot validate empty content
        # Updated system prompt
        system_prompt = f"""You are a meticulous literary editor specializing in narrative consistency for the {self.genre} genre.
Your job is to identify and flag any inconsistencies in a narrative based ONLY on the provided context."""

        # Create context for consistency check
        # Limit context to prevent excessive prompt length, focus on recent chapters
        context_chapter_limit = 5
        start_chapter = max(1, chapter_num - context_chapter_limit)
        previous_summaries = ""
        for i in range(start_chapter, chapter_num):
            if i in self.chapter_summaries:
                # Add chapter number for clarity
                previous_summaries += f"Chapter {i} Summary:\n{self.chapter_summaries[i]}\n\n"

        # Character status based on tracking up to the *previous* chapter
        character_status_list = []
        for name, data in self.characters.items():
            # Include characters who have appeared or are defined in the initial outline
            if data.get("first_appearance", 0) < chapter_num or data.get("first_appearance", 0) == 0 : # Include if appeared before this chapter or defined initially
                status = data.get("status", "unknown")
                desc = data.get("description", "N/A")
                # Get last known location and emotion if available
                location = data.get("location", "unknown")
                emotion = data.get("emotional_state", "unknown")
                # Summarize recent development
                dev_summary = "None recorded"
                if data.get("development"):
                     # Get the last recorded development entry
                     last_dev = data["development"][-1].get("development", "N/A") if data["development"] else "N/A"
                     dev_summary = f"Last noted: {last_dev} (Ch {data['development'][-1]['chapter']})" if data["development"] else "None recorded"

                character_status_list.append(f"- {name}: Status({status}), Loc({location}), Emotion({emotion}). {dev_summary}. Desc: {desc[:100]}...") # Truncate desc

        character_status_str = "\n".join(character_status_list) if character_status_list else "No character status available."


        # Add timeline information from recent chapters
        timeline_info = ""
        for i in range(start_chapter, chapter_num):
            if i in self.timeline:
                timeline_info += f"Chapter {i} End Timeline: {self.timeline[i]}\n"
        timeline_info = timeline_info if timeline_info else "No recent timeline info available."

        # Add relevant part of the chapter plan for context
        plan_extract_prompt = f"""From this detailed chapter plan for a {self.genre} novel, extract ONLY the plan
        for Chapter {chapter_num}:

        {self.chapter_plan}

        Include ONLY Chapter {chapter_num}'s detailed plan (title, summary, scenes, etc.).
        """
        this_chapter_plan = self.generate_text(plan_extract_prompt)
        if not this_chapter_plan:
             this_chapter_plan = "[Chapter plan not available for validation]"


        prompt = f"""Analyze this chapter for consistency issues compared ONLY to the provided context (previous summaries, character status, timeline, and chapter plan).

STORY PREMISE (Genre: {self.genre}): {self.story_premise}

WORLD NAME: {self.world_name}

RECENT PREVIOUS CHAPTER SUMMARIES (Chapters {start_chapter}-{chapter_num-1}):
{previous_summaries if previous_summaries else "No previous summaries provided."}

CHARACTER STATUS BEFORE THIS CHAPTER:
{character_status_str}

RECENT TIMELINE INFORMATION:
{timeline_info}

THIS CHAPTER'S ({chapter_num}) INTENDED PLAN:
{this_chapter_plan}

CURRENT CHAPTER ({chapter_num}) CONTENT TO VALIDATE:
{chapter_content}

Identify ANY specific inconsistencies between the CURRENT CHAPTER CONTENT and the PROVIDED CONTEXT related to:
1. Character details (names, status, location, established personality/abilities, relationships)
2. Setting/location names or descriptions conflicting with {self.world_name} or previous descriptions.
3. Timeline contradictions (sequence of events, time elapsed, time of day).
4. Plot contradictions (events contradicting previous summaries or character status).
5. Contradictions with the intended chapter plan provided above.
6. Incorrect use or appearance of recurring motifs based on context.
7. Genre inconsistencies (elements that strongly clash with {self.genre} conventions).

List ONLY specific inconsistencies found. Be precise.
If no inconsistencies are found based *solely* on the provided context, respond ONLY with the word "CONSISTENT".
"""
        consistency_check = self.generate_text(prompt, system_prompt)

        # Post-process the check result
        if not consistency_check:
             print(f"Warning: Consistency check for Chapter {chapter_num} failed to generate a response.")
             return "CHECK_FAILED" # Indicate failure
        elif "CONSISTENT" in consistency_check.upper() and len(consistency_check) < 20: # Check if it's likely just the word "CONSISTENT"
             return "CONSISTENT"
        else:
             # Return the list of issues
             return consistency_check


    def fix_chapter_inconsistencies(self, chapter_num, chapter_content, issues):
        """Fix identified consistency issues in a chapter"""
        if not chapter_content: return None # Cannot fix empty content
        # Updated system prompt
        system_prompt = f"""You are a professional novelist and meticulous editor specializing in the {self.genre} genre.
You excel at seamlessly fixing narrative inconsistencies while preserving the core narrative, tone, and character voices."""

        # Create context for the fix - similar context as validation
        context_chapter_limit = 5
        start_chapter = max(1, chapter_num - context_chapter_limit)
        previous_summaries = ""
        for i in range(start_chapter, chapter_num):
            if i in self.chapter_summaries:
                previous_summaries += f"Chapter {i} Summary:\n{self.chapter_summaries[i]}\n\n"

        character_status_list = []
        for name, data in self.characters.items():
             if data.get("first_appearance", 0) < chapter_num or data.get("first_appearance", 0) == 0 :
                status = data.get("status", "unknown")
                desc = data.get("description", "N/A")
                location = data.get("location", "unknown")
                emotion = data.get("emotional_state", "unknown")
                dev_summary = "None recorded"
                if data.get("development"):
                     last_dev = data["development"][-1].get("development", "N/A") if data["development"] else "N/A"
                     dev_summary = f"Last noted: {last_dev} (Ch {data['development'][-1]['chapter']})" if data["development"] else "None recorded"
                character_status_list.append(f"- {name}: Status({status}), Loc({location}), Emotion({emotion}). {dev_summary}. Desc: {desc[:100]}...")

        character_status_str = "\n".join(character_status_list) if character_status_list else "No character status available."

        timeline_info = ""
        for i in range(start_chapter, chapter_num):
            if i in self.timeline:
                timeline_info += f"Chapter {i} End Timeline: {self.timeline[i]}\n"
        timeline_info = timeline_info if timeline_info else "No recent timeline info available."

        plan_extract_prompt = f"""From this detailed chapter plan for a {self.genre} novel, extract ONLY the plan
        for Chapter {chapter_num}:

        {self.chapter_plan}

        Include ONLY Chapter {chapter_num}'s detailed plan (title, summary, scenes, etc.).
        """
        this_chapter_plan = self.generate_text(plan_extract_prompt)
        if not this_chapter_plan:
             this_chapter_plan = "[Chapter plan not available for fix context]"


        prompt = f"""Rewrite Chapter {chapter_num} of this {self.genre} novel to fix ALL the identified consistency issues, using the provided context. Maintain the original chapter's core plot, character interactions, and tone as much as possible.

STORY PREMISE (Genre: {self.genre}): {self.story_premise}

WORLD NAME: {self.world_name} (use this name consistently)

RECENT PREVIOUS CHAPTER SUMMARIES (Chapters {start_chapter}-{chapter_num-1}):
{previous_summaries if previous_summaries else "No previous summaries provided."}

CHARACTER STATUS BEFORE THIS CHAPTER:
{character_status_str}

RECENT TIMELINE INFORMATION:
{timeline_info}

THIS CHAPTER'S ({chapter_num}) INTENDED PLAN:
{this_chapter_plan}

ORIGINAL CHAPTER ({chapter_num}) CONTENT (with inconsistencies):
{chapter_content}

CONSISTENCY ISSUES TO FIX:
{issues}

Guidelines for fixing:
1. Directly address and correct every issue listed in "CONSISTENCY ISSUES TO FIX".
2. Maintain the essential plot progression and key events of the original chapter.
3. Ensure all character names, statuses, locations, abilities, and personalities align with the "CHARACTER STATUS BEFORE THIS CHAPTER" and previous summaries.
4. Use the established world name ({self.world_name}) and setting details consistently.
5. Ensure the timeline (sequence, elapsed time, time of day) flows logically from the "RECENT TIMELINE INFORMATION".
6. Align the chapter's events more closely with "THIS CHAPTER'S INTENDED PLAN" where inconsistencies were noted.
7. Preserve the original writing style and {self.genre} tone unless the inconsistency demands a change.
8. Ensure smooth transitions between scenes within the rewritten chapter.

Rewrite the COMPLETE chapter, starting with the chapter title (e.g., "## Chapter {chapter_num}: [Original or Revised Title]"). Do not include any commentary before or after the rewritten chapter content.
"""
        print(f"Attempting to fix Chapter {chapter_num}...")
        fixed_chapter = self.generate_text(prompt, system_prompt)
        # Basic validation of the fix
        if fixed_chapter and f"Chapter {chapter_num}" in fixed_chapter[:100]: # Check if it seems like a chapter
             # Optional: Run validation again on the fixed chapter? Could be slow.
             # print(f"Re-validating fixed Chapter {chapter_num}...")
             # re_check = self.validate_chapter_consistency(chapter_num, fixed_chapter) # Need to update context before re-validating
             # if re_check == "CONSISTENT":
             #     print("Fix successful based on re-validation.")
             # else:
             #     print(f"Warning: Fixed chapter still shows issues on re-validation: {re_check}")
             return fixed_chapter.strip()
        else:
             print(f"Error: Fixing Chapter {chapter_num} failed or produced invalid output.")
             if fixed_chapter: print(f"Output received:\n{fixed_chapter[:200]}...") # Log snippet if available
             return None # Indicate failure


    def generate_chapter(self, chapter_num):
        """Generate a single chapter with enhanced context awareness and consistency checks"""
        # Updated system prompt
        system_prompt = f"""You are a celebrated novelist known for writing engaging, coherent chapters
in the {self.genre} genre. Your chapters have clear narrative structure, natural flow, strong character development,
and maintain perfect consistency with previously established elements."""

        # Build context from previous chapters (limit context size)
        context_chapter_limit = 5
        start_chapter = max(1, chapter_num - context_chapter_limit)
        context_summaries = []
        for i in range(start_chapter, chapter_num):
            if i in self.chapter_summaries:
                # Add chapter number for clarity
                context_summaries.append(f"Chapter {i} Summary:\n{self.chapter_summaries[i]}")
        context = "\n\n".join(context_summaries) if context_summaries else "This is the first chapter."


        # Create character context based on tracking up to the *previous* chapter
        character_context_list = []
        for name, data in self.characters.items():
             if data.get("first_appearance", 0) < chapter_num or data.get("first_appearance", 0) == 0 :
                status = data.get("status", "unknown")
                desc = data.get("description", "N/A")
                location = data.get("location", "unknown")
                emotion = data.get("emotional_state", "unknown")
                dev_summary = "None recorded"
                if data.get("development"):
                     last_dev = data["development"][-1].get("development", "N/A") if data["development"] else "N/A"
                     dev_summary = f"Last noted: {last_dev} (Ch {data['development'][-1]['chapter']})" if data["development"] else "None recorded"
                # Provide concise context for the LLM
                character_context_list.append(f"- {name}: Status({status}), Loc({location}), Emotion({emotion}). {dev_summary}. Desc: {desc[:100]}...")

        characters_in_chapter_str = "\n".join(character_context_list) if character_context_list else "No character context available (treat as first appearance if needed)."


        # Add timeline information from the *previous* chapter's end
        timeline_context = "No previous timeline info available."
        if chapter_num > 1 and (chapter_num - 1) in self.timeline:
            timeline_context = f"End of previous chapter ({chapter_num-1}) timeline: {self.timeline[chapter_num - 1]}"

        # Add emotional arc information from the *previous* chapter's end
        emotional_context = "No previous emotional context available."
        if chapter_num > 1 and (chapter_num - 1) in self.emotional_arc:
            emotional_context = f"End of previous chapter ({chapter_num-1}) emotional state: {self.emotional_arc[chapter_num - 1]}"

        # Create chapter opener for chapters after the first
        chapter_opener = ""
        if chapter_num > 1:
            print(f"Creating opener for Chapter {chapter_num}...")
            chapter_opener = self.create_next_chapter_opener(chapter_num)
            if not chapter_opener:
                 print(f"Warning: Failed to generate specific opener for Chapter {chapter_num}.")
                 chapter_opener = "[Start chapter directly]" # Fallback instruction


        # Extract relevant part of chapter plan for this chapter
        plan_extract_prompt = f"""From this detailed chapter plan for a {self.genre} novel, extract ONLY the plan
        for Chapter {chapter_num}:

        {self.chapter_plan}

        Include ONLY Chapter {chapter_num}'s detailed plan (title, summary, scenes, character dev, plot adv, etc.).
        """
        this_chapter_plan = self.generate_text(plan_extract_prompt)
        if not this_chapter_plan:
             print(f"Critical Warning: Could not extract plan for Chapter {chapter_num}. Generation quality will be severely impacted.")
             this_chapter_plan = f"[PLAN MISSING FOR CHAPTER {chapter_num}] - Improvise based on outline and previous summaries."


        # Choose a recurring motif to include
        motif_instruction = ""
        if self.recurring_motifs:
            motif_index = (chapter_num - 1) % len(self.recurring_motifs)
            chosen_motif = self.recurring_motifs[motif_index]
            motif_instruction = f"Include the recurring motif '{chosen_motif}' somewhere in this chapter in a natural and meaningful way, relevant to the {self.genre} context."


        # Updated main prompt to include genre
        prompt = f"""Write Chapter {chapter_num} of a {self.genre} novel based on the following guidelines and context:

STORY PREMISE (Genre: {self.genre}): {self.story_premise}

WORLD NAME: {self.world_name} (use this name consistently throughout)

OVERALL STORY OUTLINE (Reference):
{self.story_outline[:1500]}...

THIS CHAPTER'S ({chapter_num}) DETAILED PLAN:
{this_chapter_plan}

CHARACTER CONTEXT (Status before this chapter):
{characters_in_chapter_str}

RECENT PREVIOUS CHAPTER SUMMARIES (Chapters {start_chapter}-{chapter_num-1}):
{context}

TIMELINE CONTEXT:
{timeline_context}

EMOTIONAL CONTEXT FROM PREVIOUS CHAPTER:
{emotional_context}

SUGGESTED CHAPTER OPENING (Use this to start the chapter unless it's '[Start chapter directly]'):
{chapter_opener}

{motif_instruction}

GUIDELINES FOR WRITING CHAPTER {chapter_num}:
- Write a complete, engaging chapter following THIS CHAPTER'S DETAILED PLAN.
- Adhere strictly to the {self.genre} genre conventions in tone, style, and content.
- Target chapter length: approximately 2500-3500 words (adjust as needed for plot).
- Include vivid descriptions, realistic dialogue, internal thoughts, and meaningful character interactions.
- Maintain 100% consistency with CHARACTER CONTEXT, WORLD NAME ({self.world_name}), TIMELINE CONTEXT, and PREVIOUS CHAPTER SUMMARIES.
- This is chapter {chapter_num} of {self.num_chapters}. Ensure it advances the plot according to the plan and overall outline.
- Do NOT re-explain established world-building elements unless necessary for context.
- Ensure the chapter has its own mini-arc (setup, rising action, climax/turning point, resolution/hook).
- Start with the chapter title using the format: "## Chapter {chapter_num}: [Chapter Title from Plan]"
- ONLY introduce new characters if specified in the plan or absolutely essential, providing proper context.
- If using the SUGGESTED CHAPTER OPENING, integrate it seamlessly as the first paragraph(s).
- End the chapter with a compelling hook, lingering question, or heightened tension suitable for {self.genre}, leading towards the next chapter's plan.

Format the chapter with clear paragraph breaks and standard dialogue formatting (e.g., using quotation marks).
Begin DIRECTLY with the chapter title (## Chapter X: Title). Do not add any introductory text.
"""
        print(f"Generating Chapter {chapter_num}...")
        chapter_content = self.generate_text(prompt, system_prompt)

        if not chapter_content or not re.match(rf"##\s*Chapter\s*{chapter_num}", chapter_content.strip()):
            print(f"Error: Failed to generate valid content for Chapter {chapter_num}.")
            # Try one regeneration attempt with a simpler prompt? (Optional)
            return f"## Chapter {chapter_num}: Generation Failed\n\n[Content generation failed for this chapter]"


        # --- Post-Generation Processing ---
        max_fix_attempts = 1 # Limit fix attempts to prevent loops
        for attempt in range(max_fix_attempts + 1):
             # Check for consistency issues
             print(f"Validating Chapter {chapter_num} (Attempt {attempt+1})...")
             consistency_check = self.validate_chapter_consistency(chapter_num, chapter_content)

             if consistency_check == "CONSISTENT":
                 print(f"Chapter {chapter_num} is consistent.")
                 break # Exit loop if consistent
             elif consistency_check == "CHECK_FAILED":
                 print(f"Consistency check failed for Chapter {chapter_num}. Proceeding with current content.")
                 break # Exit loop if check fails
             elif attempt < max_fix_attempts:
                 print(f"Consistency issues found in Chapter {chapter_num}. Issues:\n{consistency_check}\nAttempting to fix (Attempt {attempt+1}/{max_fix_attempts})...")
                 fixed_content = self.fix_chapter_inconsistencies(chapter_num, chapter_content, consistency_check)
                 if fixed_content:
                     chapter_content = fixed_content
                     print(f"Chapter {chapter_num} fixed attempt {attempt+1} successful.")
                     # Continue loop to re-validate the fixed content
                 else:
                     print(f"Error: Failed to fix consistency issues for Chapter {chapter_num} on attempt {attempt+1}. Using potentially inconsistent content.")
                     break # Exit loop if fix fails
             else:
                 print(f"Warning: Chapter {chapter_num} still has consistency issues after {max_fix_attempts} fix attempts. Issues:\n{consistency_check}\nProceeding with potentially inconsistent content.")
                 # Keep the last generated (potentially inconsistent) content


        # Create summary and update tracking *after* potential fixes
        print(f"Creating detailed summary for Chapter {chapter_num}...")
        self.create_chapter_summary(chapter_num, chapter_content) # Use potentially fixed content

        print(f"Updating character tracking for Chapter {chapter_num}...")
        self.update_character_tracking(chapter_num, chapter_content) # Use potentially fixed content

        print(f"Updating timeline information for Chapter {chapter_num}...")
        self.update_timeline(chapter_num, chapter_content) # Use potentially fixed content

        print(f"Analyzing emotional arc for Chapter {chapter_num}...")
        self.track_emotional_arc(chapter_num, chapter_content) # Use potentially fixed content

        # Add transition if not the last chapter
        if chapter_num < self.num_chapters:
            print(f"Creating transitional ending for Chapter {chapter_num}...")
            transition = self.create_chapter_transition(chapter_num, chapter_content) # Use potentially fixed content
            if transition:
                # Append the transition to the chapter content
                # Ensure there's appropriate spacing
                chapter_content = chapter_content.strip() + '\n\n' + transition
            else:
                print(f"Warning: Failed to generate or append transition for Chapter {chapter_num}.")


        return chapter_content.strip() # Return final content

    def check_chapter_transitions(self):
        """Check and improve transitions between all chapters after generation"""
        if len(self.chapters) < 2:
             print("Not enough chapters to check transitions.")
             return # Need at least two chapters

        print("Performing final check on chapter transitions...")
        improved_chapters = [self.chapters[0]] # Start with the first chapter

        for i in range(1, len(self.chapters)):
            current_chapter_text = self.chapters[i]
            prev_chapter_text = improved_chapters[i-1] # Use potentially improved previous chapter

            # Extract end of previous and start of current for context
            prev_ending_snippet = ' '.join(prev_chapter_text.strip().split()[-250:]) # More context
            current_opening_snippet = ' '.join(current_chapter_text.strip().split()[:250]) # More context

            # Updated system prompt
            system_prompt = f"""You are a professional editor specializing in narrative flow and chapter transitions for the {self.genre} genre."""

            prompt = f"""Analyze the transition between the end of Chapter {i} and the beginning of Chapter {i+1} in this {self.genre} novel.

            END OF PREVIOUS CHAPTER ({i}) (Snippet):
            ...{prev_ending_snippet}

            BEGINNING OF CURRENT CHAPTER ({i+1}) (Snippet):
            {current_opening_snippet}...

            Focus on the flow, time jump (if any), tone shift, and clarity of connection.

            If the transition is already smooth, natural, and effective for the {self.genre} genre, respond ONLY with the word "SMOOTH".

            If the transition feels abrupt, confusing, repetitive, or could be improved:
            Rewrite the opening paragraph(s) of the CURRENT chapter ({i+1}) (approx. 1-2 paragraphs, 100-200 words) to create a better bridge.
            The revised opening should:
            1. Connect more seamlessly to the previous chapter's ending mood and events.
            2. Clearly establish the time and setting for the current chapter.
            3. Avoid jarring tone shifts unless intentional and well-handled.
            4. Maintain character and plot consistency.
            5. Engage the reader immediately in the current chapter's action or theme.

            Start your response with "REVISED:" followed ONLY by the rewritten opening paragraph(s) for Chapter {i+1}.
            Do NOT include the chapter title or any other commentary.
            """

            transition_check_response = self.generate_text(prompt, system_prompt)

            if transition_check_response and transition_check_response.strip().upper() == "SMOOTH":
                 print(f"Transition to Chapter {i+1} is smooth.")
                 improved_chapters.append(current_chapter_text)
            elif transition_check_response and transition_check_response.startswith("REVISED:"):
                try:
                    revised_beginning_text = transition_check_response.split("REVISED:", 1)[1].strip()
                    if revised_beginning_text:
                        # Find the chapter title (e.g., "## Chapter X: Title")
                        title_match = re.match(r"(##\s*Chapter\s*\d+[:\s\w\W]*?\n+)", current_chapter_text) # More robust title match
                        if title_match:
                            chapter_title = title_match.group(1)
                            # Get content after the title
                            content_after_title = current_chapter_text[len(chapter_title):].strip()
                            # Split the content after title into paragraphs
                            paragraphs_after_title = re.split(r'\n\s*\n', content_after_title) # Split on blank lines

                            # Estimate how many paragraphs the revision replaces (simple heuristic)
                            num_paras_in_revision = len(re.split(r'\n\s*\n', revised_beginning_text))

                            if len(paragraphs_after_title) >= num_paras_in_revision:
                                # Join the remaining paragraphs back together
                                remaining_content = '\n\n'.join(paragraphs_after_title[num_paras_in_revision:])
                                # Construct the improved chapter
                                improved_chapter_text = chapter_title.strip() + '\n\n' + revised_beginning_text + '\n\n' + remaining_content
                                improved_chapters.append(improved_chapter_text.strip())
                                print(f"Transition improved for Chapter {i+1}.")
                            else:
                                # Fallback: Not enough paragraphs to replace, just prepend revised opening after title
                                print(f"Warning: Replacing paragraphs for Chapter {i+1} transition failed. Prepending.")
                                improved_chapters.append(chapter_title.strip() + '\n\n' + revised_beginning_text + '\n\n' + content_after_title)
                        else:
                            # Fallback: No title found, prepend revised beginning (less ideal)
                            print(f"Warning: Chapter title not found in standard format for Chapter {i+1}. Prepending revised opening to existing text.")
                            improved_chapters.append(revised_beginning_text + '\n\n' + current_chapter_text.strip())
                    else:
                        print(f"Transition for Chapter {i+1} marked for revision, but no revised text provided. Using original.")
                        improved_chapters.append(current_chapter_text)
                except Exception as e:
                    print(f"Error parsing or applying revised transition for Chapter {i+1}: {e}. Using original.")
                    improved_chapters.append(current_chapter_text)
            else:
                print(f"Transition check/revision failed for Chapter {i+1}. Using original. Response: {transition_check_response[:100] if transition_check_response else 'None'}")
                improved_chapters.append(current_chapter_text)

        self.chapters = improved_chapters
        print("Chapter transitions check complete.")


    def generate_book(self):
        """Generate the complete book with enhanced consistency checks"""
        self.get_user_input()
        self.create_story_outline()

        if not self.story_outline or not self.chapter_plan:
            print("Critical error: Story outline or chapter plan generation failed. Aborting book generation.")
            return "Book generation failed due to missing outline or plan."

        start_time = time.time()
        for i in range(1, self.num_chapters + 1):
            chapter_start_time = time.time()
            print(f"\n--- Generating Chapter {i} of {self.num_chapters} ---")
            chapter = self.generate_chapter(i)
            if chapter and f"Chapter {i}" in chapter[:100]: # Basic check for valid chapter
                self.chapters.append(chapter)
                chapter_end_time = time.time()
                print(f"--- Chapter {i} generated in {chapter_end_time - chapter_start_time:.2f} seconds ---")
            else:
                print(f"Critical error: Generation of Chapter {i} failed or produced invalid output. Aborting book generation.")
                # Add a placeholder to indicate failure but stop generation
                self.chapters.append(f"## Chapter {i}: Content Generation Failed\n\n[Generation aborted after failure in this chapter.]")
                return self.compile_book() # Compile what was generated so far


            # Optional: Add a small delay between chapters if needed
            if i < self.num_chapters:
                 # time.sleep(1) # Short delay
                 pass # Currently no delay needed unless API rate limits are hit

        # Perform final check on transitions between chapters
        if len(self.chapters) > 1: # Only check transitions if there's more than one chapter
            self.check_chapter_transitions()

        end_time = time.time()
        print(f"\n--- Book generation complete in {end_time - start_time:.2f} seconds ---")
        return self.compile_book()

    def compile_book(self):
        """Compile all chapters into a complete book, containing only title and chapters."""
        # Updated title prompt to include genre
        title_prompt = f"""Create a compelling and marketable title for a {self.genre} book with the following premise and outline snippet:

Genre: {self.genre}
Premise: {self.story_premise}

Story Outline Snippet (first few chapter ideas/themes):
{self.story_outline[:1000] if self.story_outline else "N/A"}

Respond with ONLY the book title, nothing else. Make it appropriate for the {self.genre} genre.
"""
        # Use a more robust system prompt for title generation
        title_system_prompt = f"You are an expert book marketer specializing in catchy titles for the {self.genre} genre."
        book_title = self.generate_text(title_prompt, title_system_prompt)

        # Clean up the generated title
        if not book_title:
            book_title = f"Untitled {self.genre} Story" # Fallback title
        else:
            book_title = book_title.strip().split('\n')[0] # Take first line
            # Remove potential labels like "Title:", "Book Title:", etc.
            book_title = re.sub(r"^(Title|Book Title):?\s*", "", book_title, flags=re.IGNORECASE).strip()
            # Remove surrounding quotes if present
            book_title = book_title.strip('"\'')
            if not book_title: # If cleaning leaves it empty
                book_title = f"Untitled {self.genre} Story"


        # Add Author Name (Placeholder - could be made dynamic)
        author_name = "Generated by AI Narrator" # Placeholder

        # Compile book content
        # Use Markdown formatting: Title H1, Author H3, Chapters H2
        book = f"# {book_title}\n\n"
        book += f"### By {author_name}\n\n"
        book += "---\n\n" # Add a horizontal rule after title/author

        # Iterate through the generated chapters and append them
        for i, chapter_content in enumerate(self.chapters):
            # Ensure each chapter starts with H2 title and has proper spacing
            # The generate_chapter function should already format the title correctly
            if chapter_content.strip().startswith(f"## Chapter {i+1}"):
                 book += f"{chapter_content.strip()}\n\n"
            elif chapter_content.strip().startswith(f"Chapter {i+1}"): # Handle missing '##'
                 # Add the H2 markdown if missing
                 book += f"## {chapter_content.strip()}\n\n"
            elif f"Chapter {i+1}: Generation Failed" in chapter_content:
                 # Keep failure message as is
                 book += f"{chapter_content.strip()}\n\n"
            else:
                 # Fallback: Add a generic title if missing entirely
                 print(f"Warning: Chapter {i+1} content missing standard title format. Adding generic title.")
                 book += f"## Chapter {i+1}: [Untitled Chapter]\n\n{chapter_content.strip()}\n\n"

            # Add a separator between chapters for clarity in the raw file
            if i < len(self.chapters) - 1:
                book += "---\n\n"


        return book.strip() # Remove any trailing whitespace/newlines


    def save_book(self, book_content, filename_prefix="generated_book"):
        """Save the generated book and metadata to files"""
        if not book_content or "Book generation failed" in book_content or not self.chapters:
            print(f"Book content is missing or incomplete. Not saving files.")
            return

        # Sanitize title for filename
        safe_title = re.sub(r'[\\/*?:"<>|]', "", book_title).strip()
        safe_title = re.sub(r'\s+', '_', safe_title) # Replace spaces with underscores
        filename_md = f"{filename_prefix}_{safe_title}_{self.genre}.md"
        filename_json = f"{filename_prefix}_{safe_title}_{self.genre}_metadata.json"


        try:
            with open(filename_md, "w", encoding="utf-8") as f:
                f.write(book_content)
            print(f"Book saved as {filename_md}")
        except IOError as e:
            print(f"Error saving book to {filename_md}: {e}")
            return # Stop if book saving fails


        # Prepare metadata for saving
        metadata = {
            "title_generated": book_title,
            "genre": self.genre,
            "story_premise": self.story_premise,
            "num_chapters_requested": self.num_chapters,
            "num_chapters_generated": len(self.chapters),
            "world_name": self.world_name,
            "characters": self.characters, # Save the final state
            "chapter_summaries": self.chapter_summaries,
            "recurring_motifs": self.recurring_motifs,
            "timeline": self.timeline,
            "emotional_arc": self.emotional_arc,
            "story_outline_snippet": (self.story_outline[:2000] + "..." if len(self.story_outline) > 2000 else self.story_outline) if self.story_outline else "N/A",
            "chapter_plan_snippet": (self.chapter_plan[:2000] + "..." if len(self.chapter_plan) > 2000 else self.chapter_plan) if self.chapter_plan else "N/A",
            "generation_model": self.model,
            "generation_timestamp": time.strftime("%Y-%m-%d %H:%M:%S %Z")
        }

        try:
            with open(filename_json, "w", encoding="utf-8") as f:
                # Use default=str to handle potential non-serializable data gracefully
                json.dump(metadata, f, indent=4, default=str)
            print(f"Book metadata saved as {filename_json}")
        except IOError as e:
            print(f"Error saving metadata to {filename_json}: {e}")
        except TypeError as e:
            print(f"Error serializing metadata to JSON: {e}. Some data might not be fully saved.")


if __name__ == "__main__":
    # Example usage:
    generator = BookGenerator()
    # Set base URL if your Ollama instance is running elsewhere
    # generator.base_url = "http://your_ollama_ip:11434/api/generate"
    book_text_content = generator.generate_book()

    # Check if book generation produced content before saving
    if book_text_content and "Generation Failed" not in book_text_content and generator.chapters:
        # Extract the generated title from the compiled book content for the filename
        title_match = re.match(r"#\s*(.*)", book_text_content)
        book_title = title_match.group(1).strip() if title_match else "Untitled_Book"
        generator.save_book(book_text_content, filename_prefix=f"book_{book_title[:20]}") # Use part of title in prefix
    else:
        print("\nBook generation was unsuccessful or incomplete. No file saved.")
```

```python
import requests
import json
import time
import os
import re

# --- Environment Variable for API Key ---
# Make sure to set the OPENROUTER_API_KEY environment variable before running.
# Example (Linux/macOS): export OPENROUTER_API_KEY='your_openrouter_api_key'
# Example (Windows CMD): set OPENROUTER_API_KEY=your_openrouter_api_key
# Example (Windows PowerShell): $env:OPENROUTER_API_KEY='your_openrouter_api_key'
# ---

class BookGenerator:
    def __init__(self):
        # --- OpenRouter Configuration ---
        OPENROUTER_API_KEY = "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"
        self.api_key = OPENROUTER_API_KEY
        if not self.api_key:
            raise ValueError("OPENROUTER_API_KEY environment variable not set.")

        self.base_url = "https://openrouter.ai/api/v1/chat/completions"
        # Specify an OpenRouter compatible model identifier
        # Example: "mistralai/mistral-7b-instruct", "google/gemma-7b-it", "openai/gpt-4o"
        # Check OpenRouter documentation for available models: https://openrouter.ai/docs#models
        self.model = "meta-llama/llama-4-maverick:free" # Or choose another suitable model from OpenRouter

        # Optional: Set Your App's URL and Name for OpenRouter identification
        self.site_url = "http://localhost" # Replace with your app's URL if applicable
        self.app_title = "Eeman Tries Stuff" # Replace with your app's name
        # --- End OpenRouter Configuration ---

        self.story_premise = ""
        self.genre = ""
        self.num_chapters = 0
        self.story_outline = ""
        self.chapters = []
        self.characters = {}
        self.chapter_summaries = {}
        self.settings = {}
        self.plot_events = []
        self.world_name = ""
        self.chapter_plan = ""
        self.timeline = {}
        self.emotional_arc = {}
        self.transitions = {}
        self.recurring_motifs = []

    def get_user_input(self):
        """Get the story premise, genre, and number of chapters from the user"""
        print("""
CAPABILITIES: Creates novels or fanfiction with consistency checks using OpenRouter
GENERATION TIME: Varies depending on OpenRouter model and load
STORY INPUT: One paragraph up to 830 characters with your plot idea
"""
        )
        self.story_premise = input("Please provide a paragraph about what your story is about: ")
        self.genre = input("Please specify the genre for your story (e.g., Sci-Fi, Fantasy, Mystery): ").strip()
        if not self.genre:
            self.genre = "General Fiction"
            print("No genre specified, defaulting to General Fiction.")

        while True:
            try:
                self.num_chapters = int(input("How many chapters would you like (minimum 3): "))
                if self.num_chapters >= 3:
                    break
                else:
                    print("Please enter at least 3 chapters.")
            except ValueError:
                print("Please enter a valid number.")

    def generate_text(self, prompt, system_prompt="You are a creative fiction writer."):
        """Make API call to OpenRouter with the given prompt"""
        headers = {
            "Authorization": f"Bearer {self.api_key}",
            "Content-Type": "application/json",
            # Recommended headers for OpenRouter identification
            "HTTP-Referer": self.site_url,
            "X-Title": self.app_title,
        }

        # Structure the request body according to OpenRouter/OpenAI chat completions format
        messages = []
        if system_prompt:
            messages.append({"role": "system", "content": system_prompt})
        messages.append({"role": "user", "content": prompt})

        data = {
            "model": self.model,
            "messages": messages,
            # Optional parameters (e.g., temperature, max_tokens) can be added here
            # "temperature": 0.7,
            # "max_tokens": 4096, # Adjust based on model and needs
        }

        try:
            response = requests.post(self.base_url, headers=headers, json=data, timeout=180) # Increased timeout for potentially long generations
            response.raise_for_status() # Raise HTTPError for bad responses (4xx or 5xx)

            response_data = response.json()

            # Check for errors in the response body itself
            if "error" in response_data:
                print(f"Error from OpenRouter API: {response_data['error']}")
                return None

            # Extract the generated text from the response
            if "choices" in response_data and len(response_data["choices"]) > 0:
                message = response_data["choices"][0].get("message", {})
                content = message.get("content")
                if content:
                    return content.strip()
                else:
                    print("Warning: 'content' key missing in response choice message.")
                    print(f"Full Response Choice: {response_data['choices'][0]}")
                    return None
            else:
                print("Warning: 'choices' key missing or empty in OpenRouter response.")
                print(f"Full Response: {response_data}")
                return None

        except requests.exceptions.RequestException as e:
            print(f"Error making request to OpenRouter: {e}")
            # Attempt to get more details from the response if available
            if e.response is not None:
                try:
                    error_details = e.response.json()
                    print(f"Error details: {error_details}")
                except json.JSONDecodeError:
                    print(f"Could not decode error response: {e.response.text}")
            return None
        except json.JSONDecodeError as e:
            print(f"Error decoding JSON response from OpenRouter: {e}")
            # It's possible the response was successful but not valid JSON (unlikely for OpenRouter)
            print(f"Response text: {response.text}")
            return None
        except Exception as e: # Catch any other unexpected errors
             print(f"An unexpected error occurred during API call: {e}")
             return None

    # ========================================================================
    # All other methods (extract_characters, create_story_outline, etc.)
    # remain unchanged as they primarily deal with processing text and logic,
    # not the API call itself.
    # ========================================================================

    def extract_characters(self, text):
        """Extract character information from text and create structured data"""
        if not text: # Handle case where text is None or empty
            return {}
        characters = {}
        # Look for patterns like "CHARACTER NAME: description"
        # Made pattern more robust to handle variations in spacing and line breaks
        pattern = r"([A-Z][A-Za-z\s'-]+):\s+([\s\S]*?)(?=\n[A-Z][A-Za-z\s'-]+:|\Z)"
        matches = re.findall(pattern, text)

        for match in matches:
            name = match[0].strip()
            description = match[1].strip()
            if name and description: # Ensure name and description are not empty
                characters[name] = {
                    "name": name,
                    "description": description,
                    "first_appearance": 0, # Will be updated when character first appears
                    "status": "unknown", # Initial status
                    "development": [],
                    "relationships": {},
                    "location": "unknown",
                    "emotional_state": "unknown"
                }
            else:
                print(f"Warning: Skipped potential character entry due to missing name or description: {match}")


        # Fallback if the primary pattern fails - look for capitalized names at start of lines
        if not characters:
             print("Primary character extraction pattern failed. Trying fallback.")
             # Simple fallback: look for capitalized words followed by a colon at the start of a line
             fallback_pattern = r"^\s*([A-Z][A-Za-z\s'-]+):\s*(.*)"
             lines = text.split('\n')
             for line in lines:
                 match = re.match(fallback_pattern, line)
                 if match:
                     name = match.group(1).strip()
                     description = match.group(2).strip()
                     if name and description and name not in characters: # Avoid duplicates
                         characters[name] = {
                             "name": name,
                             "description": description,
                             "first_appearance": 0,
                             "status": "unknown",
                             "development": [],
                             "relationships": {},
                             "location": "unknown",
                             "emotional_state": "unknown"
                         }

        if not characters:
            print("Warning: No characters extracted. Check the format of the character text.")

        return characters

    def extract_world_name(self, outline):
        """Extract consistent world name from the story outline"""
        if not outline: return ""
        # Generic pattern to find world names without hardcoding specific ones
        # Added more patterns and made them case-insensitive where appropriate
        patterns = [
            r"Neo-[A-Za-z]+",
            r"[A-Z][a-z]+land", r"(?i)\bthe\s+Kingdom\s+of\s+([A-Z][a-z]+)\b",
            r"(?i)\bthe\s+([A-Z][a-z]+)\s+Kingdom\b",
            r"(?i)\bthe\s+([A-Z][a-z]+)\s+Empire\b",
            r"(?i)\bthe\s+Realm\s+of\s+([A-Z][a-z]+)\b",
            r"(?i)\bthe\s+([A-Z][a-z]+)\s+Realm\b",
            r"(?i)\b([A-Z][a-z]+)\s+World\b",
            r"(?i)\b([A-Z][a-z]+)\s+City\b",
            r"(?i)\bCity\s+of\s+([A-Z][a-z]+)\b",
            r"(?i)\bPlanet\s+([A-Z][a-z]+)\b", # Added Planet
            r"(?i)\bthe\s+([A-Z][a-z]+)\s+Federation\b", # Added Federation
            r"(?i)\bthe\s+([A-Z][a-z]+)\s+Republic\b", # Added Republic
        ]

        found_names = []
        for pattern in patterns:
            # Find all matches for the pattern
            matches = re.findall(pattern, outline)
            # Add non-empty matches to the list
            found_names.extend([m.strip() for m in matches if isinstance(m, str) and m.strip()])
            # Handle patterns that return tuples (like those with capture groups)
            if matches and isinstance(matches[0], tuple):
                 found_names.extend([item.strip() for tpl in matches for item in tpl if isinstance(item, str) and item.strip()])


        if found_names:
            # Count occurrences of each name
            counts = {}
            for name in found_names:
                counts[name] = counts.get(name, 0) + 1
            # Return the most frequent name
            return max(counts, key=counts.get)

        # If no specific world name is found, let the LLM generate one
        print("No specific world name pattern matched in the outline.")
        return ""

    def create_story_outline(self):
        """Generate a high-level outline for the entire story with improved structure"""
        # Updated system prompt to include genre
        system_prompt = f"""You are a professional novelist and editor specializing in the {self.genre} genre.
You excel at creating compelling story outlines that maintain coherence and proper narrative structure.
Pay special attention to character development, consistent world-building, logical plot progression, and genre conventions for {self.genre}."""

        # Updated main prompt to include genre
        prompt = f"""Based on the following story premise, create a detailed story outline for a {self.num_chapters}-chapter {self.genre} book.

Story premise (Genre: {self.genre}): {self.story_premise}

For each chapter, provide:
1. A chapter title (evocative and relevant to the {self.genre} genre)
2. Key plot events (3-4 bullet points with specific details)
3. Character development points (which characters appear and how they develop)
4. Setting/location details (be specific, consistent, and fitting for {self.genre})

Also create a section titled "WORLD BUILDING" with:
1. The name of the main city/setting (create a unique, memorable name fitting the {self.genre} genre)
2. Key locations that will appear multiple times
3. Important technology, magic systems, or cultural elements relevant to {self.genre}

Then create a section titled "CHARACTERS" with:
1. Main protagonist (name, detailed description, motivation, arc)
2. Main antagonist (name, detailed description, motivation)
3. Supporting characters (name, role, connection to protagonist)

Also create a section titled "RECURRING MOTIFS" with 3-5 symbols, objects, or phrases that will recur throughout the story to provide thematic continuity, relevant to the {self.genre} genre.

The outline should have a clear beginning (chapters 1-2), middle (chapters 3 to {self.num_chapters-2}), and end (final {min(2, self.num_chapters-1)} chapters).
Ensure proper rising action, climax, and resolution structure appropriate for the {self.genre} genre.

For the climax chapters (chapters {self.num_chapters-2} and {self.num_chapters-1}), provide extra detail about:
1. How the final confrontation unfolds
2. What's at stake for each character
3. Specific steps in the resolution

Make sure all character names, settings, and plot elements remain 100% consistent throughout.
"""
        print("Generating detailed story outline via OpenRouter...")
        self.story_outline = self.generate_text(prompt, system_prompt)
        if not self.story_outline:
             print("Error: Failed to generate story outline. Cannot proceed.")
             # Consider adding retry logic or more robust error handling here
             raise RuntimeError("Failed to generate essential story outline.")


        # Extract character information
        # Updated system prompt for character extraction
        char_system_prompt = f"You are a literary analyst specializing in character tracking within the {self.genre} genre."
        char_prompt = f"""Based on this story outline for a {self.genre} novel, create a detailed character guide:

{self.story_outline}

For EACH character, format as:
CHARACTER NAME: Brief description, role in story, key personality traits, motivation, background (ensure consistency with the {self.genre} genre)

Include EVERY character mentioned in the outline, even minor ones. Ensure names are spelled consistently.
"""
        print("Creating detailed character profiles via OpenRouter...")
        character_text = self.generate_text(char_prompt, char_system_prompt)
        self.characters = self.extract_characters(character_text)
        if not self.characters:
            print("Warning: Character extraction yielded no results. Proceeding without pre-defined characters.")


        # Extract world name for consistency
        self.world_name = self.extract_world_name(self.story_outline)

        # If no world name was found, ask the LLM to create one
        if not self.world_name:
            print("Attempting to generate world name via OpenRouter...")
            # Updated system prompt for world name generation
            world_system_prompt = f"You are a creative world-builder specializing in the {self.genre} genre."
            world_prompt = f"""Based on this story outline for a {self.genre} novel, create a unique and memorable name for the main world/city/setting:

{self.story_outline}

The name should be a single term (or a short phrase like 'City of X'), creative, and fitting the tone and genre ({self.genre}) of the story.
Reply with ONLY the world name, nothing else.
"""
            generated_world_name = self.generate_text(world_prompt, world_system_prompt)
            if generated_world_name:
                 # Clean up potential extra text from the response
                 self.world_name = generated_world_name.strip().split('\n')[0]
                 print(f"Generated world name: {self.world_name}")
            else:
                 print("Warning: Could not generate world name. Will proceed without one.")
                 self.world_name = "The Setting" # Fallback

        # Extract recurring motifs
        # Updated system prompt for motif extraction
        motif_system_prompt = f"You are a literary analyst identifying thematic elements in {self.genre} fiction."
        motif_prompt = f"""Based on this story outline for a {self.genre} novel, identify 3-5 recurring motifs, symbols, or objects that appear throughout the story:

{self.story_outline}

Format as a simple list of items, one per line.
These should be concrete objects, symbols, or phrases relevant to the {self.genre} genre that can recur throughout chapters.
"""
        print("Identifying recurring motifs via OpenRouter...")
        motifs_text = self.generate_text(motif_prompt, motif_system_prompt)
        if motifs_text:
            # Filter empty lines and clean up potential bullet points/numbering
            self.recurring_motifs = [re.sub(r"^\s*[-\*\d]+\.?\s*", "", motif).strip()
                                     for motif in motifs_text.strip().split('\n')
                                     if motif.strip()]
        if self.recurring_motifs:
            print(f"Identified motifs: {', '.join(self.recurring_motifs)}")
        else:
            print("Warning: No recurring motifs identified.")


        # Create detailed chapter-by-chapter plan
        # Updated system prompt for chapter plan
        plan_system_prompt = f"You are a meticulous story planner specializing in {self.genre} narratives."
        chapter_plan_prompt = f"""Based on the story outline for this {self.genre} novel, create a VERY detailed chapter-by-chapter plan.

STORY OUTLINE: {self.story_outline}

GENRE: {self.genre}

For EACH chapter (1 through {self.num_chapters}), provide:
1. Chapter title (consistent with outline and genre)
2. Chapter summary (250-300 words, detailing key events and character interactions)
3. Scene breakdown (list each scene with location, characters present, and key action/dialogue points)
4. Character development in this chapter (specific changes or revelations for characters)
5. Plot advancement in this chapter (how the main plot moves forward)
6. Timeline indicators (time of day, date, or how much time has passed since previous chapter)
7. Emotional tone and tension level (1-10) at the end of the chapter
8. How this chapter connects to the next (create a narrative bridge or hook)

Be extremely specific and detailed, adhering to the {self.genre} conventions. This plan will be used to ensure narrative consistency.
"""
        print("Creating detailed chapter plan via OpenRouter...")
        self.chapter_plan = self.generate_text(chapter_plan_prompt, plan_system_prompt)
        if not self.chapter_plan:
             print("Error: Failed to generate chapter plan. Book generation quality may be affected.")
             # Consider stopping if the plan is crucial
             raise RuntimeError("Failed to generate essential chapter plan.")


    def create_chapter_summary(self, chapter_num, chapter_content):
        """Create a detailed summary of a chapter after it's written"""
        if not chapter_content: return None # Handle empty content
        # Updated system prompt
        system_prompt = f"""You are a literary analyst specializing in narrative structure and continuity for the {self.genre} genre.
Create comprehensive, detailed summaries that capture all key elements."""
        prompt = f"""Create a detailed summary of the following chapter content.
This is Chapter {chapter_num} of a {self.num_chapters}-chapter {self.genre} book.

Include in your summary:
1. All key plot developments relevant to the main story and subplots
2. Character appearances, actions, dialogue highlights, and development
3. Setting details and atmosphere established
4. Important dialogue or revelations
5. How this chapter connects to previous chapters and sets up future events
6. Emotional tone at the beginning and end of the chapter
7. Any introduction of new significant elements (characters, items, locations)

CHAPTER CONTENT:
{chapter_content}

Your summary should be comprehensive enough that another writer could use it to maintain perfect continuity for this {self.genre} story.
"""
        summary = self.generate_text(prompt, system_prompt)
        if summary:
            self.chapter_summaries[chapter_num] = summary
        else:
            print(f"Warning: Failed to generate summary for Chapter {chapter_num}.")
            self.chapter_summaries[chapter_num] = "[Summary generation failed]"
        return summary

    def update_character_tracking(self, chapter_num, chapter_content):
        """Update character tracking data based on a chapter's content"""
        if not chapter_content or not self.characters: return # Need content and characters
        # Updated system prompt
        system_prompt = f"""You are a narrative continuity expert specializing in tracking character development in {self.genre} stories.
Extract precise information about characters from text."""
        character_names = list(self.characters.keys())
        characters_str = ", ".join(character_names)
        prompt = f"""Based on the following chapter content from a {self.genre} novel, track the development of all characters mentioned.

CHAPTER CONTENT:
{chapter_content}

CHARACTERS TO TRACK: {characters_str}

For each character that APPEARS or is SIGNIFICANTLY MENTIONED in this chapter, provide updates on:
1. Current status (e.g., alive, injured, captured, location if changed significantly)
2. Key actions or decisions made in this chapter
3. Notable development or changes in motivation/personality
4. Significant interactions or changes in relationships with other characters
5. Emotional state at the end of the chapter within this context

Format as:
CHARACTER NAME: status|actions/decisions|development|relationships|emotional_state

Only include characters who actually appear or are directly impacted in this chapter. Be concise but informative.
If a character appears but has no significant changes in these areas, note that briefly (e.g., "CHARACTER NAME: present|no major changes|...").
"""
        character_updates = self.generate_text(prompt, system_prompt)

        if not character_updates:
            print(f"Warning: No character updates received or generated for Chapter {chapter_num}.")
            return

        # Parse and update character data
        # Improved pattern to handle potential missing fields and variations
        update_pattern = r"([A-Z][A-Za-z\s'-]+):\s*([^|]*)\|([^|]*)\|([^|]*)\|([^|]*)\|([^\n]*)"
        matches = re.findall(update_pattern, character_updates)

        updated_chars_in_chapter = set()

        for match in matches:
            name = match[0].strip()
            if name in self.characters:
                updated_chars_in_chapter.add(name)
                # Update fields only if they provide new information
                status = match[1].strip()
                actions = match[2].strip()
                development = match[3].strip()
                relationships = match[4].strip()
                emotion = match[5].strip()

                if status and "no major changes" not in status.lower():
                    self.characters[name]["status"] = status
                if development and "no major changes" not in development.lower():
                    self.characters[name]["development"].append({
                        "chapter": chapter_num,
                        "development": development,
                        "actions": actions # Store actions too
                    })
                if relationships:
                    # Basic update - just note the interaction mentioned
                    self.characters[name]["relationships"][f"interaction_ch_{chapter_num}"] = relationships

                # Update location and emotional state if provided
                # Simple location parsing - might need adjustment based on LLM output consistency
                if status and (' at ' in status or ' in ' in status):
                    # Attempt to split status and location if keywords are present
                    loc_keywords = [' at ', ' in ']
                    for kw in loc_keywords:
                        if kw in status:
                            parts = status.split(kw, 1)
                            self.characters[name]["status"] = parts[0].strip()
                            self.characters[name]["location"] = parts[1].strip()
                            break # Stop after first keyword match
                    else: # If no keyword found, assume status is just status
                         self.characters[name]["status"] = status


                if emotion:
                    self.characters[name]["emotional_state"] = emotion

                # Record first appearance if not already set
                if self.characters[name].get("first_appearance", 0) == 0: # Use .get for safety
                    self.characters[name]["first_appearance"] = chapter_num
            else:
                 print(f"Warning: Received update for unknown character '{name}' in Chapter {chapter_num}.")

        # Note characters present but not explicitly updated (if any were extracted)
        if updated_chars_in_chapter:
             # Find potential character names present in the text
             present_chars_pattern = r'\b(' + '|'.join(re.escape(c) for c in character_names) + r')\b'
             present_chars = set(re.findall(present_chars_pattern, chapter_content))

             for char_name in present_chars:
                 if char_name in self.characters and char_name not in updated_chars_in_chapter:
                     if self.characters[char_name].get("first_appearance", 0) == 0:
                         self.characters[char_name]["first_appearance"] = chapter_num
                     # Optionally add a note that they were present but not updated


    def update_timeline(self, chapter_num, chapter_content):
        """Extract and update timeline information for chapter"""
        if not chapter_content: return None
        system_prompt = f"""You are a literary analyst specializing in temporal structure in {self.genre} narratives."""

        prompt = f"""Based on the following chapter content from a {self.genre} novel, determine:
        1. How much time seems to have passed DURING this chapter (e.g., a few hours, one day, indeterminate)
        2. What is the approximate time of day or date at the END of the chapter (e.g., evening, night, Tuesday, Day 5)
        3. Any specific time markers mentioned within the chapter (e.g., "at dawn", "three hours later", "midnight")

        CHAPTER CONTENT:
        {chapter_content}

        Reply only with the time information in this format, keeping it concise:
        TIME_ELAPSED: [estimated time passed during chapter]
        END_TIME: [estimated time/day at chapter end]
        TIME_MARKERS: [list any specific markers mentioned, separated by commas, or 'None']
        """

        time_info = self.generate_text(prompt, system_prompt)
        if time_info:
            # Basic validation/cleaning of the response format
            cleaned_info = ""
            for line in time_info.strip().split('\n'):
                if line.startswith("TIME_ELAPSED:") or line.startswith("END_TIME:") or line.startswith("TIME_MARKERS:"):
                    cleaned_info += line + "\n"
            self.timeline[chapter_num] = cleaned_info.strip() if cleaned_info else "TIME_ELAPSED: Unknown\nEND_TIME: Unknown\nTIME_MARKERS: None"

        else:
            print(f"Warning: Failed to generate timeline info for Chapter {chapter_num}.")
            self.timeline[chapter_num] = "TIME_ELAPSED: Unknown\nEND_TIME: Unknown\nTIME_MARKERS: None"
        return self.timeline[chapter_num]


    def track_emotional_arc(self, chapter_num, chapter_content):
        """Track emotional tone and tension at the end of the chapter"""
        if not chapter_content: return None
        system_prompt = f"""You are a literary analyst specializing in emotional arcs in {self.genre} storytelling."""

        prompt = f"""Analyze the emotional tone at the end of this chapter from a {self.genre} novel:

        CHAPTER CONTENT (focus on the ending):
        {chapter_content[-1500:]} # Analyze last part for ending tone

        Determine:
        1. The primary emotion felt by the reader/protagonist at the chapter's end (e.g., tension, relief, hope, despair, mystery, anticipation)
        2. The level of narrative tension (1-10 scale, where 1 is low and 10 is high cliffhanger)
        3. The primary unresolved question or conflict driving interest into the next chapter

        Reply in this format:
        EMOTION: [primary emotion]
        TENSION: [level 1-10]
        UNRESOLVED: [main unresolved question or conflict]
        """

        emotional_status = self.generate_text(prompt, system_prompt)
        if emotional_status:
             # Basic validation/cleaning of the response format
            cleaned_info = ""
            for line in emotional_status.strip().split('\n'):
                if line.startswith("EMOTION:") or line.startswith("TENSION:") or line.startswith("UNRESOLVED:"):
                    cleaned_info += line + "\n"
            self.emotional_arc[chapter_num] = cleaned_info.strip() if cleaned_info else "EMOTION: Unknown\nTENSION: Unknown\nUNRESOLVED: Unknown"
        else:
            print(f"Warning: Failed to generate emotional arc info for Chapter {chapter_num}.")
            self.emotional_arc[chapter_num] = "EMOTION: Unknown\nTENSION: Unknown\nUNRESOLVED: Unknown"
        return self.emotional_arc[chapter_num]

    def create_chapter_transition(self, chapter_num, chapter_content):
        """Create a transition from current chapter to the next"""
        if chapter_num >= self.num_chapters or not chapter_content:
            return ""  # No transition needed for the last chapter or if content is missing

        # Updated system prompt
        system_prompt = f"""You are a master storyteller specializing in creating suspenseful
        chapter endings and smooth transitions between chapters, particularly for the {self.genre} genre."""

        # Extract relevant part of chapter plan for the next chapter
        plan_extract_prompt = f"""From this detailed chapter plan for a {self.genre} novel, extract ONLY the plan
        for Chapter {chapter_num + 1}:

        {self.chapter_plan}

        Include ONLY Chapter {chapter_num + 1}'s detailed plan (title, summary, scenes, etc.).
        """
        next_chapter_plan = self.generate_text(plan_extract_prompt)
        if not next_chapter_plan:
             print(f"Warning: Could not extract plan for Chapter {chapter_num + 1} for transition generation.")
             next_chapter_plan = "[Next chapter plan not available]"


        # Get emotional status
        emotional_status = self.emotional_arc.get(chapter_num, "EMOTION: Unknown\nTENSION: Unknown\nUNRESOLVED: Unknown")

        # Get timeline info
        timeline_info = self.timeline.get(chapter_num, "END_TIME: Unknown")

        # Choose a recurring motif to include
        chosen_motif = ""
        if self.recurring_motifs:
            # Cycle through motifs based on chapter number
            motif_index = (chapter_num -1) % len(self.recurring_motifs) # Use chapter_num - 1 for 0-based index
            chosen_motif = self.recurring_motifs[motif_index]


        prompt = f"""Create a compelling transition paragraph (or two short ones) to serve as the ending of Chapter {chapter_num}
        of this {self.genre} novel and create anticipation for Chapter {chapter_num + 1}.

        CURRENT CHAPTER CONTENT (last ~1000 characters):
        {chapter_content[-1000:]}

        NEXT CHAPTER ({chapter_num + 1}) PLAN:
        {next_chapter_plan}

        EMOTIONAL STATE AT CURRENT CHAPTER END:
        {emotional_status}

        TIMELINE INFORMATION AT CURRENT CHAPTER END:
        {timeline_info}

        RECURRING MOTIF TO INCLUDE IF POSSIBLE:
        {chosen_motif if chosen_motif else "N/A"}

        The transition should:
        1. Conclude the current chapter with a hook, an unanswered question, or lingering tension appropriate for {self.genre}.
        2. Create anticipation for what comes next based on the next chapter's plan.
        3. Avoid completely resolving all tension from the current chapter.
        4. Link to the theme or events of the upcoming chapter without being too explicit.
        5. Maintain the emotional tone but hint at potential shifts in the next chapter.
        6. Include a subtle, natural reference to the recurring motif ({chosen_motif}) if it fits the scene.

        Generate ONLY the 1-2 transition paragraphs. These will be the FINAL paragraphs of Chapter {chapter_num}.
        Do not add any extra commentary.
        """

        transition = self.generate_text(prompt, system_prompt)
        if transition:
            self.transitions[chapter_num] = transition.strip()
            return self.transitions[chapter_num]
        else:
            print(f"Warning: Failed to generate transition for Chapter {chapter_num}.")
            # Provide a generic fallback transition
            fallback_transition = f"The events of the day left much unresolved. As {self.world_name} settled into uncertainty, the stage was set for what was to come."
            self.transitions[chapter_num] = fallback_transition
            return fallback_transition


    def create_next_chapter_opener(self, chapter_num):
        """Create a strong opening for the next chapter that connects to the previous one"""
        if chapter_num <= 1:
            return ""  # First chapter doesn't need a special opener

        # Updated system prompt
        system_prompt = f"""You are a master storyteller specializing in creating
        engaging chapter openings that connect smoothly to previous events in {self.genre} narratives."""

        # Get information about the previous chapter
        prev_chapter_summary = self.chapter_summaries.get(chapter_num - 1, "[Previous summary unavailable]")
        # Use the generated transition as the end of the previous chapter
        prev_transition = self.transitions.get(chapter_num - 1, "[Previous transition unavailable]")
        prev_emotional_status = self.emotional_arc.get(chapter_num - 1, "EMOTION: Unknown\nTENSION: Unknown\nUNRESOLVED: Unknown")
        prev_timeline = self.timeline.get(chapter_num - 1, "END_TIME: Unknown")

        # Extract timeline info to know how much time has passed
        prev_end_time = "Unknown"
        if prev_timeline:
             # Try to extract END_TIME more reliably
             time_match = re.search(r"END_TIME:\s*([^\n]+)", prev_timeline)
             if time_match:
                 prev_end_time = time_match.group(1).strip()


        # Extract relevant part of chapter plan for this chapter
        plan_extract_prompt = f"""From this detailed chapter plan for a {self.genre} novel, extract ONLY the plan
        for Chapter {chapter_num}:

        {self.chapter_plan}

        Include ONLY Chapter {chapter_num}'s detailed plan (title, summary, scenes, etc.).
        """
        this_chapter_plan = self.generate_text(plan_extract_prompt)
        if not this_chapter_plan:
             print(f"Warning: Could not extract plan for Chapter {chapter_num} for opener generation.")
             this_chapter_plan = "[This chapter plan not available]"


        prompt = f"""Create a compelling opening paragraph (3-5 sentences) for Chapter {chapter_num} of this {self.genre} novel,
        connecting seamlessly with the end of Chapter {chapter_num - 1}.

        PREVIOUS CHAPTER ({chapter_num - 1}) SUMMARY (for context):
        {prev_chapter_summary}

        END OF PREVIOUS CHAPTER ({chapter_num - 1}) (The actual transition text):
        {prev_transition}

        EMOTIONAL STATE AT END OF PREVIOUS CHAPTER:
        {prev_emotional_status}

        TIME AT END OF PREVIOUS CHAPTER:
        {prev_end_time}

        THIS CHAPTER'S ({chapter_num}) PLAN:
        {this_chapter_plan}

        The opening paragraph should:
        1. Establish a clear time connection to the end of the previous chapter based on the timeline ({prev_end_time}) and this chapter's plan (e.g., "The next morning...", "Three days later...", "Moments after the confrontation...").
        2. Provide a smooth transition that builds on the tension, questions, or mood from the previous chapter's ending ({prev_transition}).
        3. Immediately orient readers in the new scene (location, characters present, initial action/observation) as per this chapter's plan.
        4. Avoid summarizing or repeating information from the previous chapter.
        5. Set the emotional tone for this chapter, aligning with the {self.genre} genre.

        Generate ONLY the single strong opening paragraph (3-5 sentences).
        Do not add any extra commentary or the chapter title.
        """

        opener = self.generate_text(prompt, system_prompt)
        if opener:
            return opener.strip()
        else:
            print(f"Warning: Failed to generate opener for Chapter {chapter_num}.")
            # Generic fallback opener
            return f"Time passed in {self.world_name}. Following the recent events, the atmosphere was charged with anticipation."


    def validate_chapter_consistency(self, chapter_num, chapter_content):
        """Check chapter for consistency issues"""
        if not chapter_content: return "CONSISTENT" # Cannot validate empty content
        # Updated system prompt
        system_prompt = f"""You are a meticulous literary editor specializing in narrative consistency for the {self.genre} genre.
Your job is to identify and flag any inconsistencies in a narrative based ONLY on the provided context."""

        # Create context for consistency check
        # Limit context to prevent excessive prompt length, focus on recent chapters
        context_chapter_limit = 5
        start_chapter = max(1, chapter_num - context_chapter_limit)
        previous_summaries = ""
        for i in range(start_chapter, chapter_num):
            if i in self.chapter_summaries and self.chapter_summaries[i] != "[Summary generation failed]":
                # Add chapter number for clarity
                previous_summaries += f"Chapter {i} Summary:\n{self.chapter_summaries[i]}\n\n"

        # Character status based on tracking up to the *previous* chapter
        character_status_list = []
        for name, data in self.characters.items():
            # Include characters who have appeared or are defined in the initial outline
            if data.get("first_appearance", 0) < chapter_num or data.get("first_appearance", 0) == 0 : # Include if appeared before this chapter or defined initially
                status = data.get("status", "unknown")
                desc = data.get("description", "N/A")
                # Get last known location and emotion if available
                location = data.get("location", "unknown")
                emotion = data.get("emotional_state", "unknown")
                # Summarize recent development
                dev_summary = "None recorded"
                if data.get("development"):
                     # Get the last recorded development entry
                     last_dev_entry = data["development"][-1] if data["development"] else {}
                     last_dev = last_dev_entry.get("development", "N/A")
                     last_dev_ch = last_dev_entry.get("chapter", "?")
                     dev_summary = f"Last noted: {last_dev} (Ch {last_dev_ch})"

                character_status_list.append(f"- {name}: Status({status}), Loc({location}), Emotion({emotion}). {dev_summary}. Desc: {desc[:100]}...") # Truncate desc

        character_status_str = "\n".join(character_status_list) if character_status_list else "No character status available."


        # Add timeline information from recent chapters
        timeline_info = ""
        for i in range(start_chapter, chapter_num):
            if i in self.timeline and "Unknown" not in self.timeline[i]:
                timeline_info += f"Chapter {i} End Timeline:\n{self.timeline[i]}\n" # Add newline for clarity
        timeline_info = timeline_info if timeline_info else "No recent timeline info available."

        # Add relevant part of the chapter plan for context
        plan_extract_prompt = f"""From this detailed chapter plan for a {self.genre} novel, extract ONLY the plan
        for Chapter {chapter_num}:

        {self.chapter_plan}

        Include ONLY Chapter {chapter_num}'s detailed plan (title, summary, scenes, etc.).
        """
        this_chapter_plan = self.generate_text(plan_extract_prompt)
        if not this_chapter_plan:
             this_chapter_plan = "[Chapter plan not available for validation]"


        prompt = f"""Analyze this chapter for consistency issues compared ONLY to the provided context (previous summaries, character status, timeline, and chapter plan).

STORY PREMISE (Genre: {self.genre}): {self.story_premise}

WORLD NAME: {self.world_name}

RECENT PREVIOUS CHAPTER SUMMARIES (Chapters {start_chapter}-{chapter_num-1}):
{previous_summaries if previous_summaries else "No previous summaries provided."}

CHARACTER STATUS BEFORE THIS CHAPTER:
{character_status_str}

RECENT TIMELINE INFORMATION:
{timeline_info}

THIS CHAPTER'S ({chapter_num}) INTENDED PLAN:
{this_chapter_plan}

CURRENT CHAPTER ({chapter_num}) CONTENT TO VALIDATE:
{chapter_content}

Identify ANY specific inconsistencies between the CURRENT CHAPTER CONTENT and the PROVIDED CONTEXT related to:
1. Character details (names, status, location, established personality/abilities, relationships contradicting prior info).
2. Setting/location names or descriptions conflicting with {self.world_name} or previous descriptions.
3. Timeline contradictions (sequence of events, time elapsed, time of day conflicting with prior info).
4. Plot contradictions (events contradicting previous summaries or character status).
5. Significant deviations from the intended chapter plan provided above without apparent reason.
6. Incorrect use or appearance of recurring motifs based on context.
7. Genre inconsistencies (elements that strongly clash with {self.genre} conventions).

List ONLY specific inconsistencies found. Be precise.
If no inconsistencies are found based *solely* on the provided context, respond ONLY with the word "CONSISTENT".
"""
        consistency_check = self.generate_text(prompt, system_prompt)

        # Post-process the check result
        if not consistency_check:
             print(f"Warning: Consistency check for Chapter {chapter_num} failed to generate a response.")
             return "CHECK_FAILED" # Indicate failure
        # Be more lenient with the "CONSISTENT" check, allowing for minor surrounding text/whitespace
        elif "CONSISTENT" in consistency_check.upper() and len(consistency_check.strip()) < 25:
             return "CONSISTENT"
        else:
             # Return the list of issues
             return consistency_check.strip()


    def fix_chapter_inconsistencies(self, chapter_num, chapter_content, issues):
        """Fix identified consistency issues in a chapter"""
        if not chapter_content: return None # Cannot fix empty content
        # Updated system prompt
        system_prompt = f"""You are a professional novelist and meticulous editor specializing in the {self.genre} genre.
You excel at seamlessly fixing narrative inconsistencies while preserving the core narrative, tone, and character voices."""

        # Create context for the fix - similar context as validation
        context_chapter_limit = 5
        start_chapter = max(1, chapter_num - context_chapter_limit)
        previous_summaries = ""
        for i in range(start_chapter, chapter_num):
            if i in self.chapter_summaries and self.chapter_summaries[i] != "[Summary generation failed]":
                previous_summaries += f"Chapter {i} Summary:\n{self.chapter_summaries[i]}\n\n"

        character_status_list = []
        for name, data in self.characters.items():
             if data.get("first_appearance", 0) < chapter_num or data.get("first_appearance", 0) == 0 :
                status = data.get("status", "unknown")
                desc = data.get("description", "N/A")
                location = data.get("location", "unknown")
                emotion = data.get("emotional_state", "unknown")
                dev_summary = "None recorded"
                if data.get("development"):
                     last_dev_entry = data["development"][-1] if data["development"] else {}
                     last_dev = last_dev_entry.get("development", "N/A")
                     last_dev_ch = last_dev_entry.get("chapter", "?")
                     dev_summary = f"Last noted: {last_dev} (Ch {last_dev_ch})"
                character_status_list.append(f"- {name}: Status({status}), Loc({location}), Emotion({emotion}). {dev_summary}. Desc: {desc[:100]}...")

        character_status_str = "\n".join(character_status_list) if character_status_list else "No character status available."

        timeline_info = ""
        for i in range(start_chapter, chapter_num):
             if i in self.timeline and "Unknown" not in self.timeline[i]:
                timeline_info += f"Chapter {i} End Timeline:\n{self.timeline[i]}\n"
        timeline_info = timeline_info if timeline_info else "No recent timeline info available."

        plan_extract_prompt = f"""From this detailed chapter plan for a {self.genre} novel, extract ONLY the plan
        for Chapter {chapter_num}:

        {self.chapter_plan}

        Include ONLY Chapter {chapter_num}'s detailed plan (title, summary, scenes, etc.).
        """
        this_chapter_plan = self.generate_text(plan_extract_prompt)
        if not this_chapter_plan:
             this_chapter_plan = "[Chapter plan not available for fix context]"


        prompt = f"""Rewrite Chapter {chapter_num} of this {self.genre} novel to fix ALL the identified consistency issues, using the provided context. Maintain the original chapter's core plot, character interactions, and tone as much as possible.

STORY PREMISE (Genre: {self.genre}): {self.story_premise}

WORLD NAME: {self.world_name} (use this name consistently)

RECENT PREVIOUS CHAPTER SUMMARIES (Chapters {start_chapter}-{chapter_num-1}):
{previous_summaries if previous_summaries else "No previous summaries provided."}

CHARACTER STATUS BEFORE THIS CHAPTER:
{character_status_str}

RECENT TIMELINE INFORMATION:
{timeline_info}

THIS CHAPTER'S ({chapter_num}) INTENDED PLAN:
{this_chapter_plan}

ORIGINAL CHAPTER ({chapter_num}) CONTENT (with inconsistencies):
{chapter_content}

CONSISTENCY ISSUES TO FIX:
{issues}

Guidelines for fixing:
1. Directly address and correct every issue listed in "CONSISTENCY ISSUES TO FIX".
2. Maintain the essential plot progression and key events of the original chapter.
3. Ensure all character names, statuses, locations, abilities, and personalities align with the "CHARACTER STATUS BEFORE THIS CHAPTER" and previous summaries.
4. Use the established world name ({self.world_name}) and setting details consistently.
5. Ensure the timeline (sequence, elapsed time, time of day) flows logically from the "RECENT TIMELINE INFORMATION".
6. Align the chapter's events more closely with "THIS CHAPTER'S INTENDED PLAN" where inconsistencies were noted.
7. Preserve the original writing style and {self.genre} tone unless the inconsistency demands a change.
8. Ensure smooth transitions between scenes within the rewritten chapter.

Rewrite the COMPLETE chapter, starting with the chapter title (e.g., "## Chapter {chapter_num}: [Original or Revised Title]"). Do not include any commentary before or after the rewritten chapter content.
"""
        print(f"Attempting to fix Chapter {chapter_num} via OpenRouter...")
        fixed_chapter = self.generate_text(prompt, system_prompt)
        # Basic validation of the fix
        if fixed_chapter and f"Chapter {chapter_num}" in fixed_chapter[:100]: # Check if it seems like a chapter
             # Optional: Run validation again on the fixed chapter? Could be slow.
             # print(f"Re-validating fixed Chapter {chapter_num}...")
             # # Need to update context before re-validating if using this
             # re_check = self.validate_chapter_consistency(chapter_num, fixed_chapter)
             # if re_check == "CONSISTENT":
             #     print("Fix successful based on re-validation.")
             # else:
             #     print(f"Warning: Fixed chapter still shows issues on re-validation: {re_check}")
             return fixed_chapter.strip()
        else:
             print(f"Error: Fixing Chapter {chapter_num} failed or produced invalid output.")
             if fixed_chapter: print(f"Output received snippet:\n{fixed_chapter[:200]}...") # Log snippet if available
             return None # Indicate failure


    def generate_chapter(self, chapter_num):
        """Generate a single chapter with enhanced context awareness and consistency checks"""
        # Updated system prompt
        system_prompt = f"""You are a celebrated novelist known for writing engaging, coherent chapters
in the {self.genre} genre. Your chapters have clear narrative structure, natural flow, strong character development,
and maintain perfect consistency with previously established elements."""

        # Build context from previous chapters (limit context size)
        context_chapter_limit = 5 # How many previous chapters' summaries to include
        start_chapter = max(1, chapter_num - context_chapter_limit)
        context_summaries = []
        for i in range(start_chapter, chapter_num):
            if i in self.chapter_summaries and self.chapter_summaries[i] != "[Summary generation failed]":
                # Add chapter number for clarity
                context_summaries.append(f"Chapter {i} Summary:\n{self.chapter_summaries[i]}")
        context = "\n\n".join(context_summaries) if context_summaries else "This is the first chapter."


        # Create character context based on tracking up to the *previous* chapter
        character_context_list = []
        for name, data in self.characters.items():
             if data.get("first_appearance", 0) < chapter_num or data.get("first_appearance", 0) == 0 :
                status = data.get("status", "unknown")
                desc = data.get("description", "N/A")
                location = data.get("location", "unknown")
                emotion = data.get("emotional_state", "unknown")
                dev_summary = "None recorded"
                if data.get("development"):
                     last_dev_entry = data["development"][-1] if data["development"] else {}
                     last_dev = last_dev_entry.get("development", "N/A")
                     last_dev_ch = last_dev_entry.get("chapter", "?")
                     dev_summary = f"Last noted: {last_dev} (Ch {last_dev_ch})"
                # Provide concise context for the LLM
                character_context_list.append(f"- {name}: Status({status}), Loc({location}), Emotion({emotion}). {dev_summary}. Desc: {desc[:100]}...")

        characters_in_chapter_str = "\n".join(character_context_list) if character_context_list else "No character context available (treat as first appearance if needed)."


        # Add timeline information from the *previous* chapter's end
        timeline_context = "No previous timeline info available."
        if chapter_num > 1 and (chapter_num - 1) in self.timeline and "Unknown" not in self.timeline[chapter_num - 1]:
            timeline_context = f"End of previous chapter ({chapter_num-1}) timeline:\n{self.timeline[chapter_num - 1]}"

        # Add emotional arc information from the *previous* chapter's end
        emotional_context = "No previous emotional context available."
        if chapter_num > 1 and (chapter_num - 1) in self.emotional_arc and "Unknown" not in self.emotional_arc[chapter_num - 1]:
            emotional_context = f"End of previous chapter ({chapter_num-1}) emotional state:\n{self.emotional_arc[chapter_num - 1]}"

        # Create chapter opener for chapters after the first
        chapter_opener = ""
        if chapter_num > 1:
            print(f"Creating opener for Chapter {chapter_num} via OpenRouter...")
            chapter_opener = self.create_next_chapter_opener(chapter_num)
            if not chapter_opener:
                 print(f"Warning: Failed to generate specific opener for Chapter {chapter_num}.")
                 chapter_opener = "[Start chapter directly]" # Fallback instruction


        # Extract relevant part of chapter plan for this chapter
        plan_extract_prompt = f"""From this detailed chapter plan for a {self.genre} novel, extract ONLY the plan
        for Chapter {chapter_num}:

        {self.chapter_plan}

        Include ONLY Chapter {chapter_num}'s detailed plan (title, summary, scenes, character dev, plot adv, etc.).
        """
        this_chapter_plan = self.generate_text(plan_extract_prompt)
        if not this_chapter_plan:
             print(f"Critical Warning: Could not extract plan for Chapter {chapter_num}. Generation quality will be severely impacted.")
             this_chapter_plan = f"[PLAN MISSING FOR CHAPTER {chapter_num}] - Improvise based on outline and previous summaries."


        # Choose a recurring motif to include
        motif_instruction = ""
        if self.recurring_motifs:
            motif_index = (chapter_num - 1) % len(self.recurring_motifs)
            chosen_motif = self.recurring_motifs[motif_index]
            motif_instruction = f"Include the recurring motif '{chosen_motif}' somewhere in this chapter in a natural and meaningful way, relevant to the {self.genre} context."


        # Updated main prompt to include genre
        prompt = f"""Write Chapter {chapter_num} of a {self.genre} novel based on the following guidelines and context:

STORY PREMISE (Genre: {self.genre}): {self.story_premise}

WORLD NAME: {self.world_name} (use this name consistently throughout)

OVERALL STORY OUTLINE (Reference):
{self.story_outline[:1500]}...

THIS CHAPTER'S ({chapter_num}) DETAILED PLAN:
{this_chapter_plan}

CHARACTER CONTEXT (Status before this chapter):
{characters_in_chapter_str}

RECENT PREVIOUS CHAPTER SUMMARIES (Chapters {start_chapter}-{chapter_num-1}):
{context}

TIMELINE CONTEXT:
{timeline_context}

EMOTIONAL CONTEXT FROM PREVIOUS CHAPTER:
{emotional_context}

SUGGESTED CHAPTER OPENING (Use this to start the chapter unless it's '[Start chapter directly]'):
{chapter_opener}

{motif_instruction}

GUIDELINES FOR WRITING CHAPTER {chapter_num}:
- Write a complete, engaging chapter following THIS CHAPTER'S DETAILED PLAN.
- Adhere strictly to the {self.genre} genre conventions in tone, style, and content.
- Target chapter length: approximately 2500-3500 words (adjust as needed for plot). Use your judgment based on the plan.
- Include vivid descriptions, realistic dialogue, internal thoughts, and meaningful character interactions.
- Maintain 100% consistency with CHARACTER CONTEXT, WORLD NAME ({self.world_name}), TIMELINE CONTEXT, and PREVIOUS CHAPTER SUMMARIES. Check details carefully.
- This is chapter {chapter_num} of {self.num_chapters}. Ensure it advances the plot according to the plan and overall outline.
- Do NOT re-explain established world-building elements unless necessary for context. Assume reader remembers previous chapters.
- Ensure the chapter has its own mini-arc (setup, rising action, climax/turning point, resolution/hook).
- Start with the chapter title using the format: "## Chapter {chapter_num}: [Chapter Title from Plan]"
- ONLY introduce new characters if specified in the plan or absolutely essential, providing proper context.
- If using the SUGGESTED CHAPTER OPENING, integrate it seamlessly as the first paragraph(s).
- End the chapter with a compelling hook, lingering question, or heightened tension suitable for {self.genre}, leading towards the next chapter's plan.

Format the chapter with clear paragraph breaks (double newline) and standard dialogue formatting (e.g., using quotation marks).
Begin DIRECTLY with the chapter title (## Chapter X: Title). Do not add any introductory text or commentary.
"""
        print(f"Generating Chapter {chapter_num} via OpenRouter ({self.model})...")
        chapter_content = self.generate_text(prompt, system_prompt)

        # Validate basic structure of the generated content
        if not chapter_content or not re.match(rf"##\s*Chapter\s*{chapter_num}", chapter_content.strip()):
            print(f"Error: Failed to generate valid content for Chapter {chapter_num}. Expected '## Chapter {chapter_num}: ...'")
            if chapter_content: print(f"Received content snippet: {chapter_content[:200]}...")
            # Optionally, attempt regeneration here
            # print("Attempting regeneration...")
            # chapter_content = self.generate_text(prompt, system_prompt) # Simple retry
            # if not chapter_content or not re.match(rf"##\s*Chapter\s*{chapter_num}", chapter_content.strip()):
            #      print("Regeneration failed.")
            #      return f"## Chapter {chapter_num}: Generation Failed\n\n[Content generation failed for this chapter]"

            return f"## Chapter {chapter_num}: Generation Failed\n\n[Content generation failed for this chapter]"


        # --- Post-Generation Processing & Consistency Loop ---
        max_fix_attempts = 1 # Limit fix attempts to prevent loops
        for attempt in range(max_fix_attempts + 1):
             # Check for consistency issues
             print(f"Validating Chapter {chapter_num} (Attempt {attempt+1})...")
             consistency_check = self.validate_chapter_consistency(chapter_num, chapter_content)

             if consistency_check == "CONSISTENT":
                 print(f"Chapter {chapter_num} is consistent.")
                 break # Exit loop if consistent
             elif consistency_check == "CHECK_FAILED":
                 print(f"Consistency check failed for Chapter {chapter_num}. Proceeding with current content.")
                 break # Exit loop if check fails
             elif attempt < max_fix_attempts:
                 print(f"Consistency issues found in Chapter {chapter_num}. Issues:\n{consistency_check}\nAttempting to fix (Attempt {attempt+1}/{max_fix_attempts})...")
                 fixed_content = self.fix_chapter_inconsistencies(chapter_num, chapter_content, consistency_check)
                 if fixed_content:
                     chapter_content = fixed_content # Update content with the fix
                     print(f"Chapter {chapter_num} fix attempt {attempt+1} successful.")
                     # Continue loop to re-validate the fixed content
                 else:
                     print(f"Error: Failed to fix consistency issues for Chapter {chapter_num} on attempt {attempt+1}. Using potentially inconsistent content.")
                     break # Exit loop if fix fails
             else: # Reached max fix attempts
                 print(f"Warning: Chapter {chapter_num} may still have consistency issues after {max_fix_attempts} fix attempts. Issues:\n{consistency_check}\nProceeding with last generated version.")
                 # Keep the last generated (potentially inconsistent) content


        # Create summary and update tracking *after* potential fixes
        print(f"Creating detailed summary for Chapter {chapter_num}...")
        self.create_chapter_summary(chapter_num, chapter_content) # Use potentially fixed content

        print(f"Updating character tracking for Chapter {chapter_num}...")
        self.update_character_tracking(chapter_num, chapter_content) # Use potentially fixed content

        print(f"Updating timeline information for Chapter {chapter_num}...")
        self.update_timeline(chapter_num, chapter_content) # Use potentially fixed content

        print(f"Analyzing emotional arc for Chapter {chapter_num}...")
        self.track_emotional_arc(chapter_num, chapter_content) # Use potentially fixed content

        # Add transition if not the last chapter
        if chapter_num < self.num_chapters:
            print(f"Creating transitional ending for Chapter {chapter_num}...")
            transition = self.create_chapter_transition(chapter_num, chapter_content) # Use potentially fixed content
            if transition:
                # Append the transition to the chapter content
                # Ensure there's appropriate spacing (double newline)
                chapter_content = chapter_content.strip() + '\n\n' + transition
            else:
                print(f"Warning: Failed to generate or append transition for Chapter {chapter_num}.")


        return chapter_content.strip() # Return final content for the chapter

    def check_chapter_transitions(self):
        """Check and improve transitions between all chapters after generation"""
        if len(self.chapters) < 2:
             print("Not enough chapters to check transitions.")
             return # Need at least two chapters

        print("Performing final check on chapter transitions...")
        improved_chapters = [self.chapters[0]] # Start with the first chapter

        for i in range(1, len(self.chapters)):
            current_chapter_text = self.chapters[i]
            # Ensure previous chapter text is valid before proceeding
            if not improved_chapters[i-1] or "Generation Failed" in improved_chapters[i-1]:
                 print(f"Skipping transition check for Chapter {i+1} due to previous chapter failure.")
                 improved_chapters.append(current_chapter_text) # Keep current chapter as is
                 continue

            prev_chapter_text = improved_chapters[i-1] # Use potentially improved previous chapter


            # Extract end of previous and start of current for context
            prev_ending_snippet = ' '.join(prev_chapter_text.strip().split()[-250:]) # More context
            current_opening_snippet = ' '.join(current_chapter_text.strip().split()[:250]) # More context

            # Updated system prompt
            system_prompt = f"""You are a professional editor specializing in narrative flow and chapter transitions for the {self.genre} genre."""

            prompt = f"""Analyze the transition between the end of Chapter {i} and the beginning of Chapter {i+1} in this {self.genre} novel.

            END OF PREVIOUS CHAPTER ({i}) (Snippet):
            ...{prev_ending_snippet}

            BEGINNING OF CURRENT CHAPTER ({i+1}) (Snippet):
            {current_opening_snippet}...

            Focus on the flow, time jump (if any), tone shift, and clarity of connection.

            If the transition is already smooth, natural, and effective for the {self.genre} genre, respond ONLY with the word "SMOOTH".

            If the transition feels abrupt, confusing, repetitive, or could be improved:
            Rewrite the opening paragraph(s) of the CURRENT chapter ({i+1}) (approx. 1-2 paragraphs, 100-200 words) to create a better bridge.
            The revised opening should:
            1. Connect more seamlessly to the previous chapter's ending mood and events.
            2. Clearly establish the time and setting for the current chapter.
            3. Avoid jarring tone shifts unless intentional and well-handled.
            4. Maintain character and plot consistency.
            5. Engage the reader immediately in the current chapter's action or theme.

            Start your response with "REVISED:" followed ONLY by the rewritten opening paragraph(s) for Chapter {i+1}.
            Do NOT include the chapter title or any other commentary.
            """

            transition_check_response = self.generate_text(prompt, system_prompt)

            if transition_check_response and transition_check_response.strip().upper() == "SMOOTH":
                 print(f"Transition to Chapter {i+1} is smooth.")
                 improved_chapters.append(current_chapter_text)
            elif transition_check_response and transition_check_response.startswith("REVISED:"):
                try:
                    revised_beginning_text = transition_check_response.split("REVISED:", 1)[1].strip()
                    if revised_beginning_text:
                        # Find the chapter title (e.g., "## Chapter X: Title")
                        # Make title regex more robust to handle variations and multi-line titles
                        title_match = re.match(r"(##\s*Chapter\s*\d+[:\s\w\W]*?\n+)", current_chapter_text)
                        if title_match:
                            chapter_title = title_match.group(1)
                            # Get content after the title
                            content_after_title = current_chapter_text[len(chapter_title):].strip()
                            # Split the content after title into paragraphs using blank lines as separators
                            paragraphs_after_title = re.split(r'\n\s*\n', content_after_title)

                            # Estimate how many paragraphs the revision replaces (simple heuristic based on blank lines in revision)
                            num_paras_in_revision = len(re.split(r'\n\s*\n', revised_beginning_text))

                            if len(paragraphs_after_title) >= num_paras_in_revision:
                                # Join the remaining paragraphs back together, preserving blank lines
                                remaining_content = '\n\n'.join(paragraphs_after_title[num_paras_in_revision:])
                                # Construct the improved chapter, ensuring proper spacing
                                improved_chapter_text = chapter_title.strip() + '\n\n' + revised_beginning_text + ('\n\n' + remaining_content if remaining_content else '')
                                improved_chapters.append(improved_chapter_text.strip())
                                print(f"Transition improved for Chapter {i+1}.")
                            else:
                                # Fallback: Not enough paragraphs to replace, just prepend revised opening after title
                                print(f"Warning: Replacing paragraphs for Chapter {i+1} transition failed (not enough original paragraphs). Prepending.")
                                improved_chapters.append(chapter_title.strip() + '\n\n' + revised_beginning_text + '\n\n' + content_after_title)
                        else:
                            # Fallback: No title found, prepend revised beginning (less ideal)
                            print(f"Warning: Chapter title not found in standard format for Chapter {i+1}. Prepending revised opening to existing text.")
                            improved_chapters.append(revised_beginning_text + '\n\n' + current_chapter_text.strip())
                    else:
                        print(f"Transition for Chapter {i+1} marked for revision, but no revised text provided. Using original.")
                        improved_chapters.append(current_chapter_text)
                except Exception as e:
                    print(f"Error parsing or applying revised transition for Chapter {i+1}: {e}. Using original.")
                    improved_chapters.append(current_chapter_text)
            else:
                # Handle cases where the check failed or didn't return SMOOTH/REVISED
                print(f"Transition check/revision failed or yielded unexpected response for Chapter {i+1}. Using original. Response snippet: {transition_check_response[:100] if transition_check_response else 'None'}")
                improved_chapters.append(current_chapter_text)

        self.chapters = improved_chapters
        print("Chapter transitions check complete.")


    def generate_book(self):
        """Generate the complete book with enhanced consistency checks"""
        try:
            self.get_user_input()
            self.create_story_outline() # This now raises RuntimeError on failure
            # create_story_outline now also generates the chapter plan
            if not self.chapter_plan:
                 # This check might be redundant if create_story_outline raises error, but good for safety
                 raise RuntimeError("Chapter plan generation failed within create_story_outline.")

            start_time = time.time()
            for i in range(1, self.num_chapters + 1):
                chapter_start_time = time.time()
                print(f"\n--- Generating Chapter {i} of {self.num_chapters} ---")
                chapter = self.generate_chapter(i) # This now returns failure message on error

                if "Generation Failed" in chapter:
                    print(f"Critical error: Generation of Chapter {i} failed. Aborting book generation.")
                    self.chapters.append(chapter) # Add the failure message chapter
                    # Compile and return what was generated so far, including the error message
                    return self.compile_book()
                else:
                    self.chapters.append(chapter)
                    chapter_end_time = time.time()
                    print(f"--- Chapter {i} generated in {chapter_end_time - chapter_start_time:.2f} seconds ---")

                # Optional: Add a small delay between chapters if needed by API rate limits
                if i < self.num_chapters:
                     time.sleep(1) # Short delay (adjust if needed)

            # Perform final check on transitions between chapters
            if len(self.chapters) > 1: # Only check transitions if there's more than one chapter
                self.check_chapter_transitions()

            end_time = time.time()
            print(f"\n--- Book generation complete in {end_time - start_time:.2f} seconds ---")
            return self.compile_book()

        except (ValueError, RuntimeError, requests.exceptions.RequestException) as e:
             # Catch errors from init (API key), outline/plan generation, or network issues
             print(f"\n--- Book Generation Error ---")
             print(f"An error occurred: {e}")
             print("Book generation cannot continue.")
             # Compile any chapters that might have been generated before the error
             if self.chapters:
                 print("Compiling partial book content...")
                 partial_book = self.compile_book()
                 # Add error note to the end
                 partial_book += "\n\n---\n\n**Note:** Book generation was interrupted due to an error."
                 return partial_book
             else:
                 return "Book generation failed due to an early error."


    def compile_book(self):
        """Compile all chapters into a complete book, containing title, author, and chapters."""
        # --- Book Title Generation ---
        book_title = f"Untitled {self.genre} Story" # Default fallback
        if self.story_premise and self.story_outline: # Only generate title if we have premise/outline
            title_prompt = f"""Create a compelling and marketable title for a {self.genre} book with the following premise and outline snippet:

Genre: {self.genre}
Premise: {self.story_premise}

Story Outline Snippet (first few chapter ideas/themes):
{self.story_outline[:1000]}

Respond with ONLY the book title, nothing else. Make it appropriate for the {self.genre} genre.
"""
            title_system_prompt = f"You are an expert book marketer specializing in catchy titles for the {self.genre} genre."
            generated_title = self.generate_text(title_prompt, title_system_prompt)

            if generated_title:
                cleaned_title = generated_title.strip().split('\n')[0]
                cleaned_title = re.sub(r"^(Title|Book Title):?\s*", "", cleaned_title, flags=re.IGNORECASE).strip()
                cleaned_title = cleaned_title.strip('"\'')
                if cleaned_title: # Use cleaned title only if it's not empty
                    book_title = cleaned_title
            else:
                 print("Warning: Failed to generate book title. Using default.")
        else:
            print("Skipping title generation due to missing premise or outline.")
        # Store the final title used
        self.final_book_title = book_title


        # --- Author Name ---
        # Could potentially ask user for author name in get_user_input
        author_name = "Generated by AI Narrator" # Placeholder


        # --- Compile Book Content ---
        book = f"# {book_title}\n\n"
        book += f"### By {author_name}\n\n"
        book += "---\n\n" # Add a horizontal rule after title/author

        if not self.chapters:
             book += "[No chapters were generated.]"
             return book.strip()

        # Iterate through the generated chapters and append them
        for i, chapter_content in enumerate(self.chapters):
            chapter_num = i + 1
            # Ensure chapter content is a string
            if not isinstance(chapter_content, str):
                 print(f"Warning: Chapter {chapter_num} content is not a string. Skipping.")
                 book += f"## Chapter {chapter_num}: [Content Error]\n\n---\n\n"
                 continue

            # Add chapter content, ensuring correct formatting
            if chapter_content.strip().startswith(f"## Chapter {chapter_num}"):
                 book += f"{chapter_content.strip()}\n\n"
            elif chapter_content.strip().startswith(f"Chapter {chapter_num}"): # Handle missing '##'
                 book += f"## {chapter_content.strip()}\n\n"
            elif f"Chapter {chapter_num}: Generation Failed" in chapter_content:
                 book += f"{chapter_content.strip()}\n\n" # Keep failure message
            else:
                 # Fallback: Add a generic title if missing entirely
                 print(f"Warning: Chapter {chapter_num} content missing standard title format. Adding generic title.")
                 book += f"## Chapter {chapter_num}: [Untitled Chapter]\n\n{chapter_content.strip()}\n\n"

            # Add a separator between chapters for clarity, except after the last one
            if i < len(self.chapters) - 1:
                book += "---\n\n"


        return book.strip() # Remove any trailing whitespace/newlines


    def save_book(self, book_content, filename_prefix="generated_book"):
        """Save the generated book and metadata to files"""
        # Check if content exists and generation wasn't aborted early
        if not book_content or not self.chapters or "generation cannot continue" in book_content.lower():
            print(f"Book content is missing or generation was incomplete/failed. Not saving files.")
            return

        # Use the final compiled title for the filename
        book_title_for_file = getattr(self, 'final_book_title', f"Untitled_{self.genre}_Story")

        # Sanitize title for filename
        # Replace invalid chars and collapse whitespace/underscores
        safe_title = re.sub(r'[\\/*?:"<>|]', "", book_title_for_file).strip()
        safe_title = re.sub(r'\s+', '_', safe_title) # Replace spaces with underscores
        safe_title = re.sub(r'_+', '_', safe_title) # Collapse multiple underscores
        safe_genre = re.sub(r'\s+', '_', self.genre) # Sanitize genre too

        # Construct filenames
        timestamp = time.strftime("%Y%m%d_%H%M%S")
        filename_base = f"{filename_prefix}_{safe_title[:30]}_{safe_genre}_{timestamp}" # Truncate long titles
        filename_md = f"{filename_base}.md"
        filename_json = f"{filename_base}_metadata.json"


        try:
            with open(filename_md, "w", encoding="utf-8") as f:
                f.write(book_content)
            print(f"Book saved as {filename_md}")
        except IOError as e:
            print(f"Error saving book to {filename_md}: {e}")
            return # Stop if book saving fails


        # --- Prepare Metadata ---
        # Ensure chapter summaries/timeline/etc. use correct chapter numbers as keys (1-based)
        # Convert integer keys to strings for JSON compatibility if needed
        def stringify_keys(d):
             return {str(k): v for k, v in d.items()} if isinstance(d, dict) else d


        metadata = {
            "title_generated": book_title_for_file,
            "genre": self.genre,
            "story_premise": self.story_premise,
            "num_chapters_requested": self.num_chapters,
            "num_chapters_generated": len(self.chapters), # Number of entries in self.chapters list
            "world_name": self.world_name,
            "characters": self.characters, # Save the final state
            "chapter_summaries": stringify_keys(self.chapter_summaries),
            "recurring_motifs": self.recurring_motifs,
            "timeline": stringify_keys(self.timeline),
            "emotional_arc": stringify_keys(self.emotional_arc),
            "story_outline_snippet": (self.story_outline[:2000] + "..." if self.story_outline and len(self.story_outline) > 2000 else self.story_outline) if self.story_outline else "N/A",
            "chapter_plan_snippet": (self.chapter_plan[:2000] + "..." if self.chapter_plan and len(self.chapter_plan) > 2000 else self.chapter_plan) if self.chapter_plan else "N/A",
            "generation_model": self.model,
            "generation_timestamp": time.strftime("%Y-%m-%d %H:%M:%S %Z")
        }

        try:
            with open(filename_json, "w", encoding="utf-8") as f:
                # Use default=str to handle potential non-serializable data gracefully (like sets if they sneak in)
                json.dump(metadata, f, indent=4, default=str)
            print(f"Book metadata saved as {filename_json}")
        except IOError as e:
            print(f"Error saving metadata to {filename_json}: {e}")
        except TypeError as e:
            # This might happen if complex objects are in character data, etc.
            print(f"Error serializing metadata to JSON: {e}. Some data might not be fully saved.")


if __name__ == "__main__":
    try:
        # Initialize the generator (this checks for API key)
        generator = BookGenerator()

        # --- Optional: Customize settings before running ---
        # generator.model = "anthropic/claude-3-haiku-20240307" # Example: Choose a different model
        # generator.site_url = "https://my-writing-app.com"
        # generator.app_title = "My Novel Generator"
        # ---

        # Generate the book
        book_text_content = generator.generate_book()

        # Save the book if generation was successful
        if book_text_content and "Generation Failed" not in book_text_content and "generation cannot continue" not in book_text_content.lower() and generator.chapters:
            # Use the final title stored during compile_book
            final_title = getattr(generator, 'final_book_title', 'Untitled_Book')
            # Use a simpler prefix or let save_book handle the full name generation
            generator.save_book(book_text_content, filename_prefix="book")
        else:
            print("\nBook generation was unsuccessful or incomplete. Check logs for errors. No file saved (or only partial content saved).")

    except ValueError as e:
        # Catch API key error from __init__
        print(f"Configuration Error: {e}")
    except RuntimeError as e:
        # Catch critical errors during outline/plan generation
        print(f"Generation Error: {e}")
    except Exception as e:
        # Catch any other unexpected errors during setup or execution
        print(f"An unexpected error occurred: {e}")
        import traceback
        traceback.print_exc() # Print full traceback for debugging


```

```python
import requests
import json
import time
import os
import re

# --- Environment Variable for API Key ---
# Make sure to set the OPENROUTER_API_KEY environment variable before running.
# Example (Linux/macOS): export OPENROUTER_API_KEY='your_openrouter_api_key'
# Example (Windows CMD): set OPENROUTER_API_KEY=your_openrouter_api_key
# Example (Windows PowerShell): $env:OPENROUTER_API_KEY='your_openrouter_api_key'
# ---

class BookGenerator:
    def __init__(self):
        # --- OpenRouter Configuration ---
        # !!! IMPORTANT: Replace this hardcoded key with environment variable loading !!!
        # Load from environment variable for security
        # self.api_key = os.environ.get("OPENROUTER_API_KEY")
        # For demonstration purposes only, using the hardcoded key:
        # Replace with your actual key OR load from env var
        self.api_key = OPENROUTER_API_KEY
        # --- End Temporary Hardcoding ---

        if not self.api_key:
            raise ValueError("OPENROUTER_API_KEY environment variable not set.")

        self.base_url = "https://openrouter.ai/api/v1/chat/completions"
        # Specify an OpenRouter compatible model identifier
        # Example: "mistralai/mistral-7b-instruct", "google/gemma-7b-it", "openai/gpt-4o"
        # Check OpenRouter documentation for available models: https://openrouter.ai/docs#models
        # NOTE: Output quality HEAVILY depends on the chosen model. Larger/more capable models generally produce better results.
        self.model = "meta-llama/llama-4-maverick:free" # Or choose another suitable model

        # Optional: Set Your App's URL and Name for OpenRouter identification
        self.site_url = "http://localhost" # Replace with your app's URL if applicable
        self.app_title = "Enhanced Book Generator" # Updated app name
        # --- End OpenRouter Configuration ---

        self.story_premise = ""
        self.genre = ""
        self.extra_info = "" # <-- New field for extra user input
        self.num_chapters = 0
        self.story_outline = ""
        self.chapters = []
        self.characters = {}
        self.chapter_summaries = {}
        self.settings = {}
        self.plot_events = []
        self.world_name = ""
        self.chapter_plan = ""
        self.timeline = {}
        self.emotional_arc = {}
        self.transitions = {}
        self.recurring_motifs = []
        self.final_book_title = "" # To store the generated title


    def get_user_input(self):
        """Get the story premise, genre, extra info, and number of chapters from the user"""
        print(f"""
CAPABILITIES: Creates novels or fanfiction with consistency checks using OpenRouter.
SELECTED MODEL: {self.model} (Note: Output quality depends heavily on the chosen model)
GENERATION TIME: Varies depending on model, load, and chapter count.
STORY INPUT: One paragraph with your plot idea.
EXTRA INFO: Optional details, desired themes, style notes, or constraints.
"""
        )
        self.story_premise = input("Please provide a paragraph about what your story is about: ").strip()
        # --- Character Limit Removed ---
        # while len(self.story_premise) > 830: # Basic length check
        #      print("Story premise is too long (max ~830 chars). Please shorten it.")
        #      self.story_premise = input("Please provide a paragraph about what your story is about: ").strip()
        # ---

        self.genre = input("Please specify the genre (e.g., Sci-Fi, Fantasy, Mystery) [Default: General Fiction]: ").strip()
        if not self.genre:
            self.genre = "General Fiction"
            print("No genre specified, defaulting to General Fiction.")

        # --- Get Extra Info ---
        self.extra_info = input("Any extra info, themes, style notes, or constraints? (Optional, press Enter to skip):\n").strip()
        # ---

        while True:
            try:
                self.num_chapters = int(input("How many chapters would you like (minimum 3): "))
                if self.num_chapters >= 3:
                    break
                else:
                    print("Please enter at least 3 chapters.")
            except ValueError:
                print("Please enter a valid number.")


    def generate_text(self, prompt, system_prompt="You are a creative fiction writer."):
        """Make API call to OpenRouter with the given prompt"""
        # [ Existing generate_text code remains largely the same, ensure error handling is robust ]
        # --- Added Robust Error Handling and Logging ---
        headers = {
            "Authorization": f"Bearer {self.api_key}",
            "Content-Type": "application/json",
            "HTTP-Referer": self.site_url,
            "X-Title": self.app_title,
        }
        messages = []
        if system_prompt:
            messages.append({"role": "system", "content": system_prompt})
        messages.append({"role": "user", "content": prompt})

        data = {
            "model": self.model,
            "messages": messages,
            # Optional: Consider adding temperature for creativity control
            # "temperature": 0.75,
            # "max_tokens": 4096, # Model dependent, check OpenRouter docs
        }

        # Rate limiting consideration (simple backoff)
        max_retries = 3
        base_delay = 5 # seconds
        for attempt in range(max_retries):
            try:
                response = requests.post(self.base_url, headers=headers, json=data, timeout=240) # Increased timeout further
                response.raise_for_status()

                response_data = response.json()

                if "error" in response_data:
                    print(f"Error from OpenRouter API: {response_data['error']}")
                    # Check for rate limiting error specifically
                    if response.status_code == 429 or ('Rate limit' in response_data.get('error', {}).get('message', '')):
                         delay = base_delay * (2 ** attempt)
                         print(f"Rate limit likely hit. Retrying in {delay} seconds...")
                         time.sleep(delay)
                         continue # Retry the request
                    else:
                         return None # Other API error

                if "choices" in response_data and len(response_data["choices"]) > 0:
                    message = response_data["choices"][0].get("message", {})
                    content = message.get("content")
                    if content:
                        return content.strip()
                    else:
                        print("Warning: 'content' key missing in response choice message.")
                        print(f"Full Response Choice: {response_data['choices'][0]}")
                        return None
                else:
                    print("Warning: 'choices' key missing or empty in OpenRouter response.")
                    print(f"Full Response: {response_data}")
                    return None

            except requests.exceptions.HTTPError as e:
                print(f"HTTP Error making request to OpenRouter: {e}")
                print(f"Response Status Code: {e.response.status_code}")
                try:
                    error_details = e.response.json()
                    print(f"Error details: {error_details}")
                    # Check for rate limits on HTTPError too
                    if e.response.status_code == 429:
                         delay = base_delay * (2 ** attempt)
                         print(f"Rate limit likely hit (HTTP 429). Retrying in {delay} seconds...")
                         time.sleep(delay)
                         continue # Retry
                except json.JSONDecodeError:
                    print(f"Could not decode error response: {e.response.text}")
                # For non-rate-limit errors or after retries fail
                if attempt == max_retries - 1:
                     print("Max retries reached for HTTP error.")
                     return None
                # Optionally retry on 5xx errors as well?
                elif str(e.response.status_code).startswith('5'):
                     delay = base_delay * (2 ** attempt)
                     print(f"Server error ({e.response.status_code}). Retrying in {delay} seconds...")
                     time.sleep(delay)
                     continue # Retry

            except requests.exceptions.Timeout:
                 print(f"Request timed out after 240 seconds.")
                 delay = base_delay * (2 ** attempt)
                 print(f"Retrying in {delay} seconds...")
                 time.sleep(delay)
                 continue # Retry on timeout

            except requests.exceptions.RequestException as e:
                print(f"Error making request to OpenRouter: {e}")
                # Maybe retry certain connection errors? For now, fail fast.
                return None
            except json.JSONDecodeError as e:
                print(f"Error decoding JSON response from OpenRouter: {e}")
                # This usually indicates a server issue returning non-JSON
                print(f"Response text: {response.text if 'response' in locals() else 'Response object not available'}")
                return None
            except Exception as e: # Catch any other unexpected errors
                print(f"An unexpected error occurred during API call: {e}")
                import traceback
                traceback.print_exc()
                return None

        print(f"API call failed after {max_retries} retries.")
        return None # Return None if all retries fail

    # ========================================================================
    # Core Logic Methods - Primarily Prompt Adjustments Below
    # ========================================================================

    def extract_characters(self, text):
        # [ Existing extract_characters code is generally fine ]
        # Consider adding a prompt *before* this if the LLM fails often,
        # asking it to reformat the character list clearly.
        if not text: return {}
        characters = {}
        # More robust pattern: handles names with hyphens, apostrophes, titles (Dr., etc.)
        # Looks for NAME: description pattern, trying to capture multi-line descriptions.
        pattern = r"^([A-Z][A-Za-z.'\- ]+(?: [A-Z][a-z.'\- ]+)*):\s*([\s\S]*?)(?=\n[A-Z][A-Za-z.'\- ]+(?: [A-Z][a-z.'\- ]+)*:|\Z)"
        matches = re.findall(pattern, text, re.MULTILINE)

        if not matches: # Try slightly simpler pattern if the first fails
             pattern = r"([A-Z][A-Za-z\s'\-]+):\s+([\s\S]*?)(?=\n[A-Z][A-Za-z\s'\-]+:|\Z)"
             matches = re.findall(pattern, text)


        for match in matches:
            name = match[0].strip()
            description = match[1].strip()
            if name and description:
                # Basic check to avoid extracting section titles like "CHARACTERS:"
                if name.upper() in ["CHARACTERS", "CHARACTER LIST", "MAIN CHARACTERS", "SUPPORTING CHARACTERS"]:
                     continue
                characters[name] = {
                    "name": name,
                    "description": description,
                    "first_appearance": 0,
                    "status": "unknown",
                    "development": [],
                    "relationships": {},
                    "location": "unknown",
                    "emotional_state": "unknown"
                }
            else:
                print(f"Warning: Skipped potential character entry due to missing name or description: {match}")

        if not characters:
            print("Warning: No characters extracted using primary patterns. Check the format.")
            # Fallback idea: Ask LLM to extract names if regex fails? (More API calls)

        return characters

    def extract_world_name(self, outline):
         # [ Existing extract_world_name code is okay, regex is hard ]
         # The LLM generation fallback is probably the most reliable way
         if not outline: return ""
         patterns = [
            # Existing patterns...
            r"world of ([A-Z][a-zA-Z]+(?:[- ][A-Z][a-zA-Z]+)*)", # World of X
            r"city of ([A-Z][a-zA-Z]+(?:[- ][A-Z][a-zA-Z]+)*)", # City of X
            r"Planet ([A-Z][a-zA-Z]+(?:[- ][A-Z][a-zA-Z]+)*)", # Planet X
            r"the ([A-Z][a-zA-Z]+(?:[- ][A-Z][a-zA-Z]+)*) Sector", # X Sector
            # Add more specific patterns if needed for the genre
         ] + [ # Keep original patterns too
            r"Neo-[A-Za-z]+", r"[A-Z][a-z]+land", r"(?i)\bthe\s+Kingdom\s+of\s+([A-Z][a-z]+)\b",
            r"(?i)\bthe\s+([A-Z][a-z]+)\s+Kingdom\b", r"(?i)\bthe\s+([A-Z][a-z]+)\s+Empire\b",
            r"(?i)\bthe\s+Realm\s+of\s+([A-Z][a-z]+)\b", r"(?i)\bthe\s+([A-Z][a-z]+)\s+Realm\b",
            r"(?i)\b([A-Z][a-z]+)\s+World\b", r"(?i)\b([A-Z][a-z]+)\s+City\b",
            r"(?i)\bCity\s+of\s+([A-Z][a-z]+)\b", r"(?i)\bPlanet\s+([A-Z][a-z]+)\b",
            r"(?i)\bthe\s+([A-Z][a-z]+)\s+Federation\b", r"(?i)\bthe\s+([A-Z][a-z]+)\s+Republic\b",
         ]

         found_names = []
         # ... rest of the extraction logic ...
         # --- Add logic from original code below ---
         for pattern in patterns:
            matches = re.findall(pattern, outline)
            found_names.extend([m.strip() for m in matches if isinstance(m, str) and m.strip()])
            if matches and isinstance(matches[0], tuple):
                 found_names.extend([item.strip() for tpl in matches for item in tpl if isinstance(item, str) and item.strip()])
         # --- End of added logic ---

         if found_names:
            counts = {}
            for name in found_names:
                counts[name] = counts.get(name, 0) + 1
            # Filter out generic terms if they accidentally match
            generic_terms = {"The City", "The Kingdom", "The Empire", "The Realm", "The World", "The Planet"}
            for term in generic_terms:
                 if term in counts:
                      del counts[term]
            if counts: # Check if counts is not empty after filtering
                 return max(counts, key=counts.get)


         print("No specific world name pattern matched in the outline.")
         return "" # Fallback to LLM generation is handled later


    def create_story_outline(self):
        """Generate a high-level outline for the entire story with improved structure and quality focus"""
        print("Generating detailed story outline...")
        # --- Enhanced System Prompt ---
        system_prompt = f"""You are a professional novelist and editor specializing in high-quality {self.genre} fiction.
You excel at creating compelling, coherent, and richly detailed story outlines.
Focus on:
- Deep character arcs with clear motivations and development.
- Logical plot progression with rising tension, meaningful stakes, and a satisfying resolution.
- Integrated and immersive world-building consistent with the {self.genre} genre.
- Adherence to {self.genre} conventions while maintaining originality.
- Foreshadowing key events and themes."""

        # --- Enhanced Main Prompt ---
        prompt = f"""Based on the following story premise and user notes, create a *detailed and high-quality* story outline for a {self.num_chapters}-chapter {self.genre} book.

Story Premise (Genre: {self.genre}):
{self.story_premise}

{f"User's Extra Info/Notes to Consider: {self.extra_info}" if self.extra_info else ""}

For each chapter, provide:
1.  Chapter Title: Evocative, relevant to {self.genre}, and hinting at the chapter's core.
2.  Key Plot Events: 3-5 specific bullet points detailing significant actions, decisions, and revelations. Focus on cause and effect.
3.  Character Development: Which characters appear? How do they change, what motivates them in this chapter, what internal/external conflicts do they face? Show, don't just state development.
4.  Setting/Atmosphere: Specific locations within the world. Describe the mood and sensory details briefly. Ensure consistency and fit the {self.genre}.
5.  Pacing Note: (e.g., Fast-paced action, Slow burn tension, Character reflection).

Structure the outline with a clear:
-   Beginning (Setup/Inciting Incident - ~Chapters 1-2)
-   Middle (Rising Action/Complications - ~Chapters 3 to {self.num_chapters-2})
-   End (Climax/Falling Action/Resolution - Final ~2 chapters)

Include these sections AT THE END of the outline:

WORLD BUILDING DETAILS:
-   World Name: Create a unique, memorable name fitting the {self.genre} genre and premise.
-   Key Locations: List 3-5 significant locations that will appear multiple times, with brief descriptions.
-   Core Concepts: Describe 1-3 essential magic systems, technologies, societal rules, or cultural elements crucial to the story and {self.genre}.

CHARACTER PROFILES:
-   Protagonist(s): Name, detailed physical/personality description, core motivation, internal/external conflict, planned character arc (beginning state -> end state).
-   Antagonist(s): Name, description, motivation (make it compelling!), methods, relationship to the protagonist.
-   Key Supporting Characters (2-3): Name, role, core trait, relationship to protagonist/plot.

RECURRING MOTIFS/THEMES:
-   List 3-5 symbols, objects, phrases, or core themes (e.g., 'Betrayal and Redemption', 'Man vs. Nature') that will recur, adding depth and thematic resonance relevant to {self.genre}.

Ensure 100% consistency in names, locations, and established rules throughout the outline. The quality of this outline is crucial for the entire book.
"""
        self.story_outline = self.generate_text(prompt, system_prompt)
        if not self.story_outline:
            print("Error: Failed to generate story outline. Cannot proceed.")
            raise RuntimeError("Failed to generate essential story outline.")

        # --- Character Extraction (Prompt mostly unchanged, relies on outline format) ---
        char_system_prompt = f"You are a literary analyst specializing in character tracking within {self.genre} narratives."
        char_prompt = f"""Based *only* on the 'CHARACTER PROFILES' section of the story outline below, create a structured character guide.

Story Outline:
{self.story_outline}

Extract information *exclusively* from the 'CHARACTER PROFILES' section provided in the outline.
For EACH character listed there (Protagonist, Antagonist, Supporting), format as:
CHARACTER NAME: Description, Motivation, Role, Character Arc/Key Traits (as provided in the outline).

Format exactly as 'CHARACTER NAME: details'. Ensure names are spelled consistently. If a profile is missing details, note that.
"""
        print("Extracting character profiles from outline...")
        character_text = self.generate_text(char_prompt, char_system_prompt)
        self.characters = self.extract_characters(character_text) # Use the improved extractor
        if not self.characters:
            print("Warning: Character extraction yielded no results. Check 'CHARACTER PROFILES' format in the outline.")


        # --- World Name Extraction/Generation (Prompt mostly unchanged) ---
        self.world_name = self.extract_world_name(self.story_outline)
        if not self.world_name:
            print("Attempting to generate world name via OpenRouter...")
            world_system_prompt = f"You are a creative world-builder specializing in naming conventions for the {self.genre} genre."
            # Try extracting from the specific section first
            world_prompt = f"""Analyze the 'WORLD BUILDING DETAILS' section of this story outline:

{self.story_outline}

Identify the specified 'World Name'. Respond with ONLY the world name itself, nothing else.
If the name isn't explicitly listed under 'World Name', respond with 'NOT_FOUND'.
"""
            generated_world_name = self.generate_text(world_prompt, world_system_prompt)

            if generated_world_name and "NOT_FOUND" not in generated_world_name:
                 self.world_name = generated_world_name.strip().split('\n')[0]
                 print(f"Extracted/Generated world name: {self.world_name}")
            else:
                # If not found, try generating based on premise
                print("World name not found in designated section. Generating based on premise.")
                world_prompt_gen = f"""Based on this story premise for a {self.genre} novel, suggest ONE unique and memorable name for the main world/city/setting:

Premise: {self.story_premise}
Genre: {self.genre}
{f"User Notes: {self.extra_info}" if self.extra_info else ""}

The name should be creative and fit the tone/genre. Reply with ONLY the suggested world name, nothing else.
"""
                generated_world_name_gen = self.generate_text(world_prompt_gen, world_system_prompt)
                if generated_world_name_gen:
                    self.world_name = generated_world_name_gen.strip().split('\n')[0]
                    print(f"Generated world name: {self.world_name}")
                else:
                    print("Warning: Could not generate world name. Using fallback.")
                    self.world_name = "The Setting" # Fallback

        # --- Motif Extraction (Prompt mostly unchanged, relies on outline format) ---
        motif_system_prompt = f"You are a literary analyst identifying thematic elements in {self.genre} fiction."
        motif_prompt = f"""Based *only* on the 'RECURRING MOTIFS/THEMES' section of this story outline, list the recurring motifs, symbols, or themes identified:

Story Outline:
{self.story_outline}

Extract items *exclusively* from the 'RECURRING MOTIFS/THEMES' section.
Format as a simple list, one item per line.
"""
        print("Identifying recurring motifs...")
        motifs_text = self.generate_text(motif_prompt, motif_system_prompt)
        if motifs_text:
            self.recurring_motifs = [re.sub(r"^\s*[-\*\d]+\.?\s*", "", motif).strip()
                                     for motif in motifs_text.strip().split('\n')
                                     if motif.strip() and "NOT FOUND" not in motif.upper()] # Basic filter
        if self.recurring_motifs:
            print(f"Identified motifs/themes: {', '.join(self.recurring_motifs)}")
        else:
            print("Warning: No recurring motifs identified or extracted from the outline.")


        # --- Detailed Chapter Plan (Prompt Enhanced) ---
        plan_system_prompt = f"You are a meticulous story planner and {self.genre} narrative architect."
        chapter_plan_prompt = f"""Based on the detailed story outline for this {self.genre} novel, create an *extremely detailed* chapter-by-chapter plan. This plan is the blueprint for writing.

STORY OUTLINE:
{self.story_outline}

GENRE: {self.genre}
{f"USER'S EXTRA INFO/NOTES: {self.extra_info}" if self.extra_info else ""}

For EACH chapter (1 through {self.num_chapters}), provide the following structure consistently:

**Chapter {{{{Chapter Number}}}}: {{Chapter Title from Outline}}**

* **Objective:** What is the main goal or purpose of this chapter in the overall narrative?
* **Summary (Detailed):** ~150-250 words expanding on the outline's points. Detail key events, character interactions, decisions, and consequences. Describe the emotional flow.
* **Scene Breakdown:**
    * Scene 1: {{Location}}. Characters: {{List}}. Action/Dialogue Points: {{Bulleted list of key moments/beats}}. Emotional Tone: {{Brief description}}.
    * Scene 2: {{Location}}. Characters: {{List}}. Action/Dialogue Points: {{Bulleted list}}. Emotional Tone: {{Brief description}}.
    * (Add scenes as needed per chapter complexity)
* **Character Development Notes:** Specific moments showing change, revelation, or reinforcement of traits for key characters in this chapter. Link back to their overall arc.
* **Plot Advancement:** How does this chapter move the main plot and any subplots forward? What new questions are raised?
* **World-Building Integration:** Any specific details about the world (setting, magic, tech, society) revealed or utilized in this chapter?
* **Timeline/Pacing:** Estimated time covered (e.g., One evening, Three days). Note the chapter's pacing (e.g., High action, Slow tension build).
* **Ending Hook:** How does the chapter end to make the reader want to continue? (e.g., Cliffhanger, Revelation, Ominous Foreshadowing, Unanswered Question).

Be extremely specific and maintain 100% consistency with the outline's characters, world, and plot points. Adhere to {self.genre} conventions. Ensure logical flow between chapters.
"""
        print("Creating detailed chapter-by-chapter plan...")
        self.chapter_plan = self.generate_text(chapter_plan_prompt, plan_system_prompt)
        if not self.chapter_plan:
            print("Error: Failed to generate chapter plan. Book generation quality may be severely affected.")
            raise RuntimeError("Failed to generate essential chapter plan.")
        # Add a basic check for plan structure
        if f"Chapter {self.num_chapters}" not in self.chapter_plan:
             print(f"Warning: Generated chapter plan might be incomplete or improperly formatted (missing Chapter {self.num_chapters} marker).")


    def create_chapter_summary(self, chapter_num, chapter_content):
        # [ Existing method is okay, prompt slightly enhanced ]
        if not chapter_content or "Generation Failed" in chapter_content : return None
        system_prompt = f"""You are a literary analyst specializing in narrative summarization and continuity tracking for the {self.genre} genre. Focus on extracting all crucial details for consistency."""
        prompt = f"""Create a comprehensive summary of the following chapter. This is Chapter {chapter_num} of a {self.num_chapters}-chapter {self.genre} book.

Your summary must capture:
1.  **Key Plot Developments:** All significant events advancing the main plot and any subplots.
2.  **Character Information:** All characters appearing, their key actions, significant dialogue, decisions, and any noticeable development or changes in their state/motivation.
3.  **Setting & Atmosphere:** Where and when did the chapter take place? What was the prevailing mood?
4.  **Important Details:** Note any new items, locations, minor characters introduced, or crucial pieces of information revealed.
5.  **Continuity Links:** How does this chapter connect to the previous one and set up the next? Note the state of things at the chapter's end.
6.  **Theme/Motif Mentions:** Were any of the story's recurring themes or motifs present? ({', '.join(self.recurring_motifs) if self.recurring_motifs else 'N/A'})

CHAPTER CONTENT:
{chapter_content}

Write a detailed summary (200-300 words) suitable for ensuring perfect continuity in this {self.genre} story. Be objective and thorough.
"""
        summary = self.generate_text(prompt, system_prompt)
        if summary:
            self.chapter_summaries[chapter_num] = summary
        else:
            print(f"Warning: Failed to generate summary for Chapter {chapter_num}.")
            self.chapter_summaries[chapter_num] = "[Summary generation failed]"
        return summary


    def update_character_tracking(self, chapter_num, chapter_content):
        # [ Existing method is okay, prompt slightly enhanced ]
        if not chapter_content or not self.characters or "Generation Failed" in chapter_content: return
        system_prompt = f"""You are a narrative continuity expert meticulously tracking character details in {self.genre} stories."""
        character_names = list(self.characters.keys())
        characters_str = ", ".join(character_names)

        # Provide current known state for context
        current_status_context = "\n".join([f"- {name}: Status({data.get('status', 'unknown')}), Loc({data.get('location', 'unknown')})" for name, data in self.characters.items()])


        prompt = f"""Analyze the provided chapter content from a {self.genre} novel to update character tracking.

CURRENT KNOWN CHARACTER STATUS (Before this chapter):
{current_status_context}

CHARACTERS TO TRACK (if they appear or are significantly mentioned): {characters_str}

CHAPTER CONTENT (Chapter {chapter_num}):
{chapter_content}

For each character from the list who APPEARS or is SIGNIFICANTLY IMPACTED in this chapter, provide concise updates using this EXACT format (use 'N/A' if no notable change):
CHARACTER NAME: Status=[New Status]|Location=[New Location]|Actions=[Key actions/decisions]|Development=[Personality/Motivation changes]|Relationships=[Changes with others]|Emotion=[End-of-chapter state]

Example:
ELARA: Status=Injured|Location=Healer's Tent|Actions=Defended the pass, Used the amulet|Development=More determined, distrusts Ronan|Relationships=Closer bond with Finn|Emotion=Weary but resolute

- Focus ONLY on changes *within this chapter*.
- Be precise and brief.
- If a listed character does not appear or have significant changes, DO NOT include them in the output.
"""
        character_updates = self.generate_text(prompt, system_prompt)

        if not character_updates:
            print(f"Info: No significant character updates identified or generated for Chapter {chapter_num}.")
            return

        # --- Improved Parsing Logic ---
        # Use a more flexible regex to capture key-value pairs, allowing for variations
        update_pattern = r"([A-Z][A-Za-z\s'\-]+):\s*Status=\[(.*?)\]\|Location=\[(.*?)\]\|Actions=\[(.*?)\]\|Development=\[(.*?)\]\|Relationships=\[(.*?)\]\|Emotion=\[(.*?)\]"
        matches = re.findall(update_pattern, character_updates)

        updated_chars_in_chapter = set()

        for match in matches:
            name = match[0].strip()
            if name in self.characters:
                updated_chars_in_chapter.add(name)
                data_map = {
                    "status": match[1].strip(),
                    "location": match[2].strip(),
                    "actions": match[3].strip(),
                    "development": match[4].strip(),
                    "relationships": match[5].strip(),
                    "emotional_state": match[6].strip(),
                }

                # Update fields only if they provide new, non-N/A information
                if data_map["status"] and data_map["status"].upper() != 'N/A':
                    self.characters[name]["status"] = data_map["status"]
                if data_map["location"] and data_map["location"].upper() != 'N/A':
                     # If location update, also update status implicitly if status wasn't updated explicitly
                     if not (data_map["status"] and data_map["status"].upper() != 'N/A'):
                          # Check if status looks like 'Alive' or similar before overwriting
                           if self.characters[name].get("status", "").lower() not in ["alive", "ok", "unharmed", "present"]:
                                self.characters[name]["status"] = f"At {data_map['location']}" # Simple update
                     self.characters[name]["location"] = data_map["location"]

                if data_map["development"] and data_map["development"].upper() != 'N/A':
                    self.characters[name]["development"].append({
                        "chapter": chapter_num,
                        "development": data_map["development"],
                        "actions": data_map["actions"] # Store associated actions
                    })
                if data_map["relationships"] and data_map["relationships"].upper() != 'N/A':
                    # Append relationship notes keyed by chapter
                    rel_key = f"ch_{chapter_num}_interactions"
                    if rel_key not in self.characters[name]["relationships"]:
                        self.characters[name]["relationships"][rel_key] = []
                    self.characters[name]["relationships"][rel_key].append(data_map["relationships"])

                if data_map["emotional_state"] and data_map["emotional_state"].upper() != 'N/A':
                    self.characters[name]["emotional_state"] = data_map["emotional_state"]

                # Record first appearance if not already set
                if self.characters[name].get("first_appearance", 0) == 0:
                    self.characters[name]["first_appearance"] = chapter_num
            else:
                print(f"Warning: Received update for unknown character '{name}' in Chapter {chapter_num}.")

        # --- Update first appearance for characters present but not explicitly updated ---
        present_chars_pattern = r'\b(' + '|'.join(re.escape(c) for c in character_names) + r')\b'
        present_chars = set(re.findall(present_chars_pattern, chapter_content))
        for char_name in present_chars:
             if char_name in self.characters and self.characters[char_name].get("first_appearance", 0) == 0:
                  self.characters[char_name]["first_appearance"] = chapter_num
                  # Optionally note presence if no other update occurred
                  if char_name not in updated_chars_in_chapter:
                       self.characters[char_name]["development"].append({
                           "chapter": chapter_num, "development": "Present in chapter, no major updates noted.", "actions": "N/A"
                       })
                       print(f"Info: Noted presence of {char_name} in Chapter {chapter_num}.")


    def update_timeline(self, chapter_num, chapter_content):
         # [ Existing method is okay, prompt slightly clarified ]
        if not chapter_content or "Generation Failed" in chapter_content: return None
        system_prompt = f"You are a literary analyst tracking temporal progression in {self.genre} narratives."
        prompt = f"""Analyze the time progression within this chapter ({chapter_num}) of a {self.genre} novel:

CHAPTER CONTENT:
{chapter_content}

Determine the following based *only* on the text provided:
1.  TIME_ELAPSED: How much time passed *during* this chapter? (e.g., A few hours, One day, Indeterminate)
2.  END_TIME: What is the approximate time/day at the chapter's END? (e.g., Evening, Midnight, Tuesday, Day 5 after event X)
3.  TIME_MARKERS: List any specific time phrases mentioned (e.g., "at dawn", "three hours later", "midnight"), separated by commas. Use 'None' if none mentioned.

Reply using the exact format below, keeping answers concise:
TIME_ELAPSED: [Your estimate]
END_TIME: [Your estimate]
TIME_MARKERS: [Your list or None]
"""
        time_info = self.generate_text(prompt, system_prompt)
        parsed_info = {}
        if time_info:
            for line in time_info.strip().split('\n'):
                if ':' in line:
                    key, value = line.split(':', 1)
                    parsed_info[key.strip()] = value.strip()
            # Ensure all keys are present, even if generation failed partially
            final_info_str = (
                 f"TIME_ELAPSED: {parsed_info.get('TIME_ELAPSED', 'Unknown')}\n"
                 f"END_TIME: {parsed_info.get('END_TIME', 'Unknown')}\n"
                 f"TIME_MARKERS: {parsed_info.get('TIME_MARKERS', 'None')}"
             )
            self.timeline[chapter_num] = final_info_str
        else:
            print(f"Warning: Failed to generate timeline info for Chapter {chapter_num}.")
            fallback_info = "TIME_ELAPSED: Unknown\nEND_TIME: Unknown\nTIME_MARKERS: None"
            self.timeline[chapter_num] = fallback_info
        return self.timeline.get(chapter_num)


    def track_emotional_arc(self, chapter_num, chapter_content):
        # [ Existing method okay, prompt slightly refined ]
        if not chapter_content or "Generation Failed" in chapter_content: return None
        system_prompt = f"You are a literary analyst specializing in tracking emotional tone and narrative tension in {self.genre} stories."
        # Analyze a larger chunk for better context, but focus on the end
        content_snippet = chapter_content[-2000:]
        prompt = f"""Analyze the emotional state and narrative tension at the *very end* of this chapter ({chapter_num}) from a {self.genre} novel:

CHAPTER CONTENT SNIPPET (focus on the concluding paragraphs):
...{content_snippet}

Determine:
1.  EMOTION: What is the primary *reader* emotion evoked by the chapter's final sentences? (e.g., Suspense, Hope, Dread, Relief, Mystery, Anticipation)
2.  TENSION: Rate the narrative tension level at the end on a scale of 1 (Very Low) to 10 (Extreme Cliffhanger).
3.  UNRESOLVED: What is the main unresolved question, threat, or conflict left lingering to hook the reader into the next chapter?

Reply ONLY in this format:
EMOTION: [Primary Emotion]
TENSION: [Level 1-10]
UNRESOLVED: [Brief description of the hook]
"""
        emotional_status = self.generate_text(prompt, system_prompt)
        parsed_info = {}
        if emotional_status:
            for line in emotional_status.strip().split('\n'):
                 if ':' in line:
                    key, value = line.split(':', 1)
                    parsed_info[key.strip()] = value.strip()
            final_info_str = (
                f"EMOTION: {parsed_info.get('EMOTION', 'Unknown')}\n"
                f"TENSION: {parsed_info.get('TENSION', 'Unknown')}\n"
                f"UNRESOLVED: {parsed_info.get('UNRESOLVED', 'Unknown')}"
            )
            self.emotional_arc[chapter_num] = final_info_str
        else:
            print(f"Warning: Failed to generate emotional arc info for Chapter {chapter_num}.")
            fallback_info = "EMOTION: Unknown\nTENSION: Unknown\nUNRESOLVED: Unknown"
            self.emotional_arc[chapter_num] = fallback_info
        return self.emotional_arc.get(chapter_num)


    def create_chapter_transition(self, chapter_num, chapter_content):
        """Create a high-quality transition paragraph for the end of the chapter"""
        if chapter_num >= self.num_chapters or not chapter_content or "Generation Failed" in chapter_content:
            return ""

        system_prompt = f"""You are a master storyteller crafting compelling, literary transitions between chapters for the {self.genre} genre. Focus on seamless flow, mood, and creating anticipation."""

        # --- Extract Next Chapter Plan (More robust extraction) ---
        next_plan_extract_prompt = f"""From the detailed chapter plan below, extract ONLY the complete plan for Chapter {chapter_num + 1}. Include title, summary, scenes, etc.

COMPLETE CHAPTER PLAN:
{self.chapter_plan}

Find the section starting with '**Chapter {chapter_num + 1}:' and return everything belonging to that chapter until the next chapter marker (e.g., '**Chapter {chapter_num + 2}:') or the end of the text. If Chapter {chapter_num + 1} plan is not found, respond with 'PLAN_NOT_FOUND'.
"""
        next_chapter_plan = self.generate_text(next_plan_extract_prompt, "You are a text extraction utility.") # Simpler system prompt for extraction
        if not next_chapter_plan or "PLAN_NOT_FOUND" in next_chapter_plan:
            print(f"Warning: Could not extract plan for Chapter {chapter_num + 1} for transition generation.")
            next_chapter_plan = "[Next chapter plan not available - focus on current chapter's end state]"
        # ---

        emotional_status = self.emotional_arc.get(chapter_num, "EMOTION: Unknown\nTENSION: Unknown\nUNRESOLVED: Unknown")
        timeline_info = self.timeline.get(chapter_num, "END_TIME: Unknown")
        chosen_motif = ""
        if self.recurring_motifs:
            motif_index = (chapter_num - 1) % len(self.recurring_motifs)
            chosen_motif = self.recurring_motifs[motif_index]

        # --- Enhanced Transition Prompt ---
        prompt = f"""Craft a compelling closing paragraph (or two short ones, ~50-150 words total) to end Chapter {chapter_num} of this {self.genre} novel, creating a powerful hook into Chapter {chapter_num + 1}.

CURRENT CHAPTER ENDING CONTEXT (last ~1000 chars):
...{chapter_content[-1000:]}

NEXT CHAPTER ({chapter_num + 1}) PLAN/THEMES:
{next_chapter_plan}

EMOTIONAL STATE AT CURRENT CHAPTER END:
{emotional_status}

TIMELINE INFORMATION AT CURRENT CHAPTER END:
{timeline_info}

{f"RECURRING MOTIF TO WEAVE IN (Subtly, if appropriate): {chosen_motif}" if chosen_motif else ""}
{f"USER'S EXTRA NOTES TO CONSIDER: {self.extra_info}" if self.extra_info else ""}

The transition should:
1.  Provide a satisfying yet *incomplete* sense of closure for Chapter {chapter_num}.
2.  Amplify the end-of-chapter EMOTION and TENSION ({emotional_status}).
3.  Directly address or hint at the UNRESOLVED hook.
4.  Subtly foreshadow elements from the NEXT CHAPTER PLAN without revealing spoilers.
5.  Maintain the established {self.genre} tone and narrative style.
6.  Use strong imagery or evocative language.
7.  If possible, naturally incorporate the motif: '{chosen_motif}'.

Generate ONLY the 1-2 transition paragraphs. These will be the FINAL paragraphs of Chapter {chapter_num}. Do not add commentary. Write with literary quality.
"""

        transition = self.generate_text(prompt, system_prompt)
        if transition:
            # Clean up potential labels
            transition = re.sub(r"^(Transition|Paragraphs?|Response):?\s*", "", transition, flags=re.IGNORECASE).strip()
            self.transitions[chapter_num] = transition
            return transition
        else:
            print(f"Warning: Failed to generate transition for Chapter {chapter_num}. Using fallback.")
            # Improved fallback
            emotion_hint = emotional_status.split("EMOTION:")[1].split("\n")[0].strip() if "EMOTION:" in emotional_status else "uncertainty"
            fallback_transition = f"The air crackled with {emotion_hint}. What lingered most was the unanswered question, a shadow stretching towards the unknown events yet to unfold in {self.world_name}."
            self.transitions[chapter_num] = fallback_transition
            return fallback_transition


    def create_next_chapter_opener(self, chapter_num):
        """Create a high-quality opening for the next chapter"""
        if chapter_num <= 1: return ""

        system_prompt = f"""You are a master storyteller crafting engaging chapter openings for {self.genre} narratives. Ensure a seamless flow from the previous chapter and immediately ground the reader."""

        # --- Gather Context (More Robust) ---
        prev_chapter_num = chapter_num - 1
        prev_summary = self.chapter_summaries.get(prev_chapter_num, "[Previous summary unavailable]")
        # Get the actual transition text that *ended* the previous chapter
        prev_transition = self.transitions.get(prev_chapter_num, "[Previous transition text unavailable]")
        prev_emotional_status = self.emotional_arc.get(prev_chapter_num, "EMOTION: Unknown\nTENSION: Unknown\nUNRESOLVED: Unknown")
        prev_timeline = self.timeline.get(prev_chapter_num, "END_TIME: Unknown")

        prev_end_time = "Unknown"
        if prev_timeline and "END_TIME:" in prev_timeline:
            try:
                 prev_end_time = prev_timeline.split("END_TIME:")[1].split('\n')[0].strip()
            except IndexError: pass # Ignore malformed timeline

        # --- Extract This Chapter Plan ---
        plan_extract_prompt = f"""From the detailed chapter plan below, extract ONLY the complete plan for Chapter {chapter_num}. Include title, summary, scenes, etc.

COMPLETE CHAPTER PLAN:
{self.chapter_plan}

Find the section starting with '**Chapter {chapter_num}:' and return everything belonging to that chapter until the next chapter marker or the end. If Chapter {chapter_num} plan is not found, respond with 'PLAN_NOT_FOUND'.
"""
        this_chapter_plan = self.generate_text(plan_extract_prompt, "You are a text extraction utility.")
        if not this_chapter_plan or "PLAN_NOT_FOUND" in this_chapter_plan:
            print(f"Warning: Could not extract plan for Chapter {chapter_num} for opener generation.")
            this_chapter_plan = "[This chapter plan not available - rely on summary and previous ending]"
        # ---

        # --- Enhanced Opener Prompt ---
        prompt = f"""Craft a compelling opening paragraph (or two short ones, ~50-150 words total) for Chapter {chapter_num} of this {self.genre} novel. It must connect *seamlessly* to the end of Chapter {prev_chapter_num}.

END OF PREVIOUS CHAPTER ({prev_chapter_num}) - The final sentences were:
"{prev_transition}"

PREVIOUS CHAPTER ({prev_chapter_num}) SUMMARY (Context Only):
{prev_summary[:500]}...

EMOTIONAL STATE AT END OF PREVIOUS CHAPTER:
{prev_emotional_status}

TIME AT END OF PREVIOUS CHAPTER:
{prev_end_time}

THIS CHAPTER'S ({chapter_num}) PLAN/GOALS:
{this_chapter_plan}

{f"USER'S EXTRA NOTES TO CONSIDER: {self.extra_info}" if self.extra_info else ""}

The opening paragraph(s) MUST:
1.  Establish a clear time/setting connection. How much time has passed since "{prev_transition}"? Where are we now? (Use {prev_end_time} and the chapter plan for guidance).
2.  Flow naturally from the mood, tension, and unresolved questions of the previous chapter's ending. Acknowledge the lingering {prev_emotional_status}.
3.  Immediately orient the reader in the *first scene* of Chapter {chapter_num} (location, character POV/focus, initial action/observation) based on its plan.
4.  Use evocative language and sensory details appropriate for {self.genre}.
5.  Avoid info-dumping or summarizing. Assume the reader just finished the previous chapter.
6.  Set the tone for this chapter.

Generate ONLY the 1-2 opening paragraphs (without the chapter title). Write with literary quality.
"""

        opener = self.generate_text(prompt, system_prompt)
        if opener:
             # Clean up potential labels
            opener = re.sub(r"^(Opener|Opening Paragraphs?|Response):?\s*", "", opener, flags=re.IGNORECASE).strip()
            return opener
        else:
            print(f"Warning: Failed to generate opener for Chapter {chapter_num}. Using fallback.")
            time_hint = f"Shortly after the previous events" if prev_end_time == "Unknown" else f"Following the {prev_end_time}"
            return f"{time_hint}, the situation in {self.world_name} remained tense. The focus now shifted to..."


    def validate_chapter_consistency(self, chapter_num, chapter_content):
        # [ Existing method is okay, prompt slightly refined for clarity ]
        if not chapter_content or "Generation Failed" in chapter_content: return "CONSISTENT" # Can't validate failure message
        system_prompt = f"""You are a meticulous continuity editor for {self.genre} fiction. Your task is to identify specific inconsistencies between the Current Chapter and the Provided Context ONLY. Do not critique style or plot choices, only factual contradictions based on the context."""

        # --- Context Gathering (Keep it concise) ---
        context_chapter_limit = 3 # Reduced limit to focus on recent events and save tokens
        start_chapter = max(1, chapter_num - context_chapter_limit)
        previous_summaries = ""
        for i in range(start_chapter, chapter_num):
             summary = self.chapter_summaries.get(i)
             if summary and "[Summary generation failed]" not in summary:
                 # Limit summary length in context
                 previous_summaries += f"**Chapter {i} Summary (Brief):**\n{summary[:500]}...\n\n"

        # Character status context (only essential info)
        character_status_list = []
        for name, data in self.characters.items():
            # Include characters who *might* appear based on appearance before *or* in this chapter's potential plan
            # (We don't have the perfect future knowledge, but this is a guess)
             first_app = data.get("first_appearance", 0)
             # Include if appeared before OR if this is chapter 1 (all chars relevant) OR if first appearance could be this chapter
             if first_app < chapter_num or chapter_num == 1 or first_app == chapter_num or first_app == 0:
                  status = data.get("status", "unknown")
                  location = data.get("location", "unknown")
                  # Last *brief* dev note
                  last_dev = "N/A"
                  if data.get("development"):
                       last_dev = data["development"][-1].get("development", "N/A")[:80] # Truncate

                  character_status_list.append(f"- {name}: Status='{status}'; Loc='{location}'; LastDev='{last_dev}...'")

        character_status_str = "\n".join(character_status_list) if character_status_list else "No specific character status context."

        # Timeline context (just the end of the last chapter)
        timeline_context = "No recent timeline info."
        prev_chap_num = chapter_num - 1
        if prev_chap_num > 0 and prev_chap_num in self.timeline:
             timeline_context = f"**End of Ch {prev_chap_num} Timeline:**\n{self.timeline[prev_chap_num]}"

        # This Chapter's Plan (Extract snippet)
        plan_extract_prompt = f"""From the detailed chapter plan, extract the 'Summary' and 'Scene Breakdown' for Chapter {chapter_num}.

COMPLETE CHAPTER PLAN:
{self.chapter_plan}

Find the section for '**Chapter {chapter_num}:' and return ONLY the 'Summary' and 'Scene Breakdown' sections for it. If not found, respond 'PLAN_SNIPPET_NOT_FOUND'.
"""
        this_chapter_plan_snippet = self.generate_text(plan_extract_prompt, "You are a text extraction utility.")
        if not this_chapter_plan_snippet or "PLAN_SNIPPET_NOT_FOUND" in this_chapter_plan_snippet:
            this_chapter_plan_snippet = "[Chapter plan snippet unavailable for validation]"
        # ---

        prompt = f"""Analyze ONLY for factual inconsistencies between the 'CURRENT CHAPTER CONTENT' and the 'PROVIDED CONTEXT'.

PROVIDED CONTEXT:
---
Genre: {self.genre}
World Name: {self.world_name}
{f"User Notes: {self.extra_info}" if self.extra_info else ""}

**Previous Chapter Summaries (Recent):**
{previous_summaries if previous_summaries else "N/A"}

**Character Status Before This Chapter:**
{character_status_str}

**Timeline Context:**
{timeline_context}

**This Chapter's ({chapter_num}) Intended Plan (Summary & Scenes):**
{this_chapter_plan_snippet}
---

CURRENT CHAPTER ({chapter_num}) CONTENT TO VALIDATE:
{chapter_content}
---

Identify SPECIFIC contradictions related ONLY to:
1.  **Character Details:** Name misspellings, incorrect status/location/abilities compared to 'Character Status', actions contradicting established personality *based on context summaries*.
2.  **Setting/World:** Contradictory descriptions of '{self.world_name}' or locations mentioned in summaries/plan.
3.  **Timeline:** Events out of sequence, time jumps conflicting with 'Timeline Context'.
4.  **Plot Points:** Events contradicting 'Previous Chapter Summaries' or the 'Intended Plan' snippet.
5.  **Object/Motif Use:** Incorrect use/appearance of significant items mentioned in context.

List ONLY clear, factual inconsistencies found. Be precise.
If NO such inconsistencies are found based *strictly* on the provided context, respond ONLY with the exact word: CONSISTENT
"""
        consistency_check = self.generate_text(prompt, system_prompt)

        if not consistency_check:
            print(f"Warning: Consistency check for Chapter {chapter_num} failed to generate a response.")
            return "CHECK_FAILED"
        # Check if the response *only* contains CONSISTENT (case-insensitive, allowing whitespace)
        if consistency_check.strip().upper() == "CONSISTENT":
            return "CONSISTENT"
        else:
            # Return the list of issues, filtering out potential conversational filler
            lines = consistency_check.strip().split('\n')
            issues = [line for line in lines if line.strip() and not line.strip().startswith(("Here are", "I found", "Okay,")) ]
            return "\n".join(issues) if issues else "CHECK_FAILED" # Return issues or failure if filtering removed everything


    def fix_chapter_inconsistencies(self, chapter_num, chapter_content, issues):
        """Attempt to fix identified consistency issues seamlessly"""
        if not chapter_content or "Generation Failed" in chapter_content: return None
        # --- Enhanced System Prompt ---
        system_prompt = f"""You are a skilled {self.genre} fiction editor. Your task is to revise the provided chapter to eliminate specific continuity errors listed, while *preserving* the original narrative voice, style, tone, and core events as much as possible. Make changes seamlessly."""

        # --- Context Gathering (Similar to validation, maybe slightly more context) ---
        context_chapter_limit = 4 # Slightly more history might help fixes
        start_chapter = max(1, chapter_num - context_chapter_limit)
        previous_summaries = ""
        for i in range(start_chapter, chapter_num):
             summary = self.chapter_summaries.get(i)
             if summary and "[Summary generation failed]" not in summary:
                 previous_summaries += f"**Chapter {i} Summary:**\n{summary[:700]}...\n\n" # More summary context

        # Character status (same as validation)
        character_status_list = []
        for name, data in self.characters.items():
             first_app = data.get("first_appearance", 0)
             if first_app < chapter_num or chapter_num == 1 or first_app == chapter_num or first_app == 0:
                  status = data.get("status", "unknown")
                  location = data.get("location", "unknown")
                  last_dev = "N/A"
                  if data.get("development"): last_dev = data["development"][-1].get("development", "N/A")[:80]
                  character_status_list.append(f"- {name}: Status='{status}'; Loc='{location}'; LastDev='{last_dev}...'")
        character_status_str = "\n".join(character_status_list) if character_status_list else "N/A"

        timeline_context = "No recent timeline info."
        prev_chap_num = chapter_num - 1
        if prev_chap_num > 0 and prev_chap_num in self.timeline:
             timeline_context = f"**End of Ch {prev_chap_num} Timeline:**\n{self.timeline[prev_chap_num]}"

        # This Chapter's Plan (Full plan might be useful for fixing deviations)
        plan_extract_prompt = f"""From the detailed chapter plan, extract the complete plan for Chapter {chapter_num}.

COMPLETE CHAPTER PLAN:
{self.chapter_plan}

Find '**Chapter {chapter_num}:' and return its full entry. If not found, respond 'PLAN_NOT_FOUND'.
"""
        this_chapter_plan = self.generate_text(plan_extract_prompt, "You are a text extraction utility.")
        if not this_chapter_plan or "PLAN_NOT_FOUND" in this_chapter_plan:
            this_chapter_plan = "[Chapter plan unavailable for fix context]"
        # ---

        # --- Enhanced Fix Prompt ---
        prompt = f"""Revise Chapter {chapter_num} of this {self.genre} novel to correct *only* the specific 'CONSISTENCY ISSUES TO FIX' listed below, using the 'CONTEXT FOR FIXING' for accuracy.

CONTEXT FOR FIXING:
---
Genre: {self.genre}
World Name: {self.world_name}
{f"User Notes: {self.extra_info}" if self.extra_info else ""}
Previous Summaries: {previous_summaries if previous_summaries else "N/A"}
Character Status Before This Chapter: {character_status_str}
Timeline Context: {timeline_context}
This Chapter's Intended Plan: {this_chapter_plan}
---

ORIGINAL CHAPTER ({chapter_num}) CONTENT (with inconsistencies):
{chapter_content}
---

CONSISTENCY ISSUES TO FIX:
{issues}
---

REVISION GUIDELINES:
1.  Address EVERY specific issue listed under 'CONSISTENCY ISSUES TO FIX'.
2.  Make corrections seamlessly integrated into the narrative.
3.  Maintain the original chapter's essential plot, pacing, tone, style, and character voices. Do NOT rewrite unnecessarily.
4.  Ensure all details align with the 'CONTEXT FOR FIXING' (character status, timeline, world name, plan).
5.  Output the *complete*, revised chapter content, starting with the chapter title (e.g., "## Chapter {chapter_num}: [Original or Revised Title]").
6.  Do NOT add any commentary before or after the revised chapter.

Rewrite the chapter now:
"""
        print(f"Attempting to fix Chapter {chapter_num} via OpenRouter...")
        fixed_chapter = self.generate_text(prompt, system_prompt)

        if fixed_chapter and f"Chapter {chapter_num}" in fixed_chapter[:100]: # Basic check
            # Clean up potential introductory remarks from the LLM
            fixed_chapter = re.sub(r"^.*?(##\s*Chapter\s*\d+)", r"\1", fixed_chapter, flags=re.DOTALL | re.IGNORECASE)
            print(f"Chapter {chapter_num} revision generated.")
            # Optionally re-validate here (can be slow and costly)
            # print("Re-validating fixed chapter...")
            # re_check = self.validate_chapter_consistency(chapter_num, fixed_chapter)
            # if re_check == "CONSISTENT": print("Re-validation passed.")
            # else: print(f"Warning: Re-validation failed or found new issues: {re_check}")
            return fixed_chapter.strip()
        else:
            print(f"Error: Fixing Chapter {chapter_num} failed or produced invalid output.")
            if fixed_chapter: print(f"Output snippet: {fixed_chapter[:200]}...")
            return None


    def generate_chapter(self, chapter_num):
        """Generate a single chapter with enhanced context, quality focus, and consistency checks"""
        # --- Enhanced System Prompt ---
        system_prompt = f"""You are a celebrated {self.genre} novelist known for writing immersive, engaging, and stylistically strong chapters.
Your writing should feature:
-   Vivid sensory details and evocative descriptions.
-   Natural, character-revealing dialogue.
-   Varied sentence structures and compelling prose.
-   Deep exploration of character thoughts and emotions ('show, don't tell').
-   Seamless pacing and tension appropriate for the chapter's goals.
-   Perfect consistency with all provided context."""

        # --- Context Building (Similar to Fix/Validate, keep concise) ---
        context_chapter_limit = 3
        start_chapter = max(1, chapter_num - context_chapter_limit)
        context_summaries = []
        for i in range(start_chapter, chapter_num):
             summary = self.chapter_summaries.get(i)
             if summary and "[Summary generation failed]" not in summary:
                  context_summaries.append(f"**Chapter {i} Summary (Brief):**\n{summary[:500]}...\n")
        context = "\n".join(context_summaries) if context_summaries else "This is the first chapter."

        # Character context
        character_context_list = []
        # Include characters potentially relevant based on past or plan
        for name, data in self.characters.items():
             first_app = data.get("first_appearance", 0)
             if first_app < chapter_num or chapter_num == 1 or first_app == chapter_num or first_app == 0:
                  status = data.get("status", "unknown")
                  location = data.get("location", "unknown")
                  emotion = data.get("emotional_state", "unknown")
                  last_dev = "N/A"
                  if data.get("development"): last_dev = data["development"][-1].get("development", "N/A")[:80]
                  # Include description snippet for key chars if available
                  desc_snippet = (data.get("description", "")[:60] + "...") if len(data.get("description", "")) > 60 else data.get("description", "")
                  character_context_list.append(f"- {name}: Status({status}), Loc({location}), Emotion({emotion}). LastDev: {last_dev}. Desc: {desc_snippet}")
        characters_context_str = "\n".join(character_context_list) if character_context_list else "No specific character context."


        timeline_context = "N/A"
        prev_chap_num = chapter_num - 1
        if prev_chap_num > 0 and prev_chap_num in self.timeline:
             timeline_context = f"End of Ch {prev_chap_num} Timeline:\n{self.timeline[prev_chap_num]}"

        emotional_context = "N/A"
        if prev_chap_num > 0 and prev_chap_num in self.emotional_arc:
             emotional_context = f"End of Ch {prev_chap_num} Emotional State:\n{self.emotional_arc[prev_chap_num]}"

        # Create chapter opener (using the improved method)
        chapter_opener = ""
        if chapter_num > 1:
            print(f"Creating opener for Chapter {chapter_num}...")
            chapter_opener = self.create_next_chapter_opener(chapter_num) # Already prints warnings
            if not chapter_opener: chapter_opener = "[Start chapter directly]" # Fallback instruction

        # Extract This Chapter's Plan (Full plan needed for generation)
        plan_extract_prompt = f"""From the detailed chapter plan, extract the complete plan for Chapter {chapter_num}.

COMPLETE CHAPTER PLAN:
{self.chapter_plan}

Find '**Chapter {chapter_num}:' and return its full entry. If not found, respond 'PLAN_NOT_FOUND'.
"""
        this_chapter_plan = self.generate_text(plan_extract_prompt, "You are a text extraction utility.")
        if not this_chapter_plan or "PLAN_NOT_FOUND" in this_chapter_plan:
            print(f"Critical Warning: Could not extract plan for Chapter {chapter_num}. Generation quality will be severely impacted.")
            # Attempt to find *something* in the outline as a last resort
            outline_fallback_prompt = f"Extract any outline points specifically for Chapter {chapter_num} from the main story outline:\n{self.story_outline}"
            outline_fallback = self.generate_text(outline_fallback_prompt, "You are text extraction utility.")
            this_chapter_plan = f"[PLAN MISSING FOR CHAPTER {chapter_num}] - Improvise based on overall outline and previous summaries. Outline hints: {outline_fallback if outline_fallback else 'None available'}"


        # Motif instruction
        motif_instruction = ""
        if self.recurring_motifs:
            motif_index = (chapter_num - 1) % len(self.recurring_motifs)
            chosen_motif = self.recurring_motifs[motif_index]
            motif_instruction = f"**Recurring Motif:** Weave '{chosen_motif}' into this chapter naturally and meaningfully, relevant to the {self.genre} context and events."

        # --- Enhanced Generation Prompt ---
        prompt = f"""Write Chapter {chapter_num} of this {self.genre} novel with high literary quality, following all instructions and context precisely.

CONTEXT & GUIDELINES:
---
**Genre:** {self.genre}
**World Name:** {self.world_name} (Use consistently)
**Story Premise:** {self.story_premise}
{f"**User's Extra Notes:** {self.extra_info}" if self.extra_info else ""}

**This Chapter's ({chapter_num}) Detailed Plan:**
{this_chapter_plan}

**Character Context (Status before this chapter):**
{characters_context_str}

**Recent Previous Chapter Summaries:**
{context}

**Timeline Context:** {timeline_context}
**Emotional Context from Previous Chapter:** {emotional_context}

**Chapter Opener:** {chapter_opener} (Integrate this as the start unless '[Start chapter directly]')

{motif_instruction}
---

WRITING INSTRUCTIONS for Chapter {chapter_num}:
-   **Follow the Plan:** Adhere strictly to THIS CHAPTER'S DETAILED PLAN for scenes, plot points, and character development.
-   **Literary Quality:** Write engaging, immersive prose. Use vivid sensory details, varied sentence structures, and strong verbs. Show emotions and motivations through actions/dialogue/thoughts (Show, Don't Tell).
-   **Style & Tone:** Maintain a consistent {self.genre} tone and narrative voice appropriate for the story.
-   **Consistency:** Ensure 100% consistency with all provided CONTEXT (characters, world, timeline, summaries). Double-check names and details.
-   **Pacing:** Manage pacing according to the plan and chapter goals – build tension, allow reflection, deliver impact.
-   **Dialogue:** Write natural-sounding dialogue that reveals character and advances the plot.
-   **Word Count:** Aim for approximately 2500-4000 words, but prioritize fulfilling the chapter plan's requirements over hitting an exact count.
-   **Formatting:** Start DIRECTLY with the chapter title: "## Chapter {chapter_num}: [Chapter Title from Plan]". Use standard paragraph breaks (double newline) and dialogue formatting.
-   **Ending:** Conclude the chapter with a compelling hook based on the plan's 'Ending Hook', leading effectively into the next chapter.

Do NOT add any author notes or commentary outside the chapter content. Begin writing the chapter now:
"""
        print(f"Generating Chapter {chapter_num} via OpenRouter ({self.model})...")
        chapter_content = self.generate_text(prompt, system_prompt)

        # Basic validation and retry placeholder (can be expanded)
        if not chapter_content or not re.match(rf"##\s*Chapter\s*{chapter_num}", chapter_content.strip(), re.IGNORECASE):
            print(f"Error: Initial generation for Chapter {chapter_num} seems invalid or failed.")
            if chapter_content: print(f"Received snippet: {chapter_content[:200]}...")
            # Simple retry logic (optional)
            # print("Attempting regeneration...")
            # time.sleep(5) # Wait before retry
            # chapter_content = self.generate_text(prompt, system_prompt)
            # if not chapter_content or not re.match(rf"##\s*Chapter\s*{chapter_num}", chapter_content.strip(), re.IGNORECASE):
            #     print("Regeneration failed.")
            #     return f"## Chapter {chapter_num}: Generation Failed\n\n[Content generation failed for this chapter after retries]"
            return f"## Chapter {chapter_num}: Generation Failed\n\n[Initial content generation failed]"


        # --- Post-Generation Processing & Consistency Loop ---
        max_fix_attempts = 1 # Limit fixes to avoid loops/excessive cost
        for attempt in range(max_fix_attempts + 1):
            print(f"Validating Chapter {chapter_num} consistency (Attempt {attempt+1})...")
            # Ensure context for validation is updated *before* calling validate
            # (Context is gathered inside validate_chapter_consistency based on chapter_num)
            consistency_check = self.validate_chapter_consistency(chapter_num, chapter_content)

            if consistency_check == "CONSISTENT":
                print(f"Chapter {chapter_num} is consistent.")
                break # Exit loop
            elif consistency_check == "CHECK_FAILED":
                print(f"Consistency check failed for Chapter {chapter_num}. Proceeding with potentially flawed content.")
                break # Exit loop
            elif attempt < max_fix_attempts:
                print(f"Consistency issues found in Chapter {chapter_num}. Attempting fix ({attempt+1}/{max_fix_attempts})...")
                print(f"Issues:\n{consistency_check}")
                # Update context for fix *before* calling fix
                # (Context is gathered inside fix_chapter_inconsistencies)
                fixed_content = self.fix_chapter_inconsistencies(chapter_num, chapter_content, consistency_check)
                if fixed_content:
                    chapter_content = fixed_content # Update with fixed version
                    print(f"Chapter {chapter_num} fix attempt {attempt+1} successful.")
                    # Continue loop to re-validate the fixed content
                else:
                    print(f"Error: Failed to fix consistency issues for Chapter {chapter_num} on attempt {attempt+1}. Using potentially inconsistent content.")
                    break # Exit loop if fix fails
            else: # Max fix attempts reached
                print(f"Warning: Chapter {chapter_num} may retain inconsistencies after {max_fix_attempts} fix attempts. Proceeding with last version.")
                print(f"Final Issues Noted:\n{consistency_check}")
                break # Exit loop

        # --- Final Processing Steps (Run on the final version of chapter_content) ---
        print(f"Creating summary for Chapter {chapter_num}...")
        self.create_chapter_summary(chapter_num, chapter_content)

        print(f"Updating character tracking for Chapter {chapter_num}...")
        self.update_character_tracking(chapter_num, chapter_content) # Uses the potentially fixed content

        print(f"Updating timeline for Chapter {chapter_num}...")
        self.update_timeline(chapter_num, chapter_content)

        print(f"Analyzing emotional arc for Chapter {chapter_num}...")
        self.track_emotional_arc(chapter_num, chapter_content)

        # Add transition (using the potentially fixed content)
        if chapter_num < self.num_chapters:
            print(f"Creating transition for Chapter {chapter_num}...")
            transition = self.create_chapter_transition(chapter_num, chapter_content)
            if transition:
                # Append transition ensuring proper spacing
                chapter_content = chapter_content.strip() + '\n\n' + transition.strip()
            else:
                print(f"Warning: Failed to generate or append transition for Chapter {chapter_num}.")

        return chapter_content.strip()


    def check_chapter_transitions(self):
         """Review and potentially improve transitions between all generated chapters"""
         # [ Existing method is okay, prompt slightly enhanced ]
         if len(self.chapters) < 2:
             print("Not enough chapters to check transitions.")
             return

         print("\n--- Performing Final Review of Chapter Transitions ---")
         improved_chapters = [self.chapters[0]] # Start with the first chapter

         for i in range(1, len(self.chapters)):
             current_chapter_idx = i
             prev_chapter_idx = i - 1
             current_chapter_num = i + 1 # Chapter numbers are 1-based
             prev_chapter_num = i

             # Get the potentially already improved previous chapter
             prev_chapter_text = improved_chapters[prev_chapter_idx]
             current_chapter_text = self.chapters[current_chapter_idx] # Original current chapter

             # Skip if previous chapter failed or current chapter failed
             if not prev_chapter_text or "Generation Failed" in prev_chapter_text or not current_chapter_text or "Generation Failed" in current_chapter_text:
                  print(f"Skipping transition check for Chapter {current_chapter_num} due to generation failure in adjacent chapter.")
                  improved_chapters.append(current_chapter_text) # Keep current chapter as is
                  continue

             # Extract snippets for analysis
             # Use regex to find the actual end, avoiding potential appended transitions if logic failed before
             # Find last non-whitespace content before potential final transition
             prev_ending_match = re.search(r'([\s\S]*\S)\s*$', prev_chapter_text) # Find last non-whitespace char/content block
             prev_ending_snippet = ('...' + prev_ending_match.group(1)[-300:]) if prev_ending_match else '[Could not extract previous ending]'

             # Find the start of the current chapter after the title
             current_opening_match = re.match(r"##\s*Chapter\s*\d+[:\s\w\W]*?\n+([\s\S]*)", current_chapter_text)
             current_opening_snippet = (current_opening_match.group(1).strip()[:300] + '...') if current_opening_match else '[Could not extract current opening]'


             system_prompt = f"You are a {self.genre} fiction editor focused on narrative flow and smooth chapter transitions."
             prompt = f"""Analyze the transition between the end of Chapter {prev_chapter_num} and the beginning of Chapter {current_chapter_num}.

END OF PREVIOUS CHAPTER ({prev_chapter_num}) SNIPPET:
{prev_ending_snippet}

BEGINNING OF CURRENT CHAPTER ({current_chapter_num}) SNIPPET:
{current_opening_snippet}

Evaluate:
-   Flow: Is the transition smooth or jarring?
-   Connection: Is the link (time, theme, character focus) clear?
-   Tone: Is the shift in mood handled well?
-   Redundancy: Does the opening repeat the ending?

If the transition is already seamless and effective for {self.genre}, respond ONLY with the word: SMOOTH

If it can be improved, rewrite *only* the opening paragraph(s) of Chapter {current_chapter_num} (~1-2 paragraphs, 50-150 words) for a better connection. The revision should:
1.  Bridge directly from the previous chapter's final mood/event.
2.  Clearly establish the new time/setting/POV.
3.  Avoid repetition.
4.  Maintain {self.genre} style.

Start the response with "REVISED:" followed ONLY by the rewritten opening paragraph(s) for Chapter {current_chapter_num}. Do not include the chapter title or commentary.
"""

             transition_check_response = self.generate_text(prompt, system_prompt)

             if transition_check_response and transition_check_response.strip().upper() == "SMOOTH":
                 print(f"Transition to Chapter {current_chapter_num} verified as smooth.")
                 improved_chapters.append(current_chapter_text) # Keep original
             elif transition_check_response and transition_check_response.strip().startswith("REVISED:"):
                 try:
                     revised_opening_text = transition_check_response.split("REVISED:", 1)[1].strip()
                     if revised_opening_text:
                         # Find the title and content *after* the title in the original current chapter
                         title_match = re.match(r"(##\s*Chapter\s*\d+[:\s\w\W]*?\n+)", current_chapter_text)
                         if title_match:
                             chapter_title_part = title_match.group(1)
                             content_after_title = current_chapter_text[len(chapter_title_part):].strip()

                             # Estimate paragraphs to replace (heuristic: count paragraphs in revised text)
                             # Split by double newline, filter empty strings
                             revised_paras = [p for p in re.split(r'\n\s*\n', revised_opening_text) if p.strip()]
                             num_paras_to_replace = len(revised_paras)

                             original_paras_after_title = [p for p in re.split(r'\n\s*\n', content_after_title) if p.strip()]

                             # Replace the first N paragraphs of the original content
                             if len(original_paras_after_title) >= num_paras_to_replace:
                                  remaining_paras = original_paras_after_title[num_paras_to_replace:]
                                  remaining_content = '\n\n'.join(remaining_paras)
                                  # Combine: Title + Revised Opening + Remaining Content
                                  improved_chapter_text = chapter_title_part.strip() + '\n\n' + revised_opening_text + ('\n\n' + remaining_content if remaining_content else '')
                                  improved_chapters.append(improved_chapter_text.strip())
                                  print(f"Transition improved for Chapter {current_chapter_num}.")
                             else:
                                  # Fallback: Not enough original paragraphs, just prepend after title (less ideal)
                                  print(f"Warning: Replacing paragraphs for Chapter {current_chapter_num} transition failed (not enough original paras). Prepending.")
                                  improved_chapter_text = chapter_title_part.strip() + '\n\n' + revised_opening_text + '\n\n' + content_after_title
                                  improved_chapters.append(improved_chapter_text.strip())
                         else:
                             # Fallback: Couldn't find title, use original (safer than guessing)
                             print(f"Warning: Could not find title in Chapter {current_chapter_num} to apply revision. Using original.")
                             improved_chapters.append(current_chapter_text)
                     else:
                         # Response started with REVISED: but text was empty
                         print(f"Revision suggested for Chapter {current_chapter_num} but no text provided. Using original.")
                         improved_chapters.append(current_chapter_text)
                 except Exception as e:
                     print(f"Error processing revised transition for Chapter {current_chapter_num}: {e}. Using original.")
                     improved_chapters.append(current_chapter_text)
             else:
                 # Handle failure or unexpected response
                 print(f"Transition check failed or yielded unexpected response for Chapter {current_chapter_num}. Using original. Response: {transition_check_response[:100] if transition_check_response else 'None'}")
                 improved_chapters.append(current_chapter_text)

         self.chapters = improved_chapters # Update the main chapters list with improved versions
         print("--- Chapter Transitions Review Complete ---")


    def generate_book(self):
        """Generate the complete book with enhanced consistency and quality checks"""
        try:
            self.get_user_input()
            # Outline and Plan generation now raise errors on failure
            self.create_story_outline() # Includes character, world, motif extraction, and plan generation

            start_time = time.time()
            for i in range(1, self.num_chapters + 1):
                chapter_start_time = time.time()
                print(f"\n--- Generating Chapter {i} of {self.num_chapters} ---")
                # generate_chapter now includes validation, fixing, summary, tracking, transition
                chapter = self.generate_chapter(i)

                if "Generation Failed" in chapter:
                    print(f"Critical error: Generation of Chapter {i} failed. Aborting further generation.")
                    self.chapters.append(chapter) # Add the failure message
                    # Compile and return partial book with error note
                    return self.compile_book(partial=True)
                else:
                    self.chapters.append(chapter)
                    chapter_end_time = time.time()
                    print(f"--- Chapter {i} generation cycle completed in {chapter_end_time - chapter_start_time:.2f} seconds ---")

                # Optional delay
                if i < self.num_chapters: time.sleep(1) # Short delay

            # Final transition check across all chapters
            if len(self.chapters) > 1:
                 self.check_chapter_transitions()

            end_time = time.time()
            print(f"\n--- Book generation complete in {end_time - start_time:.2f} seconds ---")
            return self.compile_book()

        except (ValueError, RuntimeError, requests.exceptions.RequestException) as e:
            print(f"\n--- Book Generation Error ---")
            print(f"An error occurred during setup or generation: {e}")
            import traceback
            traceback.print_exc() # More detail for debugging
            print("Book generation cannot continue.")
            # Compile partial if possible
            if self.chapters:
                print("Compiling partial book content...")
                return self.compile_book(partial=True)
            else:
                return "Book generation failed during initial setup."


    def compile_book(self, partial=False):
        """Compile chapters into a book format, optionally adding a note if partial."""
        # --- Book Title Generation (Enhanced) ---
        book_title = f"Untitled {self.genre} Story" # Default
        if self.story_premise and self.story_outline:
            title_prompt = f"""Suggest 3-5 compelling, marketable titles for a {self.genre} book based on the following:

Genre: {self.genre}
Premise: {self.story_premise}
{f"User Notes: {self.extra_info}" if self.extra_info else ""}
Outline Snippet (Themes/Tone):
{self.story_outline[:1500]}...

Consider the target audience for {self.genre}. The titles should be evocative and intriguing.
Respond ONLY with a numbered list of titles, one per line.
"""
            title_system_prompt = f"You are an expert book marketer specializing in catchy titles for the {self.genre} genre."
            generated_titles_list = self.generate_text(title_prompt, title_system_prompt)

            if generated_titles_list:
                 # Extract the first title from the list as the primary choice
                 match = re.search(r"^\s*[1-9][.\)]?\s*(.*)", generated_titles_list, re.MULTILINE)
                 if match:
                     cleaned_title = match.group(1).strip().strip('"\'')
                     if cleaned_title and len(cleaned_title) > 2: # Basic sanity check
                         book_title = cleaned_title
                         print(f"Selected book title: {book_title}")
                     else:
                         print("Warning: Generated title format invalid. Using default.")
                 else:
                     # Fallback if numbering failed: take the first non-empty line
                      lines = [line.strip() for line in generated_titles_list.split('\n') if line.strip()]
                      if lines:
                           cleaned_title = lines[0].strip().strip('"\'')
                           if cleaned_title and len(cleaned_title) > 2:
                                book_title = cleaned_title
                                print(f"Selected book title (fallback extraction): {book_title}")
                           else:
                                print("Warning: Generated title format invalid (fallback). Using default.")
                      else:
                           print("Warning: Title generation response empty. Using default.")
            else:
                print("Warning: Failed to generate book title. Using default.")
        else:
            print("Skipping title generation (missing premise/outline).")

        self.final_book_title = book_title # Store the chosen title

        # --- Author Name ---
        author_name = "Generated by AI Narrator" # Placeholder

        # --- Compile Book Content ---
        book = f"# {self.final_book_title}\n\n"
        book += f"### By {author_name}\n\n"
        book += f"*Genre: {self.genre}*\n"
        if self.extra_info:
             book += f"*Notes: {self.extra_info}*\n"
        book += "---\n\n"

        if not self.chapters:
            book += "[No chapters were generated.]"
            if partial: book += "\n\n---\n\n**Note:** Book generation was interrupted early."
            return book.strip()

        for i, chapter_content in enumerate(self.chapters):
            chapter_num = i + 1
            if not isinstance(chapter_content, str):
                print(f"Warning: Chapter {chapter_num} content is not a string. Skipping.")
                book += f"## Chapter {chapter_num}: [Content Error]\n\n---\n\n"
                continue

            # Add chapter content, ensuring correct formatting and handling failures
            content_strip = chapter_content.strip()
            if re.match(rf"##\s*Chapter\s*{chapter_num}", content_strip, re.IGNORECASE):
                book += f"{content_strip}\n\n"
            elif f"Chapter {chapter_num}: Generation Failed" in content_strip:
                 book += f"{content_strip}\n\n" # Include failure message clearly
            else:
                # Attempt to find *any* title, otherwise add generic
                title_match = re.match(r"##\s*(.*?)\n", content_strip)
                if title_match:
                     print(f"Warning: Chapter {chapter_num} title format unusual. Using found title: '{title_match.group(1)}'")
                     book += f"{content_strip}\n\n"
                else:
                     print(f"Warning: Chapter {chapter_num} content missing standard title. Adding generic title.")
                     book += f"## Chapter {chapter_num}: [Untitled Chapter]\n\n{content_strip}\n\n"

            # Add separator
            if i < len(self.chapters) - 1:
                book += "---\n\n"

        # Add note if generation was stopped prematurely
        if partial:
             book += "\n\n---\n\n**Note:** Book generation was interrupted. Content may be incomplete or inconsistent."

        return book.strip()

    def save_book(self, book_content, filename_prefix="generated_book"):
        """Save the generated book and detailed metadata"""
        # Basic check if content seems valid before saving
        if not book_content or not self.chapters or "generation failed during initial setup" in book_content.lower():
            print("Book content seems invalid or generation failed very early. Skipping file save.")
            return

        safe_title = re.sub(r'[\\/*?:"<>|]', "", self.final_book_title).strip()
        safe_title = re.sub(r'\s+', '_', safe_title)
        safe_title = re.sub(r'_+', '_', safe_title)
        safe_genre = re.sub(r'\s+', '_', self.genre)
        timestamp = time.strftime("%Y%m%d_%H%M%S")
        filename_base = f"{filename_prefix}_{safe_title[:40]}_{safe_genre}_{timestamp}" # Slightly longer title part
        filename_md = f"{filename_base}.md"
        filename_json = f"{filename_base}_metadata.json"

        try:
            with open(filename_md, "w", encoding="utf-8") as f:
                f.write(book_content)
            print(f"Book saved as Markdown: {filename_md}")
        except IOError as e:
            print(f"Error saving book to {filename_md}: {e}")
            return # Stop if book saving fails

        # --- Prepare Detailed Metadata ---
        def stringify_keys(d):
             return {str(k): v for k, v in d.items()} if isinstance(d, dict) else d

        # Truncate potentially huge text fields for metadata file size
        outline_snippet = (self.story_outline[:5000] + "...") if self.story_outline and len(self.story_outline) > 5000 else self.story_outline
        plan_snippet = (self.chapter_plan[:5000] + "...") if self.chapter_plan and len(self.chapter_plan) > 5000 else self.chapter_plan

        metadata = {
            "generation_info": {
                 "model_used": self.model,
                 "generation_timestamp": time.strftime("%Y-%m-%d %H:%M:%S %Z"),
                 "app_title": self.app_title,
            },
            "book_details": {
                 "generated_title": self.final_book_title,
                 "genre": self.genre,
                 "world_name": self.world_name,
                 "num_chapters_requested": self.num_chapters,
                 "num_chapters_generated": len(self.chapters), # Actual count
                 "generation_successful": "Generation Failed" not in book_content and "generation cannot continue" not in book_content.lower()
            },
            "user_input": {
                "story_premise": self.story_premise,
                "extra_info": self.extra_info
            },
            "narrative_elements": {
                 "characters": self.characters, # Save final state
                 "recurring_motifs_themes": self.recurring_motifs,
            },
            "generation_artifacts": {
                 # Use stringify_keys for dictionaries with integer keys
                 "chapter_summaries": stringify_keys(self.chapter_summaries),
                 "timeline_tracking": stringify_keys(self.timeline),
                 "emotional_arc_tracking": stringify_keys(self.emotional_arc),
                 "chapter_transitions_generated": stringify_keys(self.transitions),
                 "story_outline_snippet": outline_snippet if outline_snippet else "N/A",
                 "chapter_plan_snippet": plan_snippet if plan_snippet else "N/A",
            }
        }

        try:
            with open(filename_json, "w", encoding="utf-8") as f:
                # Use default=str for safety against non-serializable types
                json.dump(metadata, f, indent=4, default=str)
            print(f"Book metadata saved as JSON: {filename_json}")
        except IOError as e:
            print(f"Error saving metadata to {filename_json}: {e}")
        except TypeError as e:
            print(f"Error serializing metadata to JSON: {e}. Some data might not be fully saved.")


if __name__ == "__main__":
    try:
        generator = BookGenerator() # Checks API key

        # --- Optional: Customize settings ---
        # generator.model = "openai/gpt-4o" # Example: Try a more powerful model if available/affordable
        # ---

        book_text_content = generator.generate_book()

        # Save only if generation produced something potentially useful
        if book_text_content and "generation failed during initial setup" not in book_text_content.lower():
            generator.save_book(book_text_content, filename_prefix="book")
        else:
            print("\nBook generation failed early or produced no usable content. No files saved.")

    except ValueError as e: # Catch API key error
        print(f"Configuration Error: {e}")
    except RuntimeError as e: # Catch critical outline/plan errors
        print(f"Generation Error: {e}")
    except Exception as e: # Catch other unexpected errors
        print(f"An unexpected error occurred: {e}")
        import traceback
        traceback.print_exc()
```
